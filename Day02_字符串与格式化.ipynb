#### 索引(index) - 取出第几个字符

> **痛点**:字符串 "Python" 想取出第 1 个字符 P,怎么做?
> **类比**:索引就像座位号 - 从 0 开始编号。
> **解释**:`字符串[索引]` 取出对应位置的字符。

> **ASCII 内存图**

```
字符串:  P  y  t  h  o  n
正向索引: 0  1  2  3  4  5
反向索引:-6 -5 -4 -3 -2 -1
```

````python
# 正向索引(从 0 开始)s = "Python"print(s[0])   # Pprint(s[1])   # yprint(s[5])   # n# 反向索引(从 -1 开始)print(s[-1])  # nprint(s[-2])  # o
````

````python
# --- 执行过程 ---# 第 2 行 s[0]:#   ① 取出索引 0 的字符 -> 'P'#   ② print 输出: P## 第 5 行 s[5]:#   ① 取出索引 5(最后 1 个) -> 'n'## 第 8 行 s[-1]:#   ① 反向索引,-1 是倒数第 1 个 -> 'n'#   ② 不需要计算长度,s[-1] 永远是最后一个print(s[0])print(s[-1])
````

> **逐行解剖**
- `s[0]` 是第 1 个字符,**不是第 0 个**
- `s[-1]` 负数索引 = 从右往左,-1 是最后 1 个
- 索引超出范围 -> IndexError

> **常见错误**
1. 索引从 1 开始:`s[1]` 是第 2 个,不是第 1 个
2. 索引越界:`s = "Hi"; s[5]` -> IndexError
3. 忘方括号:`s(0)` -> TypeError

> 问自己:
> - 字符串长度为 6,合法索引范围?
> - s[0] 和 s[-6] 是同一个字符吗?
> - 怎么取出最后一个字符?

#### 切片(slice) - 取出一段子串

> **痛点**:想取出 "apple" 中的 "app",怎么做?
> **类比**:切片 = 用刀切一段。`s[开始:结束]`,结束本身不包含。
> **解释**:`s[start:end]` 取出 start 到 end-1 的字符。

> **ASCII 内存图:切片原理**

```
字符串:  a  p  p  l  e
索引:    0  1  2  3  4

s[1:4]  ->  p  p  l    (含 1,不含 4)
s[:3]   ->  a  p  p    (从头开始)
s[2:]   ->  p  l  e    (到末尾)
```

````python
s = "apple"# 基础切片print(s[1:4])   # pplprint(s[:3])    # appprint(s[2:])    # ple# 常用技巧phone = "13812345678"print(phone[:3])          # 138print(phone[-4:])         # 5678print(phone[3:7])         # 1234
````

````python
# --- 执行过程 ---# 第 4 行 s[1:4]:#   ① 取索引 1 到 3(不含 4)#   ② 'p','p','l' -> "ppl"## 第 5 行 s[:3]:#   ① 省略 start,默认从 0 开始#   ② 取 0 到 2 -> "app"## 第 7 行 s[2:]:#   ① 省略 end,默认到末尾#   ② 取 2 到末 -> "ple"s = "apple"print(s[1:4])print(s[:3])print(s[2:])
````

> **逐行解剖**
- `s[start:end]`:含 start,不含 end(左闭右开)
- `s[:end]`:省略 start,从头开始
- `s[start:]`:省略 end,到末尾
- 切片不会越界报错(自动截断)

> **常见错误**
1. 以为含结束:s[1:4] 取第 1,2,3,不含第 4
2. start > end:s[4:1] -> 空字符串
3. 忘冒号:s[1,4] -> TypeError

> 问自己:
> - s[2:2] 取出什么?s[3:1]?
> - 怎么取前 3 个?后 3 个?
> - s[:] 取出什么?有什么用?

#### 切片进阶 - 步长与反转

> **解释**
`s[start:end:step]` 三个参数:step 是步长。`s[::-1]` 反转字符串。

````python
s = "abcdefghij"# 步长:每隔 step-1 个取一个print(s[::2])     # acegiprint(s[1::2])    # bdfhjprint(s[::3])     # adgj# 反转print(s[::-1])    # jihgfedcba
````

````python
# --- 执行过程 ---# 第 4 行 s[::2]:#   ① step=2,每隔 1 个取 1 个#   ② 索引 0,2,4,6,8 -> 'a','c','e','g','i'## 第 7 行 s[::-1]:#   ① step=-1,反向走#   ② 从末尾往头部取 -> 反转s = "abcdefghij"print(s[::2])print(s[::-1])
````

> **常见错误**
1. 步长理解错:s[::2] 是"隔 1 取 1",不是"取 2 个"

#### f-string - 字符串"填空"

> **痛点**:想输出"小明今年 18 岁",原来的写法:
`print("姓名:" + name + ",年龄:" + str(age))`
加号和引号满天飞,要转字符串。

> **类比**:f-string = 在字符串里留"填空位"`f"姓名:{name},年龄:{age}"`

> **解释**:引号前加 f,用 {} 包起变量,自动替换。

````python
name = "小明"age = 18height = 1.75# f-string 基本用法print(f"姓名:{name}")print(f"姓名:{name},年龄:{age}")# 表达式和格式化print(f"明年{age + 1}岁")print(f"身高{height:.1f}米")
````

````python
# --- 执行过程 ---# 第 6 行 print(f"姓名:{name},年龄:{age}"):#   ① Python 看到 f"..." 前缀,这是 f-string#   ② 扫描字符串,找到 {name} 占位符#   ③ 读取变量 name 的值 -> "小明"#   ④ 找到 {age},读取 age -> 18#   ⑤ 替换后 -> "姓名:小明,年龄:18"#   ⑥ print 输出name = "小明"age = 18print(f"姓名:{name},年龄:{age}")
````

> **逐行解剖**
- `f"..."`:前缀 f 表示格式化字符串
- `{}` 里放变量名、表达式、函数调用
- `{height:.1f}` 格式说明:保留 1 位小数

> **常见错误**
1. 忘加 f:`"姓名:{name}"` → 输出字面量
2. 括号不匹配:`f"姓名:{name"` → SyntaxError
3. 想输出字面量 {:`f"{{}}"`(双花括号)

> 问自己:

> - 怎么保留 2 位小数?

#### 常用方法 - 查找/替换/清洗/分割

> **解释**:字符串有很多内置方法,用 `.方法名()` 调用。

````python
# 查找s = "apple pie"print(s.find('p'))       # 1print(s.find('z'))       # -1(找不到)print(s.rfind('p'))      # 7print(s.count('p'))      # 3# 替换print(s.replace('p', 'P'))  # aPPle Pieprint(s.replace(' ', '_'))  # apple_pie# 清洗print("  hello  ".strip())   # helloprint("  hello  ".lstrip())  # hello(左)print("HELLO".lower())       # hello# 分割和连接csv = "苹果,香蕉,橙子"fruits = csv.split(',')print(fruits)print(','.join(fruits))
````

````python
# --- 执行过程 ---# 第 2 行 s.find('p'):#   ① 从左往右搜索 'p'#   ② 索引 1 找到 -> 返回 1## 第 4 行 s.find('z'):#   ① 没找到 -> 返回 -1(不是报错!)## 第 9 行 s.replace('p','P'):#   ① 创建新字符串,把 'p' 改成 'P'#   ② 原字符串 s 不变(不可变!)#   ③ 返回新字符串 'aPPle Pie's = "apple pie"print(s.find('p'))print(s.find('z'))print(s.replace('p', 'P'))
````

> **逐行解剖**
- find():找不到返回 -1
- replace():返回新字符串,不修改原字符串
- split():返回列表,join()连接列表

> **常见错误**
1. 以为 find 找不到报错:实际返回 -1
2. 以为 replace 修改原字符串:实际不修改
3. split 忘传参:s.split() 按空白分割

> 问自己:
> - find 找不到时返回什么?
> - replace 后原字符串变了吗?

#### 字符串不可变 - 不能原地修改

> **解释**:字符串一旦创建,不能通过索引修改单个字符。

````python
s = "hello"try:    s[0] = "H"except TypeError as e:    print(f"报错: {e}")# 正确做法:创建新字符串s1 = "H" + s[1:]print(s1)print(s)  # 原字符串没变
````

> **常见错误**
1. 想通过索引修改:s[0] = 'A' -> TypeError
2. 以为 replace 修改原字符串

> 问自己:
> - 为什么字符串要设计成不可变的?
> - 怎么"修改"字符串的某个字符?

> 怎么"修改":用切片 + 拼接(s = s[:i] + '新字符' + s[i+1:])
