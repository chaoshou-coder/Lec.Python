**第 1 讲:while 循环**

语法:**while 条件:** —— 只要条件为 True,就持续执行循环体。口诀:**while = "当...时就继续"**。

三要素:**起始值、终止条件、递增步进**,缺一不可,否则就是死循环。

```python
# 打印 1 到 5
count = 1                # 1. 起始值
while count <= 5:        # 2. 终止条件
    print(count)
    count += 1            # 3. 递增步进
print("循环结束")
```

---

**第 2 讲:死循环与 break 退出**

忘记写递增步进,或者条件永远为 True,就会形成 **死循环**。

刻意构造死循环 **while True:** 时,配合 **break** 主动退出 —— 这是菜单类程序的常见骨架。

```python
# 用 break 退出死循环 —— 输入 "quit" 才结束
while True:
    word = input("输入 quit 退出:")
    if word == "quit":
        break           # 立刻跳出整个循环
print("你退出了")
```

---

**第 3 讲:for 循环 —— 遍历序列**

**for item in 序列:** 依次取出序列中的每个元素,对每个元素执行一次循环体。最适合"已知长度、要处理每一个元素"的场景。

```python
# 遍历字符串:每次取一个字符
for ch in "Python":
    print(ch)           # P y t h o n 各一行
```

```python
# 遍历列表:每次取一个元素
fruits = ["苹果", "香蕉", "橘子"]
for fruit in fruits:
    print(f"我喜欢吃 {fruit}")
```

---

**第 4 讲:range() —— 生成整数数列**

**range(start, stop, step)** 生成从 start 开始(含)到 stop 结束(不含)、步长为 step 的整数序列。

没有现成的循环对象时,用 range 配合 for 循环最方便。

```python
# range(stop):从 0 到 stop-1
for i in range(5):
    print(i)            # 0 1 2 3 4
```

```python
# range(start, stop):从 start 到 stop-1
for i in range(1, 6):
    print(i)            # 1 2 3 4 5
```

```python
# range(start, stop, step):按步长跳着取
for i in range(0, 10, 2):
    print(i)            # 0 2 4 6 8
```

```python
# step 为负数:倒着数
for i in range(100, 0, -2):
    print(i)            # 100 98 96 ... 2
```

---

**第 5 讲:累加器与累乘器**

**累加器**:初始值为 0,循环中不断 `+=`。口诀:**加 0 乘 1,初始值不要搞反**。

累乘时初始值必须是 1(写成 0 结果永远为 0)。

```python
# 求 1 + 2 + ... + 100 的和
total = 0                  # 累加初始值为 0
for i in range(1, 101):
    total += i             # 累加
print(total)               # 5050
```

```python
# 求 1 到 9 所有奇数的乘积
product = 1                # 累乘初始值为 1(不是 0!)
for i in range(1, 10, 2):
    product *= i           # 累乘
print(product)             # 945
```

---

**第 6 讲:break 与 continue**

**break** 彻底结束整个循环(辞职);**continue** 只跳过本次,继续下一趟(请假)。

```python
# break:遇到 5 直接退出,后续不再执行
for i in range(10):
    if i == 5:
        break
    print(i)               # 输出 0 1 2 3 4
```

```python
# continue:偶数跳过不打印
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)               # 输出 1 3 5 7 9
```

---

**第 7 讲:嵌套循环 —— 九九乘法表**

外层循环控制行,内层循环控制列。配合 **end=""** 让内层不换行,外层 print() 负责换行。

```python
# 九九乘法表
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j}*{i}={i*j}", end="\t")   # \t 制表符对齐,不换行
    print()                                  # 每行结束换行
```

**今日小结**

今天你把代码从"顺序执行"升级到了"**循环**":
- **while** 当条件成立就继续,三要素:起始/终止/步进
- **for** 遍历序列中的每个元素
- **range** 生成整数数列,左闭右开
- **累加器/累乘器** 初始值:加 0 乘 1
- **break** 彻底退出;**continue** 跳过本次
- **嵌套循环** 外层管行,内层管列

> 易错点:**while** 缺少步进会变死循环;累乘初值写成 0 结果恒 0;**range** 第二个参数不含,写倒着数时要注意 stop 值。
