# Pandas 进阶

在 Pandas 基础之上,进一步掌握**多条件过滤、query、分组聚合、表合并、拼接、透视表、缺失值处理**等进阶操作,这是数据分析从"看数"到"问数"的关键一步。

## 布尔过滤 多条件加括号

多条件过滤时,**每个条件必须用括号包裹**,再用 **& 且** 或 **| 或** 连接,否则会因运算符优先级导致报错。

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "age": [18, 20, 19, 21, 22],
    "score": [88, 92, 75, 85, 90]
})

# 单条件:分数大于 80
print(df[df["score"] > 80])
```

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "age": [18, 20, 19, 21, 22],
    "score": [88, 92, 75, 85, 90]
})

# 多条件且:分数 > 80 且 年龄 >= 20,每个条件必须加括号
print(df[(df["score"] > 80) & (df["age"] >= 20)])

# 多条件或:分数 > 90 或 年龄 < 19
print(df[(df["score"] > 90) | (df["age"] < 19)])
```

## query 方法引用外部变量

**query** 可以把过滤条件写成字符串,更简洁。使用 **@变量名** 可以引用外部变量。

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "age": [18, 20, 19, 21, 22],
    "score": [88, 92, 75, 85, 90]
})

# query 写法:条件写成字符串,更直观
print(df.query("score > 80 and age >= 20"))
```

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "age": [18, 20, 19, 21, 22],
    "score": [88, 92, 75, 85, 90]
})

# 定义外部变量
min_score = 80
min_age = 20

# 使用 @变量名 引用外部变量,不用拼接字符串
print(df.query("score > @min_score and age >= @min_age"))
```

## sort_values 排序

**sort_values** 支持单列、多列排序,通过 **ascending** 控制升降序,多列排序传入列表。

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "age": [18, 20, 19, 21, 22],
    "score": [88, 92, 75, 85, 90]
})

# 按分数升序排序
print(df.sort_values("score"))

# 按分数降序排序
print(df.sort_values("score", ascending=False))
```

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "age": [18, 20, 19, 21, 22],
    "score": [88, 92, 75, 85, 90]
})

# 多列排序:先按年龄升序,年龄相同按分数降序
print(df.sort_values(["age", "score"], ascending=[True, False]))
```

## groupby 分组聚合

**groupby** 把数据按某列分组后做聚合运算,核心思想是**拆分、应用、合并**。

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "class": ["A", "A", "B", "B", "B"],
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "score": [88, 92, 75, 85, 90]
})

# 按班级分组,计算每个班的平均分
print(df.groupby("class")["score"].mean())
```

## agg 多种聚合

**agg** 可以一次性做多种聚合运算,支持传入**函数列表**或使用**命名聚合**自定义列名。

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "class": ["A", "A", "B", "B", "B"],
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "score": [88, 92, 75, 85, 90]
})

# agg + 列表:一次算多个指标,自动命名列名
print(df.groupby("class")["score"].agg(["mean", "max", "min", "count"]))
```

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "class": ["A", "A", "B", "B", "B"],
    "name": ["张三", "李四", "王五", "赵六", "孙七"],
    "score": [88, 92, 75, 85, 90]
})

# 命名聚合:自定义输出列名,格式为 新列名=("原列名", "函数名")
print(df.groupby("class").agg(
    avg=("score", "mean"),
    max=("score", "max"),
    count=("score", "count")
))
```

## merge 表合并

**merge** 类似于 SQL 的 JOIN,根据关联字段将两张表合并。**how** 参数控制连接方式:inner、left、right、outer。

```python
# 导入 pandas
import pandas as pd

# 创建两张示例表,通过 id 关联
df_stu = pd.DataFrame({
    "id": ["S1", "S2", "S3", "S4"],
    "name": ["张三", "李四", "王五", "赵六"]
})
df_score = pd.DataFrame({
    "id": ["S1", "S2", "S3", "S5"],
    "score": [88, 92, 75, 80]
})

# inner 合并:只保留两边都有的键
print(pd.merge(df_stu, df_score, on="id", how="inner"))
```

```python
# 导入 pandas
import pandas as pd

# 创建两张示例表,通过 id 关联
df_stu = pd.DataFrame({
    "id": ["S1", "S2", "S3", "S4"],
    "name": ["张三", "李四", "王五", "赵六"]
})
df_score = pd.DataFrame({
    "id": ["S1", "S2", "S3", "S5"],
    "score": [88, 92, 75, 80]
})

# left 合并:以左表为主,右表匹配不到填 NaN
print(pd.merge(df_stu, df_score, on="id", how="left"))
```

```python
# 导入 pandas
import pandas as pd

# 创建两张示例表,通过 id 关联
df_stu = pd.DataFrame({
    "id": ["S1", "S2", "S3", "S4"],
    "name": ["张三", "李四", "王五", "赵六"]
})
df_score = pd.DataFrame({
    "id": ["S1", "S2", "S3", "S5"],
    "score": [88, 92, 75, 80]
})

# right 合并:以右表为主,左表匹配不到填 NaN
print(pd.merge(df_stu, df_score, on="id", how="right"))

# outer 合并:取两边键的并集,匹配不到填 NaN
print(pd.merge(df_stu, df_score, on="id", how="outer"))
```

## concat 拼接

**concat** 沿指定轴拼接多张表。**axis=0** 纵向堆叠行,**axis=1** 横向拼接列。

```python
# 导入 pandas
import pandas as pd

# 创建两个示例 DataFrame
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"A": [7, 8], "B": [9, 10]})

# 纵向拼接:堆叠行,ignore_index 重置索引为连续整数
print(pd.concat([df1, df2], axis=0, ignore_index=True))
```

```python
# 导入 pandas
import pandas as pd

# 创建两个示例 DataFrame
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"C": [7, 8], "D": [9, 10]})

# 横向拼接:按列拼接,行索引对齐
print(pd.concat([df1, df2], axis=1))
```

## pivot_table 透视表

**pivot_table** 把长表变成宽交叉表,通过 **index、columns、values、aggfunc** 指定行键、列键、值列、聚合函数。

```python
# 导入 pandas
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "day": ["周一", "周一", "周二", "周二"],
    "city": ["北京", "上海", "北京", "上海"],
    "sales": [120, 85, 96, 110]
})

# 透视表:行为 day,列为 city,值为 sales,聚合函数为求和
pt = df.pivot_table(
    index="day",
    columns="city",
    values="sales",
    aggfunc="sum"
)
print(pt)
```

## 缺失值处理

缺失值是数据分析常见问题。处理流程:**发现**(isnull().sum)、**删除**(dropna)、**填充**(fillna)。

```python
# 导入 pandas 和 numpy(NaN 来自 numpy)
import pandas as pd
import numpy as np

# 创建含缺失值的示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六"],
    "age": [18, np.nan, 19, np.nan],
    "score": [88, 92, np.nan, 85]
})

# isnull().sum() 统计每列缺失值数量
print(df.isnull().sum())
```

```python
# 导入 pandas 和 numpy
import pandas as pd
import numpy as np

# 创建含缺失值的示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六"],
    "age": [18, np.nan, 19, np.nan],
    "score": [88, 92, np.nan, 85]
})

# dropna() 删除含缺失值的行
print(df.dropna())
```

```python
# 导入 pandas 和 numpy
import pandas as pd
import numpy as np

# 创建含缺失值的示例 DataFrame
df = pd.DataFrame({
    "name": ["张三", "李四", "王五", "赵六"],
    "age": [18, np.nan, 19, np.nan],
    "score": [88, 92, np.nan, 85]
})

# fillna(常数):用固定值填充所有缺失值
print(df.fillna(0))

# fillna(均值):用该列的均值填充缺失值
print(df.fillna(df.mean(numeric_only=True)))

# ffill:用前一个有效值前向填充
print(df.fillna(method="ffill"))
```
