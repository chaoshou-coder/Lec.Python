# 异常值检测：IQR 箱线图与 3σ 可视化

IQR = Q3 - Q1，超过 [Q1 - 1.5×IQR, Q3 + 1.5×IQR] 视为异常值。对于近似正态的列，可用 3σ 原则：超出 [均值 ± 3×标准差] 视为异常值。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["Arial Unicode MS", "SimHei"]
plt.rcParams["axes.unicode_minus"] = False

data = {"姓名": ["张三", "李四", "王五", "赵六", "孙七", "周八", "吴九", "郑十", "钱十一", "陈十二", "林十三", "黄十四", "杨十五", "朱十六", "秦十七", "尤十八", "许十九", "何二十", "吕二一", "施二二", "张二三", "孔二四", "曹二五", "严二六"], "部门": ["销售", "技术", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场"], "薪资": [8000, 15000, 13000, 9000, 11000, 7500, null, 8500, 10000, 7000, 16000, 8800, null, 7200, 14000, 9200, 10500, 6800, 15500, null, 10800, 120000, 13500, 8700], "年龄": [25, 32, 29, null, 35, 28, 31, 27, null, 26, 33, 30, 40, 24, null, 29, 38, 27, 34, 28, 36, 31, 30, null], "绩效评分": [78, 92, 88, 65, 74, 82, 95, 70, 80, 83, 86, 68, 90, 77, 91, 66, 72, 84, 89, 55, 81, 79, 87, 73], "入职年限": [1, 5, 4, 2, 6, 3, 7, 2, 5, 1, 8, 3, 10, 1, 6, 2, 9, 2, 7, 3, 8, 4, 5, 2]}
df = pd.DataFrame(data)

# IQR 箱线图
df[["薪资"]].boxplot()
plt.title("薪资箱线图(IQR 异常值检测)")
plt.ylabel("薪资(元)")
plt.show()

s = df["薪资"].dropna()
q1 = s.quantile(0.25)
q3 = s.quantile(0.75)
iqr = q3 - q1
lower_iqr = q1 - 1.5 * iqr
upper_iqr = q3 + 1.5 * iqr
print(f"IQR 正常范围: [{lower_iqr}, {upper_iqr}]")
outliers_iqr = df[(df["薪资"] < lower_iqr) | (df["薪资"] > upper_iqr)]
print(f"IQR 异常值:\n{outliers_iqr[["姓名", "薪资"]]}\n")

# 3σ 方法
mu = df["薪资"].mean()
sigma = df["薪资"].std()
lower_3s = mu - 3 * sigma
upper_3s = mu + 3 * sigma
print(f"3σ 范围: [{lower_3s:.0f}, {upper_3s:.0f}]")

df["薪资"].dropna().hist(bins=10, edgecolor="black", color="skyblue")
plt.axvline(lower_3s, color="red", linestyle="--", label="下界")
plt.axvline(upper_3s, color="red", linestyle="--", label="上界")
plt.title("薪资分布与 3σ 异常边界")
plt.xlabel("薪资(元)")
plt.ylabel("频数")
plt.legend()
plt.show()
```

# 类型修正与新增特征

把浮点年龄转成整数；通过现有列派生新特征，例如按薪资划分等级。

```python
import pandas as pd
import numpy as np

data = {"姓名": ["张三", "李四", "王五", "赵六", "孙七", "周八", "吴九", "郑十", "钱十一", "陈十二", "林十三", "黄十四", "杨十五", "朱十六", "秦十七", "尤十八", "许十九", "何二十", "吕二一", "施二二", "张二三", "孔二四", "曹二五", "严二六"], "部门": ["销售", "技术", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场"], "薪资": [8000, 15000, 13000, 9000, 11000, 7500, null, 8500, 10000, 7000, 16000, 8800, null, 7200, 14000, 9200, 10500, 6800, 15500, null, 10800, 120000, 13500, 8700], "年龄": [25, 32, 29, null, 35, 28, 31, 27, null, 26, 33, 30, 40, 24, null, 29, 38, 27, 34, 28, 36, 31, 30, null], "绩效评分": [78, 92, 88, 65, 74, 82, 95, 70, 80, 83, 86, 68, 90, 77, 91, 66, 72, 84, 89, 55, 81, 79, 87, 73], "入职年限": [1, 5, 4, 2, 6, 3, 7, 2, 5, 1, 8, 3, 10, 1, 6, 2, 9, 2, 7, 3, 8, 4, 5, 2]}
df = pd.DataFrame(data)
df["薪资"] = df["薪资"].fillna(df["薪资"].median())
df["年龄"] = df["年龄"].fillna(df["年龄"].median())

# 类型修正：年龄转整数
df["年龄"] = df["年龄"].astype(int)

# 新增特征：薪资等级
def salary_level(x):
    if x >= 15000:
        return "高"
    elif x >= 10000:
        return "中"
    else:
        return "低"

df["薪资等级"] = df["薪资"].apply(salary_level)

# 新增特征：绩效等级
def perf_level(x):
    if x >= 90:
        return "优秀"
    elif x >= 75:
        return "良好"
    else:
        return "待提升"

df["绩效等级"] = df["绩效评分"].apply(perf_level)

print("类型修正与新增特征后的数据:")
print(df.head(10))
```

# 分组聚合：groupby + agg

groupby("部门") 按部门分组后聚合。agg() 可同时应用多个聚合函数，并用 tuples 指定每列聚合方式。

```python
import pandas as pd

data = {"姓名": ["张三", "李四", "王五", "赵六", "孙七", "周八", "吴九", "郑十", "钱十一", "陈十二", "林十三", "黄十四", "杨十五", "朱十六", "秦十七", "尤十八", "许十九", "何二十", "吕二一", "施二二", "张二三", "孔二四", "曹二五", "严二六"], "部门": ["销售", "技术", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场"], "薪资": [8000, 15000, 13000, 9000, 11000, 7500, 12000, 8500, 10000, 7000, 16000, 8800, 9500, 7200, 14000, 9200, 10500, 6800, 15500, 11500, 10800, 10800, 13500, 8700], "年龄": [25, 32, 29, 30, 35, 28, 31, 27, 33, 26, 33, 30, 40, 24, 33, 29, 38, 27, 34, 28, 36, 31, 30, 32], "绩效评分": [78, 92, 88, 65, 74, 82, 95, 70, 80, 83, 86, 68, 90, 77, 91, 66, 72, 84, 89, 55, 81, 79, 87, 73], "入职年限": [1, 5, 4, 2, 6, 3, 7, 2, 5, 1, 8, 3, 10, 1, 6, 2, 9, 2, 7, 3, 8, 4, 5, 2]}
df = pd.DataFrame(data)

# 按部门分组计算薪资均值
dept_salary_mean = df.groupby("部门")["薪资"].mean()
print("各部门平均薪资:")
print(dept_salary_mean.round(0))

# 按部门分组统计人数
dept_count = df.groupby("部门")["姓名"].count()
print("\n各部门人数:")
print(dept_count)

# agg 命名聚合
result = df.groupby("部门").agg(
    薪资均值=("薪资", "mean"),
    薪资最高=("薪资", "max"),
    绩效均值=("绩效评分", "mean"),
    人数=("姓名", "count"),
)
print("\n各部门多维度聚合结果:")
print(result.round(1))
```

# 排序、筛选与量化洞察

sort_values 按列排序；query 用字符串表达式筛选行。最终用具体数字输出 3 条量化发现。

```python
import pandas as pd

data = {"姓名": ["张三", "李四", "王五", "赵六", "孙七", "周八", "吴九", "郑十", "钱十一", "陈十二", "林十三", "黄十四", "杨十五", "朱十六", "秦十七", "尤十八", "许十九", "何二十", "吕二一", "施二二", "张二三", "孔二四", "曹二五", "严二六"], "部门": ["销售", "技术", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场", "财务", "销售", "技术", "市场"], "薪资": [8000, 15000, 13000, 9000, 11000, 7500, 12000, 8500, 10000, 7000, 16000, 8800, 9500, 7200, 14000, 9200, 10500, 6800, 15500, 11500, 10800, 10800, 13500, 8700], "年龄": [25, 32, 29, 30, 35, 28, 31, 27, 33, 26, 33, 30, 40, 24, 33, 29, 38, 27, 34, 28, 36, 31, 30, 32], "绩效评分": [78, 92, 88, 65, 74, 82, 95, 70, 80, 83, 86, 68, 90, 77, 91, 66, 72, 84, 89, 55, 81, 79, 87, 73], "入职年限": [1, 5, 4, 2, 6, 3, 7, 2, 5, 1, 8, 3, 10, 1, 6, 2, 9, 2, 7, 3, 8, 4, 5, 2]}
df = pd.DataFrame(data)

# 按薪资降序排列
df_sorted = df.sort_values(by="薪资", ascending=False)
print("按薪资降序(前 5):")
print(df_sorted.head())

# query 筛选
r1 = df.query('部门 == "技术" and 薪资 > 13000')
print("\n技术部门薪资 > 13000:")
print(r1)

# === 量化洞察 ===
dept_salary = df.groupby("部门")["薪资"].mean()
top_dept = dept_salary.idxmax()
top_val = dept_salary.max()
bottom_dept = dept_salary.idxmin()
bottom_val = dept_salary.min()
print(f"发现 1：{top_dept} 部门平均薪资最高，为 {top_val:.0f} 元，"
      f"是 {bottom_dept} 部门（{bottom_val:.0f} 元）的 {top_val / bottom_val:.1f} 倍。")

corr = df["绩效评分"].corr(df["入职年限"])
relation = "正" if corr > 0 else "负"
strength = "较强" if abs(corr) > 0.3 else "较弱"
print(f"发现 2：绩效评分与入职年限的相关系数为 {corr:.2f}，"
      f"两者{relation}相关关系{strength}。")

high_ratio = (df["薪资"] >= 14000).mean()
high_count = int(high_ratio * len(df))
print(f"发现 3：薪资不低于 14000 元的员工占比 {high_ratio:.0%}，共 {high_count} 人。")
```
