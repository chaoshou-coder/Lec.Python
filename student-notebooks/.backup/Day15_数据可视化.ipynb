# Day15 数据可视化

**matplotlib** 是 Python 最基础的绘图库,可以绘制折线图、柱状图、散点图、直方图、饼图等常见图表。每个 cell 独立 import、独立运行、数据硬编码。

```
# 导入 matplotlib 的 pyplot 模块,别名为 plt
import matplotlib.pyplot as plt

# 设置中文字体为 SimHei,解决中文显示为方框的问题
plt.rcParams["font.sans-serif"] = ["SimHei"]
# 解决负号显示为方框的问题
plt.rcParams["axes.unicode_minus"] = False

# 硬编码数据:x 为月份列表,y 为销售额列表
x = ["1月", "2月", "3月", "4月", "5月"]
y = [120, 150, 170, 160, 200]

# 绘制折线图
plt.plot(x, y)
# 显示图表
plt.show()
```

**柱状图** 适合对比不同类别的数据。使用 `bar()` 绘制,通过 `title`、`xlabel`、`ylabel` 设置标题与坐标轴标签。

```
# 导入 pyplot
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 类别数据与对应数值
categories = ["苹果", "香蕉", "橙子", "葡萄", "西瓜"]
values = [30, 25, 40, 35, 50]

# 绘制柱状图
plt.bar(categories, values)
# 设置图表标题
plt.title("水果销量")
# 设置 x 轴标签
plt.xlabel("水果种类")
# 设置 y 轴标签
plt.ylabel("销量(吨)")
# 显示图表
plt.show()
```

**散点图** 用于观察两个变量之间的相关性。使用 `scatter()` 绘制,每个点由 (x, y) 坐标决定。

```
# 导入 pyplot
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 学习时长与考试分数数据
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
scores = [50, 55, 65, 70, 75, 80, 88, 92]

# 绘制散点图
plt.scatter(study_hours, scores)
# 设置标题与标签
plt.title("学习时长与分数关系")
plt.xlabel("学习时长(小时)")
plt.ylabel("考试分数")
# 显示图表
plt.show()
```

**直方图** 用于观察数据的分布情况。使用 `hist()` 将数据分成若干区间,统计每个区间的频数。

```
# 导入 pyplot
import matplotlib.pyplot as plt
# 导入 random 用于生成随机数据
import random

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 固定随机种子,保证每次运行结果一致
random.seed(42)
# 生成 100 个 40~100 的随机成绩
grades = [random.randint(40, 100) for _ in range(100)]

# 绘制直方图,bins=6 表示分为 6 个区间
plt.hist(grades, bins=6)
plt.title("成绩分布直方图")
plt.xlabel("分数区间")
plt.ylabel("人数")
# 显示图表
plt.show()
```

**饼图** 用于展示各部分占整体的比例。使用 `pie()` 绘制,`autopct` 显示百分比。

```
# 导入 pyplot
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 各平台用户占比数据
platforms = ["安卓", "iOS", "Windows", "Mac"]
ratios = [45, 30, 20, 5]

# 绘制饼图,autopct 显示百分比保留 1 位小数
plt.pie(ratios, labels=platforms, autopct="%.1f%%")
plt.title("操作系统市场份额")
# 显示图表
plt.show()
```

**图例、颜色、线型、标记** 可以丰富图表信息。`legend()` 显示图例,`color` 设置颜色,`linestyle` 设置线型,`marker` 设置数据点标记。

```
# 导入 pyplot
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 两个城市 1-5 月平均气温
months = [1, 2, 3, 4, 5]
beijing = [0, 3, 10, 18, 25]
shanghai = [5, 8, 13, 20, 27]

# 绘制北京折线图,红色虚线,圆形标记
plt.plot(
    months, beijing,
    color="red", linestyle="--", marker="o",
    label="北京"
)
# 绘制上海折线图,蓝色实线,方形标记
plt.plot(
    months, shanghai,
    color="blue", linestyle="-", marker="s",
    label="上海"
)
# 显示图例,内容由每次 plot 的 label 参数提供
plt.legend()
plt.title("城市月均气温对比")
plt.xlabel("月份")
plt.ylabel("温度(℃)")
# 显示图表
plt.show()
```

**subplots** 可在一个画布上创建多个子图。`plt.subplots(2, 2)` 创建 2 行 2 列共 4 个子图,`tight_layout()` 自动调整间距,`savefig()` 保存图片。

```
# 导入 pyplot
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 创建 2x2 子图,fig 为画布,axes 为二维子图数组
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# 第一个子图:折线图
axes[0, 0].plot([1, 2, 3], [1, 4, 9])
axes[0, 0].set_title("折线图")

# 第二个子图:柱状图
axes[0, 1].bar(["A", "B", "C"], [3, 7, 2])
axes[0, 1].set_title("柱状图")

# 第三个子图:散点图
axes[1, 0].scatter([1, 2, 3, 4], [2, 3, 5, 7])
axes[1, 0].set_title("散点图")

# 第四个子图:饼图
axes[1, 1].pie([30, 40, 30], labels=["甲", "乙", "丙"])
axes[1, 1].set_title("饼图")

# 自动调整子图间距,避免重叠
plt.tight_layout()

# 保存图片到文件,dpi 控制分辨率,bbox_inches 裁剪空白
plt.savefig("subplots_demo.png", dpi=100, bbox_inches="tight")
# 显示图表
plt.show()
```

**seaborn** 是基于 matplotlib 的高级绘图库,提供更美观的统计图表。以下演示箱线图与热力图。

```
# 导入 seaborn,别名为 sns
import seaborn as sns
# 导入 pyplot 用于显示
import matplotlib.pyplot as plt
# 导入 pandas 构造数据
import pandas as pd

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 构造三组成绩数据
data = pd.DataFrame({
    "班级": ["一班"] * 5 + ["二班"] * 5 + ["三班"] * 5,
    "分数": [80, 85, 90, 75, 88,
             70, 78, 82, 65, 75,
             90, 92, 95, 88, 93]
})

# 绘制箱线图:x 为班级,y 为分数
sns.boxplot(x="班级", y="分数", data=data)
plt.title("各班级成绩箱线图")
# 显示图表
plt.show()
```

**热力图 heatmap** 常用于展示相关系数矩阵。`annot=True` 在格子中显示数值,`cmap` 控制颜色映射。

```
# 导入 seaborn
import seaborn as sns
# 导入 pyplot
import matplotlib.pyplot as plt
# 导入 pandas 构造数据
import pandas as pd

# 设置中文字体
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 构造相关系数矩阵数据
df = pd.DataFrame({
    "语文": [1.0, 0.8, 0.3],
    "数学": [0.8, 1.0, 0.5],
    "英语": [0.3, 0.5, 1.0]
}, index=["语文", "数学", "英语"])

# 绘制热力图,annot=True 显示数值,cmap 使用蓝色系配色
sns.heatmap(df, annot=True, cmap="Blues")
plt.title("学科相关性热力图")
# 显示图表
plt.show()
```
