#### 为什么需要抽象（定义规范，隐藏实现）

**痛点：为什么需要它**  
多态只管'有没有方法'，但团队项目里没法保证每个子类真的都实现了；调试到运行时才发现某个子类漏写，BUG 难追溯。

**类比：生活例子**  
国家发布 USB-C 接口规范：每个厂商的设备必须有 type-c 口；用户只管插，不管内部怎么实现。

**解释：它是什么**  
抽象类 = 契约/规范，**不能实例化**，只定义子类**必须实现**的方法；至于怎么实现，子类自己决定。

```python
import abc

# 抽象类：不能直接实例化
class Shape(abc.ABC):
    @abc.abstractmethod
    def area(self):
        # 子类必须实现这个方法
        ...

    @abc.abstractmethod
    def perimeter(self):
        ...

# 尝试实例化会报错
try:
    s = Shape()
except TypeError as e:
    print('报错:', e)
```

```python
# --- 执行过程 ---
# ① import abc: 引入抽象基类模块
# ② class Shape(abc.ABC):
#    Shape 继承自 abc.ABC，成为抽象基类
# ③ @abc.abstractmethod 装饰 area 方法
#    -> area 变成抽象方法，没有具体实现
# ④ @abc.abstractmethod 装饰 perimeter 方法
#    -> perimeter 也变成抽象方法
# ⑤ 尝试 Shape()
#    -> Python 发现 area 和 perimeter 是抽象方法
#    -> 抛出 TypeError，不允许实例化
# ⑥ 被 try/except 捕获，打印错误信息
```

**逐行解剖**

- `abc.ABC` 作为父类，让当前类成为抽象基类。
- `@abc.abstractmethod` 装饰的方法**没有函数体**（用 `...` 占位）。
- 抽象类**不能实例化**，否则报 TypeError。
- 想使用，必须创建**子类并实现所有抽象方法**。

#### abc.ABC 抽象基类 + @abstractmethod

**痛点：为什么需要它**  
抽象方法只有声明没有实现，子类必须一一实现，否则就报错；这套机制靠 abc 模块实现。

**类比：生活例子**  
合同写明'乙方必须提供发票'（抽象方法），至于发票开什么内容，乙方自己定。

**解释：它是什么**  
- abc.ABC 让类拥有'抽象'属性。
- @abstractmethod 让方法变成强制规范。
- 子类**必须实现所有抽象方法**，才能被实例化。

```python
import abc

class Payment(abc.ABC):
    @abc.abstractmethod
    def pay(self, amount):
        ...

class Alipay(Payment):
    # 实现了抽象方法
    def pay(self, amount):
        print(f'支付宝支付 {amount} 元')

class WeChatPay(Payment):
    # 也实现了抽象方法，但实现不同
    def pay(self, amount):
        print(f'微信支付 {amount} 元')

alipay = Alipay()
wechat = WeChatPay()

alipay.pay(100)   # 支付宝支付 100 元
wechat.pay(200)   # 微信支付 200 元
```

```python
# --- 执行过程 ---
# ① import abc
# ② class Payment(abc.ABC): 抽象基类
# ③ @abc.abstractmethod + def pay(...): 抽象方法
# ④ class Alipay(Payment):
#    实现 pay -> 覆盖抽象方法
# ⑤ class WeChatPay(Payment):
#    实现 pay -> 覆盖抽象方法
# ⑥ alipay = Alipay() -> 所有抽象方法已实现，允许实例化
# ⑦ wechat = WeChatPay() -> 同理
# ⑧ alipay.pay(100) -> 调用 Alipay.pay，输出'支付宝支付'
# ⑨ wechat.pay(200) -> 调用 WeChatPay.pay，输出'微信支付'
```

**逐行解剖**

- 抽象基类中可以同时存在抽象方法和普通方法。
- @abstractmethod 下面可以是 `...`、pass、或带 raise NotImplementedError。
- 继承后**每个**抽象方法都必须实现，漏一个都不行。

```python
# ============ 参考答案 ============
import abc

class Notifier(abc.ABC):
    @abc.abstractmethod
    def send(self, msg):
        ...

    @abc.abstractmethod
    def channel(self):
        ...

class EmailNotifier(Notifier):
    def send(self, msg):
        print(f'邮件: {msg}')

    def channel(self):
        return 'email'

class SMSNotifier(Notifier):
    def send(self, msg):
        print(f'短信: {msg}')

    def channel(self):
        return 'sms'

e = EmailNotifier()
s = SMSNotifier()
e.send('订单已发货')
print(e.channel())
s.send('验证码 1234')
print(s.channel())
```

#### 接口概念（子类必须实现）

**痛点：为什么需要它**  
多人协作写插件系统，各人写各人的插件，如果没有统一规范，后期集成发现方法名不一致、参数不匹配。

**类比：生活例子**  
公司面试流程：简历、笔试、面试、Offer 每轮必须做，但每轮具体题目每个公司不同。

**解释：它是什么**  
接口 = 只包含抽象方法的抽象类，**纯粹定义规范**；实现接口 = 子类负责填完每个方法体。

```python
import abc

class Plugin(abc.ABC):
    # 纯接口：所有方法都是抽象的
    @abc.abstractmethod
    def name(self):
        ...

    @abc.abstractmethod
    def run(self, data):
        ...

class UpperPlugin(Plugin):
    def name(self):
        return 'upper'

    def run(self, data):
        return data.upper()

class ReversePlugin(Plugin):
    def name(self):
        return 'reverse'

    def run(self, data):
        return data[::-1]

plugins = [UpperPlugin(), ReversePlugin()]
text = 'hello world'

for p in plugins:
    print(f'插件 {p.name()}: {p.run(text)}')
```

```python
# --- 执行过程 ---
# ① class Plugin(abc.ABC): 纯接口，两个抽象方法
# ② UpperPlugin 实现 name 和 run
# ③ ReversePlugin 实现 name 和 run
# ④ plugins = [UpperPlugin(), ReversePlugin()]
#    -> 创建两个插件对象放入列表
# ⑤ text = 'hello world'
# ⑥ for p in plugins:
#    -> 第一次 p = UpperPlugin()
#    -> p.name() -> 'upper'
#    -> p.run('hello world') -> data.upper() -> 'HELLO WORLD'
#    -> 第二次 p = ReversePlugin()
#    -> p.name() -> 'reverse'
#    -> p.run('hello world') -> 'dlrow olleh'
```

**逐行解剖**

- 接口类只有抽象方法，没有普通方法和属性。
- 子类必须**全部实现**，否则仍是抽象类，也无法实例化。
- Python 没有 interface 关键字，用**全抽象方法的 ABC** 模拟。

```python
# ============ 学员代码区 ============
import abc

class Serializer(abc.ABC):
    # 定义 serialize 抽象方法
    pass

# class JsonSerializer(Serializer): ...
pass
```

```python
# ============ 参考答案 ============
import abc

class Serializer(abc.ABC):
    @abc.abstractmethod
    def serialize(self, obj):
        ...

    @abc.abstractmethod
    def deserialize(self, text):
        ...

class JsonSerializer(Serializer):
    def serialize(self, obj):
        # 模拟 JSON 格式
        return "{" + f"data: {obj}" + "}"

    def deserialize(self, text):
        return text

sers = [JsonSerializer()]
obj = "hello"

for s in sers:
    txt = s.serialize(obj)
    print('序列化:', txt)
    print('反序列化:', s.deserialize(txt))
```

#### 常见魔术方法（__len__ / __getitem__ / __add__）

**痛点：为什么需要它**  
自写的类想用 len()、[]、+ 这些内置操作，直接报错；靠原生语法才能写出 Pythonic 代码。

**类比：生活例子**  
电视机装好 HDMI 口才能接电脑，**魔术方法**就是类与 Python 内置函数之间的接口。

**解释：它是什么**  
双下划线开头和结尾的方法，Python 在特定场景下**自动调用**；例如 len(obj) 自动执行 obj.__len__()，a + b 自动执行 a.__add__(b)。

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        # len(obj) 自动调用
        return len(self.members)

    def __getitem__(self, index):
        # obj[index] 自动调用
        return self.members[index]

    def __add__(self, other):
        # obj1 + obj2 自动调用
        return Team(self.members + other.members)

t1 = Team(['张三', '李四'])
t2 = Team(['王五'])

# len() 触发 __len__
print(len(t1))        # 2

# [] 触发 __getitem__
print(t1[0])          # 张三
print(t1[1])          # 李四

# + 触发 __add__
t3 = t1 + t2
print(len(t3))        # 3
print(t3[2])          # 王五
```

```python
# --- 执行过程 ---
# ① class Team: 定义类
# ② __init__: members 绑定列表
# ③ __len__: 注册到类，处理 len(obj) 请求
# ④ __getitem__: 注册到类，处理 obj[index]
# ⑤ __add__: 注册到类，处理 obj1 + obj2
# ⑥ t1 = Team(['张三', '李四']) -> 绑定属性
# ⑦ t2 = Team(['王五']) -> 绑定属性
# ⑧ len(t1)
#    -> Python 自动调用 t1.__len__()
#    -> 返回 len(['张三', '李四']) -> 2
# ⑨ t1[0]
#    -> Python 自动调用 t1.__getitem__(0)
#    -> return ['张三', '李四'][0] -> '张三'
# ⑩ t1 + t2
#    -> Python 自动调用 t1.__add__(t2)
#    -> 拼接 members，返回新 Team
# ⑪ len(t3) -> 3
# ⑫ t3[2] -> '王五'
```

**逐行解剖**

- __len__ 必须返回整数，被 len() 调用。
- __getitem__ 让对象能被**索引甚至切片**。
- __add__ 让对象支持 + 运算符，应返回**新对象**而非修改原对象。
- 常见魔术方法还有 __str__、__eq__、__contains__ 等。

#### 综合练习：把抽象四个知识点串起来

设计一个插件系统：
1. 抽象类 Filter：抽象方法 name()、apply(text)
2. 子类 UpperFilter：apply 返回 text.upper()
3. 子类 ReverseFilter：apply 返回 text[::-1]
4. 子类 TrimFilter：apply 返回 text.strip()
5. 所有 Filter 实现 __len__，返回 name 长度
6. 写函数 run_filters(filters, text)，循环调用每个插件的 apply
7. 用 len() 测每个插件 name 长度

> 问自己:
> - 抽象类在这一整套系统里扮演了什么角色?
> - run_filters 函数体现了哪个特性?

```python
# ============ 学员代码区 ============
import abc

class Filter(abc.ABC):
    # 定义 name()、apply(text) 抽象方法
    # 实现 __len__，返回 name 长度
    pass

# class UpperFilter(Filter): ...
# class ReverseFilter(Filter): ...
# class TrimFilter(Filter): ...
# def run_filters(filters, text): ...
pass
```

```python
# ============ 参考答案 ============
import abc

class Filter(abc.ABC):
    @abc.abstractmethod
    def name(self):
        ...

    @abc.abstractmethod
    def apply(self, text):
        ...

    def __len__(self):
        # 魔术方法：返回名字长度
        return len(self.name())

class UpperFilter(Filter):
    def name(self):
        return 'upper'

    def apply(self, text):
        return text.upper()

class ReverseFilter(Filter):
    def name(self):
        return 'reverse'

    def apply(self, text):
        return text[::-1]

class TrimFilter(Filter):
    def name(self):
        return 'trim'

    def apply(self, text):
        return text.strip()

def run_filters(filters, text):
    # 多态：循环调用每个插件的 apply
    for f in filters:
        print(f'{f.name()}: {f.apply(text)}')

filters = [UpperFilter(), ReverseFilter(), TrimFilter()]
text = '  Hello World  '
run_filters(filters, text)

for f in filters:
    # len() 自动调用 __len__
    print(f'{f.name()} 名字长度: {len(f)}')
```
