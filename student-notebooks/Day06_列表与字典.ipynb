#### 知识点 1: 列表创建与索引切片

**痛点**: 如果你有 100 个名字,只用变量存就要写 100 行,一旦要整组操作(比如找最大值)就很麻烦。

**类比**: 列表就像一排抽屉,每个槽位有编号(从 0 开始),可以按号取货。

**解释**: 列表(list)是 Python 内置的有序集合,用方括号 `[]` 标记,元素用逗号分隔。
通过**索引**(`lst[i]`)或**切片**(`lst[start:end]`)访问元素。

```python
# 创建列表
fruits = ["苹果", "香蕉", "橙子", "菠萝"]
print("列表:", fruits)
print("长度:", len(fruits))

# 索引访问: 从 0 开始计数
print(fruits[0])   # 第 1 个元素
print(fruits[1])   # 第 2 个元素
print(fruits[-1])  # 最后 1 个元素(负向索引)

# 切片: 取出一段子列表
print(fruits[1:3])  # 取索引 1,2, 不包含 3
print(fruits[:2])   # 开始到索引 2(不含)
print(fruits[2:])   # 索引 2 到结尾
print(fruits[::2])  # 步长 2, 每隔一个取一个
print(fruits[::-1]) # 逆序
```

**--- 执行过程 ---**

① 第 2 行: 创建一个含 4 个字符串的列表赋给 fruits
② 第 5 行: fruits[0] 取索引 0 的结果是 '苹果'
③ 第 7 行: fruits[-1] 负向索引 结果是最后一个元素
④ 第 10 行: fruits[1:3] 返回 [香蕉, 橙子](左闭右开)
⑤ 第 13 行: fruits[::2] 步长为 2 取元素
⑥ 第 14 行: fruits[::-1] 直接逆序返回新列表

**逐行解剖**

- `[]`: 创建列表的语法符
- `[i]`: 正向索引从 0 开始,负数从右往左数(-1 是最后一个)
- `[start:end]`: 左闭区间,不包含 end
- `[:end]` / `[start:]`: 省略一端表示到头/到尾
- `[::step]`: 步长,为负数时表示反向

#### 知识点 2: 列表的可变性

**痛点**: 字符串是不可变的,改一个字符就要重新生成一个新对象。
列表可以直接修改内容,更高效。

**类比**: 字符串像已印好的书,想改就要重印一本;
列表像白板,随时在上面改加删。

**解释**: 赋值语句 `lst[i] = x` 会直接修改原列表,
不会创建新对象,这就是"可变(mutable)"。

```python
# 修改列表元素
colors = ["红", "绿", "蓝"]
print("修改前:", colors)

colors[1] = "黄"   # 直接替换索引 1 的元素
print("修改后:", colors)

# 对比字符串(不可变)
s = "abc"
# s[0] = "x"  # 这行会报 TypeError, 证明字符串不支持索引赋值
```

**--- 执行过程 ---**

① 第 2 行: 创建列表 [红, 绿, 蓝]
② 第 5 行: colors[1] = 黄 将索引 1 原元素替换
③ 第 6 行: 输出修改后的列表
④ 第 10 行: 字符串不支持这种操作, 注释掉了

**逐行解剖**

- `lst[i] = x`: 找到索引 i 位置, 把值改成 x
- 特点: 原列表被直接修改, 没有新对象产生
- 字符串是不可变类型, 不支持这种赋值

#### 知识点 3: 列表 CRUD - 增删查

**痛点**: 操作列表时,添加/删除/排序都要自己写复杂的循环。

**类比**: 列表就像一个自动调整的购物车,
把东西推进去它会自动扩大,把东西拉出来它会自动缩小。

**解释**: 列表内置了一系列方法来管理元素:
添加(append/insert)、删除(pop/remove)、排序(sort/reverse)、查找(index/count)。

```python
# 列表 CRUD 操作(每一步用新列表,互相不影响)

# --- 增: append / insert ---
basket = ["苹果", "香蕉"]
basket.append("橙子")
print("append 后:", basket)
basket.insert(1, "菠萝")
print("insert 后:", basket)

# --- 删: pop ---
basket2 = ["苹果", "香蕉", "橙子"]
item = basket2.pop()
print("弹出:", item, "剩余:", basket2)

# --- 删: remove ---
basket3 = ["苹果", "香蕉", "橙子"]
basket3.remove("香蕉")
print("remove 后:", basket3)

# --- 排序与反转 ---
nums = [3, 1, 4, 1, 5]
nums.sort()
print("sort 后:", nums)
nums.reverse()
print("reverse 后:", nums)
```

**--- 执行过程 ---**

① 第 3 行: 初始列表 [苹果, 香蕉]
② 第 4 行: append(橙子) 添加到尾部
③ 第 6 行: insert(1, 菠萝) 在索引 1 插入,原元素右移
④ 第 11 行: pop() 弹出并返回最后一个元素
⑤ 第 16 行: remove 找到香蕉并删除
⑥ 第 20-25 行: sort 升序排序, reverse 反转

**逐行解剖**

- `append(x)`: 在末尾添加一个元素
- `insert(i, x)`: 在索引 i 插入 x, 原有元素后移
- `pop(i)`: 删除索引 i 并返回它, 只写 pop() 就删除最后一个
- `remove(x)`: 删除第一个等于 x 元素, 不存在会报 ValueError
- `sort()`: 小到大排序(源列表被修改, 返回 None!)
- `reverse()`: 反转列表

#### 知识点 4: 字典创建与取值

**痛点**: 列表只能用数字索引,要"找某人电话"时记不住排第几号。

**类比**: 字典就像电话簿,通过姓名(key)查电话(value),而不用记序号。

**解释**: 字典(dict)是 key-value 对的集合。
用花括号 `{}` 标记。取值有两种: `d[key]` 和 `d.get(key, default)`。

```python
# 创建字典
phone = {"张三": "138-xxxx", "李四": "139-xxxx"}

# 取值方式 1: d[key]
print(phone["张三"])

# 取值方式 2: d.get(key, default)
print(phone.get("张三"))        # 存在则返回
print(phone.get("王五"))        # 不存在返回 None
print(phone.get("王五", "未找到"))  # 指定默认值
```

**--- 执行过程 ---**

① 第 2 行: 创建字典, 关联姓名和电话
② 第 5 行: 通过 key 直接取 value
③ 第 8-10 行: get() 安全取值
④ 第 11 行: 当 key 不存在时提供默认值避免 None

**逐行解剖**

- `{}`: 创建字典的语法符
- `key: value`: 键值对
- `d[key]`: 不存在时报 KeyError(不安全)
- `d.get(key, default)`: 安全取值, 可设默认值

#### 知识点 5: 字典 CRUD - 增删查

**痛点**: 电话簿要添加新联系人、删掉旧联系人或修改号码。

**类比**: 字典就像纸账本,可以随时添加、删除和修改记录。

**解释**: 字典支持增(d[key]=v)、删(del/pop)、改(d[key]=v)、查(in)。
所有操作都靠 key 定位。

```python
# 字典 CRUD
phone = {"张三": "138-xxxx"}

# 增: 直接赋值新 key
phone["李四"] = "139-xxxx"
print("增后:", phone)

# 改: 重新赋值已有 key
phone["张三"] = "137-xxxx"
print("改后:", phone)

# 删: del 删除指定 key
del phone["张三"]
print("del 后:", phone)

# 删: pop 删除并返回 value
val = phone.pop("李四")
print("pop 值:", val)

# 查: 判断 key 是否存在
phone2 = {"张三": "138"}
print("张三" in phone2)
print("李四" in phone2)
```

**--- 执行过程 ---**

① 第 2 行: 初始化字典
② 第 5 行: phone[李四] 添加新联系人
③ 第 9 行: 重新赋值修改已有联系人
④ 第 13 行: del 删除指定键值对
⑤ 第 17 行: pop 删除并返回对应值
⑥ 第 21-22 行: in 查看 key 是否存在

**逐行解剖**

- `d[key] = value`: 增加或修改键值对
- `del d[key]`: 删除键值对, 不存在则报 KeyError
- `d.pop(key)`: 删除并返回 value
- `key in d`: 判断 key 是否存在, 返回 True/False

#### 知识点 6: 字典遍历

**痛点**: 字典存了很多数据,怎么一一拿出来?

**类比**: 电话簿要把所有人的姓名、电话都列出来,就要"遍历"。

**解释**: `d.keys()` 返回所有键, `d.values()` 返回所有值,
`d.items()` 返回所有键值对(key, value)。

```python
# 字典遍历
score = {"数学": 90, "英语": 85, "语文": 78}

# 遍历键
for k in score.keys():
    print("科目:", k)

# 遍历值
for v in score.values():
    print("分数:", v)

# 遍历键值对(最常用)
for k, v in score.items():
    print(k, ":", v)
```

**--- 执行过程 ---**

① 第 2 行: 创建字典
② 第 5-6 行: keys() 遍历所有键
③ 第 9-10 行: values() 遍历所有值
④ 第 13-14 行: items() 同时获取键和值

**逐行解剖**

- `for k in d.keys()`: 遍历所有 key
- `for v in d.values()`: 遍历所有 value
- `for k, v in d.items()`: 同时获得 key 和 value(最强迭代方式)
- 可直接 `for k in d` 默认遍历键

#### 知识点 7: 列表推导式

**痛点**: 创建一个带规律的列表(1-10 的平方)要写 5 行 for 循环。

**类比**: 就像用模具一次压出 100 个饼干,比一个一个手工快得多。

**解释**: 列表推导式(comprehension)是一行生成列表的极简语法。
格式: `[expression for var in iterable]`
可加条件: `[expr for var in iterable if cond]`

```python
# 最简单: 生成 0-9 的平方
squares = [x ** 2 for x in range(10)]
print("平方列表:", squares)

# 带 if 条件: 只保留偶数平方
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print("偶数平方:", even_squares)

# 对字符串列表操作
words = ["你好", "hello", "世界"]
upper_words = [w.upper() for w in words]
print("大写:", upper_words)
```

**--- 执行过程 ---**

① 第 2 行: range(10) 产生 0-9, 对每个 x 计算 x**2
② 第 3 行: 输出 [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
③ 第 6 行: 先筛选偶数再平方
④ 第 10-11 行: 对字符串列表打大写

**逐行解剖**

- `[expr for x in iterable]`: 核心格式, 遍历 + 计算
- `if cond`: 可选筛选条件
- 比 for 循环 + append 简洁得多
- 适合简单转换, 复杂逻辑还是用 for
