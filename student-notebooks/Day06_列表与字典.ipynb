**1. 列表 CRUD(增删改查)**

类比:**列表** = 一排长格子粘在一起,从 0 
开始编号(index)。与字符串很像:有序、支持索引、支持切片。关键区别:**字符串不可变**(改了会创建新串),**列表可变**(原地修改)。


```python
# 列表创建
fruits = ["苹果", "香蕉", "橘子"]
numbers = [1, 2, 3, 4, 5]
empty = []                       # 空列表
print(fruits, numbers, empty)    # ['苹果', '香蕉', '橘子'] [1, 2, 3, 4, 5] []

# 索引与切片(规则与字符串完全一致)
fruits = ["苹果", "香蕉", "橘子", "葡萄", "西瓜"]
print(fruits[0])     # 苹果(第一个)
print(fruits[-1])    # 西瓜(倒数第一个)
print(fruits[1:4])   # ['香蕉', '橘子', '葡萄'](左闭右开)
print(fruits[:3])    # ['苹果', '香蕉', '橘子'](省略 start)
print(fruits[::2])   # ['苹果', '橘子', '西瓜'](步长 2)
```

增用 \`append\`(末尾追加)和 \`insert\`(指定位置插入)。删用 \`pop\`(按索引删,返回被删的值)、\`remove\`(按值删)、\`del\`(按索引删)。


```python
# 增
fruits = ["苹果", "香蕉"]
fruits.append("橘子")           # 末尾追加
fruits.insert(1, "葡萄")        # 索引 1 处插入
print(fruits)                   # ['苹果', '葡萄', '香蕉', '橘子']

# 删
fruits = ["苹果", "香蕉", "橘子", "葡萄"]
last = fruits.pop()             # 删除并返回末尾元素
print(fruits, last)             # ['苹果', '香蕉', '橘子'] 葡萄
fruits.remove("香蕉")           # 按值删除
print(fruits)                   # ['苹果', '橘子']
del fruits[0]                   # 按索引删除
print(fruits)                   # ['橘子']
```

**2. 字典 CRUD(键值对)**

类比:**列表** = 一排长格子(按编号取),**字典** = 快递柜(按 key 取 value)。key 
必须是不可变类型(字符串/数字/元组),value 可以是任何类型。


```python
# 字典创建
phone = {"张三": "13800138000", "李四": "13900139000"}
empty = {}                                   # 空字典
print(phone)

# 增(键不存在则新增)
phone["王五"] = "13700137000"
print(phone)

# 改(键存在则覆盖)
phone["张三"] = "13600136000"

# 查
print(phone["张三"])                         # 13600136000
print(phone.get("赵六", "未找到"))           # 未找到(key 不存在时不报错)

# 删
val = phone.pop("李四")                     # 删除并返回值
print(phone, val)
```

两个超级常用的方法:\`keys()\` 返回所有 key,\`items()\` 返回所有 (key, value) 对。配合 \`for\` 循环遍历字典。


```python
phone = {"张三": "13800138000", "李四": "13900139000"}

# 遍历三种方式
for key in phone:               # 遍历 key
    print(key, phone[key])

for key, value in phone.items():  # 同时遍历 key 和 value
    print(f"{key}: {value}")

print(list(phone.keys()))       # ['张三', '李四']
print(list(phone.values()))     # ['13800138000', '13900139000']
print(list(phone.items()))      # [('张三', '13800138000'), ('李四', '13900139000')]
```

**3. 列表推导式与综合实战**

**列表推导式**:用一行代码生成列表,初看别扭,写多了极其顺手。公式:\`[表达式 for 变量 in 序列 if 条件]\`


```python
# 列表推导式
squares = [x**2 for x in range(1, 6)]
print(squares)              # [1, 4, 9, 16, 25]

evens = [x for x in range(10) if x % 2 == 0]
print(evens)                # [0, 2, 4, 6, 8]
```
