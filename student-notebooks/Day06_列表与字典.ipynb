#### 列表创建与索引切片 —— 把一排数据粘在一起> **痛点**之前用独立变量 `name1`, `name2`, `name3` 存多个数据,变量散落在各处,无法批量操作(比如一次性打印全部)。> **类比****列表** = 一排带编号的储物柜,编号从 0 开始。你只要说"第 3 号柜子",就能直接拿到里面的东西。> **解释**列表用**方括号** `[]` 表示,元素之间用**逗号**分隔。列表和字符串很像:**有序、支持索引和切片**。关键区别:字符串不可变,列表**可以修改**(后面会讲)。

**ASCII 内存图 —— 列表的存储结构**```fruits = ["苹果", "香蕉", "橙子", "葡萄", "西瓜"]内存中:  变量 fruits ──→  ┌─────┬─────┬─────┬─────┬─────┐                   │ 苹果 │ 香蕉 │ 橙子 │ 葡萄 │ 西瓜 │                   └─────┴─────┴─────┴─────┴─────┘  索引(正):          0     1     2     3     4  索引(负):         -5    -4    -3    -2    -1```**规则**:索引从 0 开始,负数从 -1 倒着数。

````python
# 示例 1:创建列表,用索引访问元素fruits = ["苹果", "香蕉", "橙子", "葡萄", "西瓜"]print(fruits[0])    # 第 1 个(索引 0)print(fruits[2])    # 第 3 个(索引 2)print(fruits[-1])   # 最后一个(索引 -1)print(fruits[-2])   # 倒数第 2 个# --- 执行过程 ---# 第 1 行 fruits = [...]:#   ① 在内存中分配连续空间,存入 5 个字符串#   ② 变量 fruits 指向这块空间(列表对象)## 第 3 行 fruits[0]:#   ① 从列表开头偏移 0 个位置 → "苹果"## 第 5 行 fruits[-1]:#   ① 负数索引 = 长度 + (-1) = 5 - 1 = 4#   ② 取索引 4 的元素 → "西瓜"
````

````python
# 示例 2:切片 —— 取一段连续元素fruits = ["苹果", "香蕉", "橙子", "葡萄", "西瓜"]print(fruits[1:4])   # 索引 1~3(含头不含尾)print(fruits[:3])    # 开头到索引 2print(fruits[3:])    # 索引 3 到末尾print(fruits[::-1])  # 逆序(步长 -1)# --- 执行过程 ---# fruits[1:4]:#   ① 起点索引 1(含),终点索引 4(不含)#   ② 取出索引 1、2、3 → ["香蕉", "橙子", "葡萄"]## fruits[::-1]:#   ① 省略起点终点 = 整个列表#   ② 步长 -1 = 从后往前取 → 逆序
````

**逐行解剖**- `fruits[0]` 索引访问,时间复杂度 O(1)- `fruits[1:4]` 切片,**含头不含尾**(左闭右开)- `fruits[::-1]` 步长为 -1,等价于反转列表- 切片不会修改原列表,而是**返回新列表**

**常见错误**1. **索引越界**:`fruits[5]` → IndexError   列表只有 5 个元素,最大索引是 4。2. **切片边界搞反**:以为 `fruits[1:4]` 包含索引 4   实际是**含头不含尾**,索引 4 不包含。

#### 列表可变性 —— 直接改,不报错> **痛点**字符串不能修改单个字符:`s = "abc"; s[0] = "X"` 报错。但现实中经常需要"更新第 3 个学生的成绩"这种操作。> **类比****字符串** = 刻在石头上的字,刻了就改不了。**列表** = 写在白板上的清单,随时可以擦掉重写。> **解释**列表是**可变对象**(mutable),可以通过索引直接赋值修改。这是列表和字符串的**核心区别**。

````python
# 示例 1:列表可变 —— 直接通过索引修改fruits = ["苹果", "香蕉", "橘子"]fruits[0] = "榴莲"   # 把第 1 个元素替换成 "榴莲"print(fruits)         # ["榴莲", "香蕉", "橘子"]# --- 执行过程 ---# 第 1 行 fruits = [...]:#   ① 创建列表,fruits 指向它## 第 2 行 fruits[0] = "榴莲":#   ① 找到索引 0 的位置#   ② 把该位置的引用从 "苹果" 改为 "榴莲"#   ③ 列表长度不变,只是内容变了
````

````python
# 示例 2:对比 —— 字符串不可变s = "abc"# s[0] = "X"  # 取消注释会报错!# TypeError: 'str' object does not support item assignment# 字符串要修改,只能重建一个新字符串s = "X" + s[1:]print(s)  # Xbc# --- 执行过程 ---# s[1:] 切片取出 "bc"# "X" + "bc" 拼接成新字符串 "Xbc"# 变量 s 指向新字符串,原字符串 "abc" 没变
````

**逐行解剖**- `fruits[0] = "榴莲"` 直接修改列表元素,OK- `s[0] = "X"` 对字符串做同样操作 → TypeError- 字符串修改只能**重建**:`s = "X" + s[1:]`- 列表修改是**原地**:不创建新列表,直接改内容

**常见错误**1. **以为修改列表会创建新列表**:   赋值 `fruits[0] = "X"` 是原地修改,列表地址不变。2. **对字符串用索引赋值**:   `s[0] = "X"` → TypeError,字符串不支持。

#### 列表 CRUD(增删)—— append / insert / pop / remove> **痛点**创建列表后,经常需要动态添加或删除元素。比如购物车:用户每买一件商品,就往列表里加一个。> **类比**列表是一个**队伍**,`append` = 排在队伍末尾,`insert` = 插队(指定位置),`pop` = 从队伍里移除一个人。> **解释**CRUD = Create(增) / Read(查) / Update(改) / Delete(删)。本节先讲**增**和**删**,改就是索引赋值(上节已学)。

**ASCII 内存图 —— append 和 pop 的过程**```fruits = ["苹果", "香蕉"]append("橘子") 后:  ┌─────┬─────┬─────┐  │ 苹果 │ 香蕉 │ 橘子 │  ← 末尾追加  └─────┴─────┴─────┘pop() 后:  ┌─────┬─────┐  │ 苹果 │ 香蕉 │  ← 末尾被移除  └─────┴─────┘```

````python
# 示例 1:增 —— append / insert / extendfruits = ["苹果", "香蕉"]fruits.append("橘子")       # 末尾追加print(fruits)                # ["苹果", "香蕉", "橘子"]fruits.insert(1, "榴莲")    # 在索引 1 处插入print(fruits)                # ["苹果", "榴莲", "香蕉", "橘子"]fruits.extend(["葡萄"])     # 合并另一个列表# 等价于 fruits += ["葡萄"]# --- 执行过程 ---# append("橘子"):#   ① 在列表末尾分配新空间,存入 "橘子"#   ② 列表长度 +1## insert(1, "榴莲"):#   ① 把索引 1 及之后的元素全部右移一格#   ② 在索引 1 处放入 "榴莲"## extend(["葡萄"]):#   ① 遍历参数列表,逐个追加到末尾
````

````python
# 示例 2:删 —— pop / remove / clearfruits = ["苹果", "香蕉", "橘子", "葡萄"]last = fruits.pop()          # 删末尾,返回被删元素print(last)                  # 葡萄print(fruits)                # ["苹果", "香蕉", "橘子"]first = fruits.pop(0)        # 删索引 0print(first)                 # 苹果print(fruits)                # ["香蕉", "橘子"]fruits.remove("香蕉")        # 按值删第一个匹配项print(fruits)                # ["橘子"]# fruits.clear()             # 清空整个列表# --- 执行过程 ---# pop():#   ① 取出末尾元素 "葡萄"#   ② 列表长度 -1,返回 "葡萄"## pop(0):#   ① 取出索引 0 的元素 "苹果"#   ② 后续元素全部左移一格## remove("香蕉"):#   ① 从头到尾找第一个值为 "香蕉" 的元素#   ② 删除它,后续元素左移#   ③ 如果找不到 → ValueError
````

**逐行解剖**- `append(x)` 末尾追加,时间复杂度 O(1)- `insert(i, x)` 在索引 i 处插入,后续元素右移- `pop()` 删末尾并返回;`pop(i)` 删索引 i 并返回- `remove(x)` 按值删第一个匹配项,找不到报错- `clear()` 清空列表,变成 `[]`

**常见错误**1. **pop 取值后以为原列表不变**:   `pop()` 会**修改原列表**,不是只取值。2. **remove 不存在的值**:   `remove("不存在的值")` → ValueError。3. **insert 索引超出范围**:   不会报错,超出就等价于 append。

#### 列表排序与查询—— sort / index / count / in> **痛点**拿到一堆数据后,经常需要:排序(从高到低看排名)、查找(某个值在哪)、计数(某个值出现几次)、判断(某个值在不在)。> **类比**列表是一副扑克牌:`sort` = 按大小排一下,`index` = 找到红桃 A 在第几张,`count` = 数一下有几个 A,`in` = 看看这副牌里有没有大王。> **解释**Python 为列表内置了排序和查询方法,不需要手写循环,直接调用即可。

````python
# 示例 1:排序 —— sort (原地排序,返回 None)nums = [3, 1, 4, 1, 5, 9, 2, 6]nums.sort()                  # 升序print(nums)                  # [1, 1, 2, 3, 4, 5, 6, 9]nums.sort(reverse=True)      # 降序print(nums)                  # [9, 6, 5, 4, 3, 2, 1, 1]# --- 执行过程 ---# sort():#   ① 在原列表上进行排序(原地)#   ② 返回 None## 警告:不要写 result = nums.sort()# 因为 sort() 返回 None!## sort(reverse=True):#   ① 排序后反转,得到降序
````

````python
# 示例 2:查询 —— index / count / innums = [3, 1, 4, 1, 5, 9, 2, 6]idx = nums.index(4)          # 第一次出现的索引print(idx)                   # 2cnt = nums.count(1)          # 值 1 出现几次print(cnt)                   # 2print(5 in nums)             # True(在不在)print(7 in nums)             # False# --- 执行过程 ---# index(4):#   ① 从左到右扫描,找到第一个值为 4 的元素#   ② 返回其索引 2;找不到 → ValueError## count(1):#   ① 遍历整个列表,统计值为 1 的个数#   ② 返回 2## 5 in nums:#   ① 等价于遍历 + 比较,找到返回 True
````

**逐行解剖**- `sort()` **原地排序**,返回 None,原列表被修改- `sort(reverse=True)` 降序排列- `index(x)` 返回 x 第一次出现的索引,找不到报错- `count(x)` 返回 x 出现的次数- `x in lst` 返回布尔值,判断 x 是否在列表中

**常见错误**1. **sort 返回 None**:   `sorted_nums = nums.sort()`   → sorted_nums 是 None!   正确做法:先 `nums.sort()`,再 `print(nums)`。2. **index 找不到值**:   `nums.index(99)` → ValueError。   建议先用 `if 99 in nums` 判断。

#### 字典创建与取值 —— 带 key 的快递柜> **痛点**列表用数字索引(0, 1, 2)访问,但现实中数据通常有名字。比如学生信息:姓名、年龄、城市 —— 用列表存,你还得记住"索引 0 是姓名,索引 1 是年龄",容易出错。> **类比****字典** = 带标签的快递柜。你只要说"取 key 为 '姓名' 的柜子",就能拿到值。不需要记住"第几号柜子"。> **解释**字典用**花括号** `{}` 表示,元素是 **key: value** 对。key 必须是不可变类型(字符串、数字),value 任意。

**ASCII 内存图 —— 字典的存储结构**```stu = {"name": "小明", "age": 18, "city": "成都"}内存中:  变量 stu ──→  ┌─────────┬─────────┐                │ key     │ value   │                ├─────────┼─────────┤                │ "name"  │ "小明"  │                │ "age"   │ 18      │                │ "city"  │ "成都"  │                └─────────┴─────────┘取值: stu["name"] → "小明"(通过 key 找 value)```

````python
# 示例 1:创建字典,用 key 取值stu = {"name": "小明", "age": 18, "city": "成都"}print(stu["name"])    # 小明print(stu["age"])     # 18# print(stu["phone"])  # KeyError! key 不存在# --- 执行过程 ---# 第 1 行 stu = {...}:#   ① 创建字典对象,存入 3 个 key-value 对#   ② 变量 stu 指向该字典## 第 3 行 stu["name"]:#   ① 在字典中查找 key "name"#   ② 找到 → 返回对应的 value "小明"## 第 6 行(注释掉的) stu["phone"]:#   ① 查找 key "phone" → 找不到 → KeyError
````

````python
# 示例 2:安全取值 —— get() 带默认值stu = {"name": "小明", "age": 18}print(stu.get("name", "未知"))    # 小明print(stu.get("phone", "未填写"))  # 未填写(不报错)# --- 执行过程 ---# get("name", "未知"):#   ① 查找 key "name" → 找到 → 返回 "小明"#   ② 如果找不到 → 返回默认值 "未知"## get("phone", "未填写"):#   ① 查找 key "phone" → 找不到#   ② 返回默认值 "未填写",不报错
````

**逐行解剖**- `d[key]` 直接取值,key 不存在 → KeyError- `d.get(key, default)` 安全取值,不存在返回 default- key 必须是**不可变类型**(字符串、数字、元组)- value 可以是任意类型(包括列表、字典)

**常见错误**1. **用不存在的 key 直接取值**:   `stu["phone"]` → KeyError。   建议用 `stu.get("phone", "默认值")`。2. **key 大小写敏感**:   `stu["Name"]` 和 `stu["name"]` 是不同的 key。

#### 字典 CRUD —— 增 / 改 / 删> **痛点**创建字典后,经常需要动态添加新字段、修改已有字段、删除不需要的字段。> **类比**字典是一个**快递柜系统**:"增" = 新开一个柜子,放入东西;"改" = 打开已有柜子,换里面的东西;"删" = 拆掉一个柜子。> **解释**字典的增和改语法一样:给 key 赋值。删除用 `pop(key)`,类似列表的 pop。

````python
# 示例 1:增 / 改 / 删 综合演示stu = {"name": "小明", "age": 18}# 增:给新 key 赋值stu["city"] = "成都"print(stu)  # {"name": "小明", "age": 18, "city": "成都"}# 改:给已有 key 赋值(覆盖)stu["age"] = 19print(stu)  # {"name": "小明", "age": 19, "city": "成都"}# 删:pop 删除 key,返回 valueremoved = stu.pop("city")print(removed)  # 成都print(stu)      # {"name": "小明", "age": 19}# --- 执行过程 ---# stu["city"] = "成都":#   ① 查找 key "city" → 不存在#   ② 新增一对 key-value: "city" → "成都"## stu["age"] = 19:#   ① 查找 key "age" → 存在#   ② 把 value 从 18 改为 19## pop("city"):#   ① 查找 key "city" → 找到#   ② 删除该 key-value 对,返回 value#   ③ key 不存在 → KeyError
````

**逐行解剖**- `d[key] = value` 增/改通用语法- key 不存在 → 新增;key 存在 → 覆盖- `d.pop(key)` 删除并返回 value,不存在报错- `d.clear()` 清空字典,变成 `{}`

**常见错误**1. **pop 不存在的 key**:   `stu.pop("phone")` → KeyError。   建议先用 `if "phone" in stu` 判断。2. **以为修改 key 会新增**:   如果 key 已存在,赋值是**修改**,不是新增。

#### 字典遍历 —— keys / values / items> **痛点**拿到一个字典,经常需要:打印所有 key、统计所有 value、同时看 key 和 value。用 `for key in dict` 只能拿到 key,怎么同时拿 value?> **类比**字典是一排带标签的快递柜:`keys()` = 只看所有标签,`values()` = 只看所有柜子里的东西,`items()` = 同时看标签和东西(最常用)。> **解释**`items()` 是字典**最强的迭代方式** ——同时拿到 key 和 value,一次遍历搞定。

````python
# 示例 1:三种遍历方式stu = {"name": "小明", "age": 18, "city": "成都"}# 方式 1:遍历 key(最基础)for key in stu:    print(key)# 方式 2:遍历 valuefor value in stu.values():    print(value)# 方式 3:同时遍历 key 和 value(最强)for key, value in stu.items():    print(f"{key}: {value}")# --- 执行过程 ---# for key in stu:#   ① 依次取出每个 key: "name", "age", "city"## for value in stu.values():#   ① 依次取出每个 value: "小明", 18, "成都"## for key, value in stu.items():#   ① items() 返回 (key, value) 元组#   ② 每次迭代解包: key="name", value="小明"
````

````python
# 示例 2:用 items() 做条件筛选scores = {"张三": 85, "李四": 92, "王五": 78}for name, score in scores.items():    if score >= 90:        print(f"{name} 优秀")# --- 执行过程 ---# 第 3 行 for name, score in ...:#   ① 第 1 次: name="张三", score=85#   ② 85 >= 90? → False → 不打印#   ③ 第 2 次: name="李四", score=92#   ④ 92 >= 90? → True → 打印
````

**逐行解剖**- `for key in d` 等价于 `for key in d.keys()`- `d.keys()` 返回所有 key 的视图- `d.values()` 返回所有 value 的视图- `d.items()` 返回 (key, value) 元组列表- `items()` 是最常用的遍历方式

**常见错误**1. **遍历字典时修改字典**:   在 `for k in d` 循环里 `d.pop(k)` 会报错。2. **以为 items() 返回列表**:   它返回的是**视图对象**,   需要 `list(d.items())` 转列表。

````python
# ============ 学员代码区 ============# menu = {"红烧肉": 38, "鱼香肉丝": 28,#         "麻婆豆腐": 18, "宫保鸡丁": 26}# 1. 用 items() 遍历,打印 "菜名 - XX 元"# 2. 找出价格最低的菜(用 min + values)pass# ============ 参考答案 ============menu = {"红烧肉": 38, "鱼香肉丝": 28,        "麻婆豆腐": 18, "宫保鸡丁": 26}for name, price in menu.items():    print(f"{name} - {price} 元")cheapest_price = min(menu.values())for name, price in menu.items():    if price == cheapest_price:        print(f"最便宜: {name}")
````

#### 综合练习 —— 通讯录 v1现在把**列表和字典**串成一个小项目。**需求清单**:1. 用**字典**存储通讯录,key = 姓名, value = 电话2. 用**菜单循环**让用户选择操作3. 功能:添加、查找、删除、查看全部、退出**关键语法**:- `contacts[name] = phone` 添加- `contacts.get(name, "未找到")` 安全查找- `contacts.pop(name)` 删除- `for name, phone in contacts.items()` 遍历

**ASCII 内存图 —— 通讯录的数据结构**```contacts = {    "小明": "13800138000",    "小红": "13900139000",}内存中:  ┌─────────┬──────────────┐  │ key     │ value        │  ├─────────┼──────────────┤  │ "小明"  │ "13800138000" │  │ "小红"  │ "13900139000" │  └─────────┴──────────────┘```

````python
# 示例:通讯录 v1 完整实现contacts = {}while True:    print("\n=== 通讯录 ===")    print("1.添加  2.查找  3.删除  4.全部  0.退出")    choice = input("请选择:")    if choice == "1":        name = input("姓名:")        phone = input("电话:")        contacts[name] = phone        print("添加成功")    elif choice == "2":        name = input("查找姓名:")        print(contacts.get(name, "未找到"))    elif choice == "3":        name = input("删除姓名:")        if name in contacts:            contacts.pop(name)            print("删除成功")        else:            print("查无此人")    elif choice == "4":        if not contacts:            print("通讯录为空")        else:            for name, phone in contacts.items():                print(f"{name}: {phone}")    elif choice == "0":        print("再见!")        break    else:        print("无效选择")# --- 执行过程 ---# 第 1 行 contacts = {}:#   ① 创建空字典## 第 3 行 while True:#   ① 进入无限循环,直到遇到 break## 第 11 行 contacts[name] = phone:#   ① 新增或修改 key-value 对## 第 15 行 contacts.get(name, "未找到"):#   ① 安全查找,不存在返回默认值## 第 25 行 for name, phone in ...:#   ① items() 遍历所有联系人
````
