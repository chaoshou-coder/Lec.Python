**第 1 讲:列表创建、索引、切片**

**列表** = 一排长格子粘在一起,从 0 开始编号(index)。与字符串很像:**有序、支持索引、支持切片**。关键区别:**字符串不可变**(改了会创建新串),**列表可变**(原地修改)。

```python
# 列表创建:用方括号包裹元素,逗号分隔
fruits = ["苹果", "香蕉", "橘子"]
numbers = [1, 2, 3, 4, 5]
empty = []                       # 空列表,没有任何元素
print(fruits, numbers, empty)    # 打印三个列表

# ===== 索引与切片(规则与字符串完全一致) =====
fruits = ["苹果", "香蕉", "橘子", "葡萄", "西瓜"]
print(fruits[0])     # 苹果 —— 第一个元素,索引从 0 开始
print(fruits[-1])    # 西瓜 —— 倒数第一个,负索引从 -1 开始
print(fruits[1:4])   # ['香蕉', '橘子', '葡萄'] —— 左闭右开,含 1 不含 4
print(fruits[:3])    # ['苹果', '香蕉', '橘子'] —— 省略 start 默认从 0 开始
print(fruits[::2])   # ['苹果', '橘子', '西瓜'] —— 步长 2,每隔一个取一个
print(fruits[::-1])  # ['西瓜', '葡萄', '橘子', '香蕉', '苹果'] —— 逆序
```

列表是**可变的**,跟字符串不一样!可以直接通过索引修改某个位置的元素,字符串不允许这种操作。

```python
# 列表可变:通过索引直接修改元素
fruits = ["苹果", "香蕉"]
fruits[0] = "榴莲"           # 把索引 0 的 "苹果" 改成 "榴莲"
print(fruits)                # ['榴莲', '香蕉']

# 字符串不可变:下面这行取消注释会报 TypeError
# s = "hello"
# s[0] = "H"                 # 报错!字符串不支持索引赋值
# 正确做法是用切片拼新字符串: s = "H" + s[1:]
```

**第 2 讲:列表 CRUD —— append / insert / pop / remove / sort**

CRUD = 增删改查。**增**用 `append`(末尾追加)和 `insert`(指定位置插入);**删**用 `pop`(按索引删,返回被删的值)、`remove`(按值删);**改 + 查**用 `sort` / `index` / `count`。

```python
# ===== 增:append / insert / extend =====
fruits = ["苹果", "香蕉"]

# append:末尾追加一个元素
fruits.append("橘子")
print(fruits)   # ['苹果', '香蕉', '橘子']

# insert:在指定索引位置插入元素,原位置及之后的元素右移
fruits.insert(1, "葡萄")
print(fruits)   # ['苹果', '葡萄', '香蕉', '橘子']

# extend:合并另一列表,等价于 +=
fruits.extend(["西瓜", "芒果"])
print(fruits)   # ['苹果', '葡萄', '香蕉', '橘子', '西瓜', '芒果']
```

```python
# ===== 删:pop / remove / del =====
fruits = ["苹果", "香蕉", "橘子", "葡萄"]

# pop():不带参数删末尾元素,返回被删的值
last = fruits.pop()
print(fruits, "被删:", last)   # ['苹果', '香蕉', '橘子'] 被删: 葡萄

# pop(0):带参数删指定索引的元素
first = fruits.pop(0)
print(fruits, "被删:", first)  # ['香蕉', '橘子'] 被删: 苹果

# remove:按值删除第一个匹配项,不返回值
fruits.remove("香蕉")
print(fruits)                  # ['橘子']

# remove 删除不存在的值会报 ValueError,实际应用时先判断
# fruits.remove("西瓜")        # 报错 ValueError!
```

```python
# ===== 改 + 查:sort / index / count / in =====
nums = [3, 1, 4, 1, 5]

# sort():原地排序,原列表直接改变,返回 None(不要赋值!)
nums.sort()
print(nums)                 # [1, 1, 3, 4, 5]

# sort(reverse=True):降序排列
nums.sort(reverse=True)
print(nums)                 # [5, 4, 3, 1, 1]

# in 运算符:判断元素是否存在于列表
if 3 in nums:
    print("找到了 3")       # 找到了 3

# index():返回元素第一次出现的索引
print(nums.index(3))        # 2

# count():统计元素出现次数
print(nums.count(1))        # 2
```

**第 3 讲:字典 CRUD —— [] / get / items**

类比:**列表**用编号(index)取元素,**字典**用标签(key)取值(value)。key 必须是不可变类型(字符串/数字/元组),value 可以是任何类型。口诀:**列表用索引找元素,字典用 key 找 value**。

```python
# ===== 字典创建 =====
# 用花括号包裹键值对,key 和 value 用冒号分隔
stu = {"name": "小明", "age": 18, "city": "成都"}
empty_dict = {}              # 空字典
print(stu)

# ===== 取值:两种方式 =====
# 方式 1:[]直接取值,key 不存在会报 KeyError
print(stu["name"])           # 小明
# print(stu["phone"])        # 报错 KeyError!

# 方式 2:get()取值,key 不存在返回默认值,不报错
print(stu.get("phone", "未填写"))   # 未填写 —— 安全取值首选 get
```

```python
# ===== 增:直接给新 key 赋值 =====
stu = {"name": "小明", "age": 18}
stu["city"] = "成都"         # key 不存在 → 新增
print(stu)                   # {'name': '小明', 'age': 18, 'city': '成都'}

# ===== 改:给已有 key 赋值,直接覆盖 =====
stu["age"] = 19              # key 存在 → 覆盖
print(stu)                   # {'name': '小明', 'age': 19, 'city': '成都'}

# ===== 删:pop 删除指定 key,返回对应的 value =====
val = stu.pop("city")        # 删除 "city",返回 "成都"
print(stu, "被删的值:", val) # {'name': '小明', 'age': 19} 被删的值: 成都

# clear():清空整个字典
# stu.clear()                # 执行后 stu = {}
```

```python
# ===== 遍历字典:keys / values / items =====
stu = {"name": "小明", "age": 18, "city": "成都"}

# 方式 1:直接遍历 —— 遍历的是 key
for key in stu:
    print(key)               # name  age  city

# 方式 2:keys() —— 显式遍历 key(与方式 1 等价)
for key in stu.keys():
    print(key)

# 方式 3:values() —— 遍历 value
for value in stu.values():
    print(value)             # 小明  18  成都

# 方式 4:items() —— 同时遍历 key 和 value(最常用!)
for key, value in stu.items():
    print(f"{key}: {value}") # name: 小明  age: 18  city: 成都
```

**第 4 讲:列表推导式**

列表推导式 = 用一行代码生成列表。公式:**`[表达式 for 变量 in 序列 if 条件]`**。初看别扭,写多了极其顺手,比 `for` 循环 + `append` 更简洁更高效。

```python
# ===== 基础形式:[表达式 for 变量 in 序列] =====
# 生成 1~5 的平方列表
squares = [x**2 for x in range(1, 6)]
print(squares)              # [1, 4, 9, 16, 25]

# 生成 0~9 中所有偶数的列表
evens = [x for x in range(10) if x % 2 == 0]
print(evens)                # [0, 2, 4, 6, 8]

# ===== 带 if 条件的筛选 =====
# 从列表中筛选出长度大于 2 的字符串
words = ["猫", "老虎", "兔子", "蛇"]
long_words = [w for w in words if len(w) > 1]
print(long_words)           # ['老虎', '兔子']

# ===== 等价对比:推导式 vs 普通循环 =====
# 普通循环写法
result = []
for x in range(5):
    result.append(x * 2)
print(result)               # [0, 2, 4, 6, 8]

# 推导式写法(一行搞定)
result2 = [x * 2 for x in range(5)]
print(result2)              # [0, 2, 4, 6, 8]
```

**今日小结**

本节进入"数据结构"的世界。**列表** = 有序排队,按索引取;**字典** = 带标签盒子,按 key 取。三个必须掌握的操作:

- 列表 CRUD:`append` / `insert` / `pop` / `remove` / `sort`
- 字典 CRUD:`d[key] = value` / `d.pop(key)` / `d.get(key, 默认值)`
- 遍历字典:`for k, v in d.items():`

易错点:`list.sort()` 是原地排序,返回 `None`,不要写 `sorted_lst = lst.sort()`;`d[key]` 找不到会 `KeyError`,安全取值用 `d.get(key, default)`。
