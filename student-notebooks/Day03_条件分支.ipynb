#### 比较运算符 —— 判断两个值的关系

> **痛点**

程序需要判断"年龄是否大于 18"、"密码是否正确",
但之前学的 `=` 只能赋值,没法做比较。

> **类比**

比较运算符就像天平:
`>` 看哪边重,`==` 看两边是否一样重,
`!=` 看两边是不是不一样。

> **解释**

比较运算符对两个值做比较,
结果为 **True**(真)或 **False**(假),
这两种值叫做 **布尔值**(bool)。

**ASCII 内存图 —— 比较运算的结果**

```
print(3 > 1)

计算过程:
  ① 取出 3 和 1
  ② 比较:3 大于 1 吗? → 是
  ③ 返回 True(布尔值)

内存:
  变量 a = True  ← 比较结果存为布尔值
```

**关键**:比较运算的结果是 **bool** 类型,
可以直接赋值给变量,也可以用于 if 判断。

````
# 示例 1:六种比较运算符
# --- 执行过程 ---# 这行:计算 3 > 1,输出结果到屏幕# 这行:计算 3 < 1,输出结果到屏幕# 这行:计算 3 == 3,输出结果到屏幕# 这行:计算 3 != 4,输出结果到屏幕# 这行:计算 3 >= 3,输出结果到屏幕# 这行:计算 3 <= 2,输出结果到屏幕print(3 > 1)     # True —— 大于
print(3 < 1)     # False —— 小于
print(3 == 3)    # True —— 双等号:比较相等
print(3 != 4)    # True —— 不等
print(3 >= 3)    # True —— 大于等于
print(3 <= 2)    # False —— 小于等于
````

````
# 示例 2:比较结果存入变量
# --- 执行过程 ---# 这行:计算右边 18,把结果存入变量 age# 行:执行 is_adult = age >= 18    # 比较结果赋值给变量# 这行:计算 is_adult,输出结果到屏幕# 这行:计算 type(is_adult,输出结果到屏幕age = 18
is_adult = age >= 18    # 比较结果赋值给变量

print(is_adult)         # True
print(type(is_adult))   # <class 'bool'>
````

**逐行解剖**

① `3 > 1` 比较 3 是否大于 1 → True
② `3 == 3` **双等号**是比较,不是赋值
③ `!=` 是不等,`>=` 是大于等于
④ `age >= 18` 返回布尔值,可存入变量
⑤ `type(is_adult)` 确认类型是 bool

**要点**:`=` 是赋值,`==` 是比较,
这是初学者最容易搞混的一对。

**常见错误**

1. **错误现象**:`if age = 18:` 报错 SyntaxError
   **原因**:条件里用了单等号(赋值),应该用 `==`

2. **错误现象**:`print(3 > 1 > 0)` 结果出乎意料
   **原因**:Python 支持链式比较,但可读性差,
   建议拆成 `3 > 1 and 1 > 0`

> - `3 >= 3` 是 True 还是 False?为什么?

#### if 基础 —— 单向判断(如果...就...)

> **痛点**

程序需要"满足某条件时才做某事",
比如"满 18 岁才显示成人内容",
没有 if,只能对所有用户显示相同内容。

> **类比**

小区门禁:**如果**你是业主,**就**开门;
不是业主?什么也不做,直接走开。
这就是"单向判断" —— 只处理成立的情况。

> **解释**

`if` 后面跟一个条件表达式,
条件为 True 时执行缩进块内的代码,
条件为 False 时跳过整个缩进块。

**ASCII 内存图 —— if 的执行流程**

```
age = 20
if age >= 18:
    print("已成年")     ← 条件为 True,执行
print("程序结束")       ← 无论条件如何,都执行

执行流程:
  ① age = 20
  ② 判断 age >= 18 → True
  ③ 进入缩进块 → 打印"已成年"
  ④ 跳出 if → 打印"程序结束"

如果 age = 15:
  ② 判断 age >= 18 → False
  ③ 跳过缩进块
  ④ 直接打印"程序结束"
```

**关键**:Python 用 **缩进**(4 个空格)表示代码块,
不需要像 C/Java 那样写花括号。

````
# 示例 1:最简单的 if
# --- 执行过程 ---# 这行:计算右边 20,把结果存入变量 age# 行:判断条件 age >= 18 → 如果 True 执行下面缩进块    # 这行:计算 "你已经成年了",输出结果到屏幕    # 这行:计算 "可以考取驾照",输出结果到屏幕# 这行:计算 "这行始终执行",输出结果到屏幕age = 20

if age >= 18:
    print("你已经成年了")      # 缩进 4 格
    print("可以考取驾照")      # 同一代码块

print("这行始终执行")          # 没有缩进,不属于 if
````

````
# 示例 2:条件不成立时,跳过 if 块
# --- 执行过程 ---# 这行:计算右边 15,把结果存入变量 age# 行:判断条件 age >= 18 → 如果 True 执行下面缩进块    # 这行:计算 "你已经成年了",输出结果到屏幕# 这行:计算 "这行始终执行",输出结果到屏幕age = 15

if age >= 18:
    print("你已经成年了")      # 不会执行

print("这行始终执行")          # 会执行
````

**逐行解剖**

① `if age >= 18:` if + 条件 + **冒号**
② `print("你已经成年了")` 缩进 4 格,属于 if 块
③ `print("这行始终执行")` 无缩进,不属于 if

**要点**:
- 冒号 `:` 不能忘
- 缩进必须统一(4 个空格)
- 混用 Tab 和空格会报错

**常见错误**

1. **错误现象**:忘记写冒号
   ```python
   if age >= 18    # 错!缺少冒号
   if age >= 18:   # 对!
   ```

2. **错误现象**:缩进不一致
   ```python
   if age >= 18:
       print("A")   # 4 空格
         print("B") # 8 空格 → IndentationError
   ```

> **问自己**:
> - if 末尾为什么要加冒号?
> - 缩进 2 个空格可以吗?8 个呢?
> - 如果条件是 False,if 块里的代码会执行吗?

````
# ============ 学员代码区 ============
# 请定义变量 temperature = 38,
# 如果 temperature >= 37,
# 打印 "体温异常,请就医"。
# 最后打印 "检查完成"。
#
# temperature = 38
# if ... :
#     ...
# print("检查完成")
pass

# ============ 参考答案 ============
temperature = 37.5
if temperature >= 37:
    print("体温异常,请就医")
print("检查完成")
````

#### if-else —— 双向判断(如果...否则...)

> **痛点**

if 只能处理"成立"的情况,
但很多时候"不成立"时也要做点什么,
比如"成年显示 A,未成年显示 B"。

> **类比**

红绿灯:**如果是绿灯,就走;否则,就停。**
两条路必走其一,没有第三种可能。

> **解释**

`if-else` 让程序二选一:
条件成立执行 if 块,不成立执行 else 块,
**有且只有一条路**会被执行。

**ASCII 内存图 —— if-else 的执行流程**

```
age = 15
if age >= 18:
    print("成年人")     ← 条件为 True 时走这里
else:
    print("未成年人")   ← 条件为 False 时走这里
print("判断结束")       ← 无论走哪条,都会到这里

执行流程(age = 15):
  ① 判断 age >= 18 → False
  ② 跳过 if 块
  ③ 进入 else 块 → 打印"未成年人"
  ④ 跳出 if-else → 打印"判断结束"
```

**关键**:if 和 else 是 **互斥** 的,
执行了 if 就不会执行 else,反之亦然。

````
# 示例 1:if-else 基本用法
# --- 执行过程 ---# 这行:计算右边 15,把结果存入变量 age# 行:判断条件 age >= 18 → 如果 True 执行下面缩进块    # 这行:计算 "成年人",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "未成年人",输出结果到屏幕# 这行:计算 "判断结束",输出结果到屏幕age = 15

if age >= 18:
    print("成年人")
else:
    print("未成年人")

print("判断结束")
````

````
# 示例 2:if-else 配合 input 使用
# password = input("请输入密码: ")
# --- 执行过程 ---# 这行:计算右边 "123456"  # 模拟输入,把结果存入变量 password# 行:判断条件 password == "123456" → 如果 True 执行下面缩进块    # 这行:计算 "密码正确,欢迎登录",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "密码错误,拒绝访问",输出结果到屏幕password = "123456"  # 模拟输入

if password == "123456":
    print("密码正确,欢迎登录")
else:
    print("密码错误,拒绝访问")
````

**逐行解剖**

① `if age >= 18:` 判断条件
② `print("成年人")` 条件为 True 时执行
③ `else:` **不需要条件**,表示"其他所有情况"
④ `print("未成年人")` 条件为 False 时执行

**要点**:
- else 后面 **没有条件**,只有冒号
- if 和 else 的缩进要对齐
- else 不能单独存在,必须跟在一个 if 后面

**常见错误**

1. **错误现象**:else 后面写了条件
   ```python
   else age < 18:   # 错!else 不带条件
   else:             # 对!
   ```

2. **错误现象**:if 和 else 缩进不对齐
   ```python
   if age >= 18:
       print("A")
     else:           # 错!缩进不一致
       print("B")
   ```

> - if 和 else 能同时执行吗?
> - else 后面能加条件吗?
> - 如果 if 条件成立,else 块会执行吗?

#### if-elif-else —— 多向判断(多选一)

> **痛点**

if-else 只能二选一,
但现实中经常需要多个分支:
成绩分 A/B/C/D/F,年龄分老/中/青/少。

> **类比**

考试评分:**90 分以上 → A,80-89 → B,
70-79 → C,60-69 → D,60 以下 → F。**
从上到下,第一个满足的就"命中",然后跳出。

> **解释**

`elif` 是 **else if** 的缩写,
可以写多个,Python 从上到下依次判断,
**第一个为 True 的分支执行后,整个结构结束**。

**ASCII 内存图 —— if-elif-else 的执行流程**

```
score = 85
if score >= 90:
    print("A")         ← 85 >= 90? 否,跳过
elif score >= 80:
    print("B")         ← 85 >= 80? 是,执行,跳出
elif score >= 70:
    print("C")         ← 不再判断
else:
    print("F")         ← 不再判断

执行流程:
  ① 判断 score >= 90 → False
  ② 判断 score >= 80 → True → 打印 B
  ③ 跳出整个结构,不再判断后续条件
```

**关键**:elif 是 **顺序敏感** 的,
大范围条件要写在小范围前面,否则会被提前拦截。

````
# 示例 1:成绩等级判断
# --- 执行过程 ---# 这行:计算右边 85,把结果存入变量 score# 行:判断条件 score >= 90 → 如果 True 执行下面缩进块    # 这行:计算 "A",输出结果到屏幕# 行:判断条件 elscore >= 80 → 如果 True 执行下面缩进块    # 这行:计算 "B",输出结果到屏幕# 行:判断条件 elscore >= 70 → 如果 True 执行下面缩进块    # 这行:计算 "C",输出结果到屏幕# 行:判断条件 elscore >= 60 → 如果 True 执行下面缩进块    # 这行:计算 "D",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "F",输出结果到屏幕score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
````

````
# 示例 2:顺序敏感 —— 错误示范
# --- 执行过程 ---# 这行:计算右边 95,把结果存入变量 score# 行:判断条件 score >= 60 → 如果 True 执行下面缩进块    # 这行:计算 "D",输出结果到屏幕# 行:判断条件 elscore >= 90 → 如果 True 执行下面缩进块    # 这行:计算 "A",输出结果到屏幕score = 95

# 错误:小范围写在前面,大范围永远执行不到
if score >= 60:
    print("D")         # 95 >= 60,命中,跳出
elif score >= 90:
    print("A")         # 永远不会执行!
````

````
# 示例 3:顺序正确 —— 大范围在前
# --- 执行过程 ---# 这行:计算右边 95,把结果存入变量 score# 行:判断条件 score >= 90 → 如果 True 执行下面缩进块    # 这行:计算 "A",输出结果到屏幕# 行:判断条件 elscore >= 80 → 如果 True 执行下面缩进块    # 这行:计算 "B",输出结果到屏幕# 行:判断条件 elscore >= 70 → 如果 True 执行下面缩进块    # 这行:计算 "C",输出结果到屏幕# 行:判断条件 elscore >= 60 → 如果 True 执行下面缩进块    # 这行:计算 "D",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "F",输出结果到屏幕score = 95

if score >= 90:
    print("A")         # 95 >= 90,命中,跳出
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
````

**逐行解剖**

① `if score >= 90:` 第一个条件
② `elif score >= 80:` 隐含条件:90 > score >= 80
③ `elif score >= 70:` 隐含条件:80 > score >= 70
④ `else:` 兜底,以上都不满足时执行

**要点**:
- elif 可以写多个,也可以不写
- else 可以省略(可能所有分支都不执行)
- 顺序很重要:从严格到宽松

**常见错误**

1. **错误现象**:顺序写反,大范围永远执行不到
   ```python
   if score >= 60: print("D")    # 先写宽松条件
   elif score >= 90: print("A")  # 永远不会执行
   ```

2. **错误现象**:忘了 elif 需要条件
   ```python
   elif:           # 错!elif 必须带条件
   elif score >= 80:  # 对!
   ```

#### 逻辑运算符 —— 组合多个条件

> **痛点**

现实中的条件往往不是单一的:
"年龄 >= 18 **且** 有身份证"才能开户,
"是会员 **或** 消费满 100"才能打折。
需要把多个条件 **组合** 在一起。

> **类比**

- **and**(且):考试两科都及格才通过
- **or**(或):有学生证或身份证就能入场
- **not**(非):不是会员就不能打折

> **解释**

逻辑运算符把多个布尔条件组合,
返回一个新的布尔值:
- `and`:两边都 True 才 True
- `or`:一边为 True 就 True
- `not`:取反,True 变 False

````
# 示例 1:and —— 两边都满足才成立
# --- 执行过程 ---# 这行:计算右边 25,把结果存入变量 age# 这行:计算右边 True,把结果存入变量 has_id# 行:判断条件 age >= 18 and has_id → 如果 True 执行下面缩进块    # 这行:计算 "可以开户",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "无法开户",输出结果到屏幕age = 25
has_id = True

if age >= 18 and has_id:
    print("可以开户")
else:
    print("无法开户")
````

````
# 示例 2:or —— 任意一个满足就成立
# --- 执行过程 ---# 这行:计算右边 False,把结果存入变量 is_member# 这行:计算右边 150,把结果存入变量 total_price# 行:判断条件 is_member or total_price >= 100 → 如果 True 执行下面缩进块    # 这行:计算 "享受折扣",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "原价",输出结果到屏幕is_member = False
total_price = 150

if is_member or total_price >= 100:
    print("享受折扣")
else:
    print("原价")
````

````
# 示例 3:not —— 取反
# --- 执行过程 ---# 这行:计算右边 False,把结果存入变量 is_raining# 行:判断条件 not is_raining → 如果 True 执行下面缩进块    # 这行:计算 "今天不下雨,可以出门",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "今天下雨,带伞",输出结果到屏幕is_raining = False

if not is_raining:
    print("今天不下雨,可以出门")
else:
    print("今天下雨,带伞")
````

**逐行解剖**

① `age >= 18 and has_id` 两个条件都为 True → True
② `is_member or total_price >= 100` 一个为 True → True
③ `not is_raining` 把 False 变成 True

**要点**:
- `and` 和 `or` 两边必须是 **完整表达式**
- 不能写 `if age >= 18 and <= 60:`
- 要写 `if age >= 18 and age <= 60:`

**常见错误**

1. **错误现象**:and 两边不是完整表达式
   ```python
   if age >= 18 and <= 60:  # 错!右边不完整
   if age >= 18 and age <= 60:  # 对!
   ```

#### 嵌套条件 —— if 里面再写 if

> **痛点**

有些判断需要 **分步骤**:
先看"是否成年",再看"是否退休",
外层不满足时内层根本不用看。

> **类比**

入职流程:**先**看年龄是否达标,
**再**看学历是否达标,
**最后**看经验是否达标。
每一层都通过,才能进入下一层。

> **解释**

在一个 if 的缩进块内部再写一个 if,
就是 **嵌套条件**。
内层 if 只有在外层条件为 True 时才会被判断。

**ASCII 内存图 —— 嵌套条件的执行流程**

```
age = 25
if age >= 18:                    ← 外层判断
    print("成年人")
    if age >= 60:                ← 内层判断
        print("老年人")
    else:
        print("劳动年龄人口")
else:
    print("未成年人")

执行流程(age = 25):
  ① 判断 age >= 18 → True
  ② 进入外层 if 块 → 打印"成年人"
  ③ 判断 age >= 60 → False
  ④ 进入内层 else → 打印"劳动年龄人口"
  ⑤ 跳出整个结构
```

**关键**:嵌套不宜超过 2 层,
超过建议用 `and` 扁平化。

````
# 示例 1:嵌套条件 —— 判断年龄阶段
# --- 执行过程 ---# 这行:计算右边 25,把结果存入变量 age# 行:判断条件 age >= 18 → 如果 True 执行下面缩进块    # 这行:计算 "成年人",输出结果到屏幕    # 行:判断条件 age >= 60 → 如果 True 执行下面缩进块        # 这行:计算 "老年人",输出结果到屏幕    # 行:上面所有条件都不满足,执行这里        # 这行:计算 "劳动年龄人口",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "未成年人",输出结果到屏幕age = 25

if age >= 18:
    print("成年人")
    if age >= 60:
        print("老年人")
    else:
        print("劳动年龄人口")
else:
    print("未成年人")
````

````
# 示例 2:用 and 扁平化(推荐)
# --- 执行过程 ---# 这行:计算右边 25,把结果存入变量 age# 行:判断条件 age >= 60 → 如果 True 执行下面缩进块    # 这行:计算 "老年人",输出结果到屏幕# 行:判断条件 elage >= 18 → 如果 True 执行下面缩进块    # 这行:计算 "劳动年龄人口",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "未成年人",输出结果到屏幕age = 25

if age >= 60:
    print("老年人")
elif age >= 18:
    print("劳动年龄人口")
else:
    print("未成年人")
````

**逐行解剖**

① `if age >= 18:` 外层判断
② `print("成年人")` 外层为 True 时执行
③ `if age >= 60:` 内层判断(只有外层为 True 才判断)
④ `else:` 内层 else

**要点**:
- 嵌套 if 的缩进要 **再缩进 4 格**
- 两层以内可读性好,超过两层建议用 elif + and
- 能用扁平化就不用嵌套

**常见错误**

1. **错误现象**:内层 if 和外层 if 缩进相同
   ```python
   if age >= 18:
       print("A")
       if age >= 60:    # 正确:再缩进 4 格
       print("B")       # 错误:和 if 对齐了
   ```

2. **错误现象**:嵌套超过 3 层,代码难以阅读
   ```python
   if a:
       if b:
           if c:
               if d:  # 太深了!建议重构
   ```

> **问自己**:
> - 嵌套 if 的内层条件什么时候才会被判断?
> - 嵌套和扁平化(and/elif)哪个更好?
> - 什么情况下必须用嵌套?

#### 三元表达式 —— 一行搞定简单二选一

> **痛点**

简单的 if-else 要写 4 行,
如果只是"根据条件选一个值",
能不能更简洁?

> **类比**

点餐:**如果**有会员价,**就**按会员价;
**否则**按原价。
一句话就能说清,不需要写一大段。

> **解释**

三元表达式语法:**值1 if 条件 else 值2**
条件成立返回值1,不成立返回值2。
只适合 **简单赋值**,复杂逻辑不要强塞一行。

````
# 示例 1:三元表达式基本用法
# --- 执行过程 ---# 这行:计算右边 20,把结果存入变量 age# 行:执行 status = "成年" if age >= 18 else "未成年"# 这行:计算 status,输出结果到屏幕age = 20
status = "成年" if age >= 18 else "未成年"

print(status)  # 成年
````

````
# 示例 2:三元表达式 vs 普通 if-else
# 普通写法(4 行)
# --- 执行过程 ---# 这行:计算右边 20,把结果存入变量 age# 行:判断条件 age >= 18 → 如果 True 执行下面缩进块    # 这行:计算右边 "成年",把结果存入变量 status# 行:上面所有条件都不满足,执行这里    # 这行:计算右边 "未成年",把结果存入变量 status# 行:执行 status = "成年" if age >= 18 else "未成年"# 这行:计算 status,输出结果到屏幕age = 20
if age >= 18:
    status = "成年"
else:
    status = "未成年"

# 三元表达式(1 行)
status = "成年" if age >= 18 else "未成年"

print(status)  # 成年
````

````
# 示例 3:三元表达式的典型场景 —— 求绝对值
# --- 执行过程 ---# 这行:计算右边 -5,把结果存入变量 num# 行:执行 abs_value = num if num >= 0 else -num# 这行:计算 abs_value,输出结果到屏幕num = -5
abs_value = num if num >= 0 else -num

print(abs_value)  # 5
````

**逐行解剖**

① `"成年" if age >= 18 else "未成年"` 三元表达式
② 先判断 `age >= 18`
③ 为 True → 返回 `"成年"`
④ 为 False → 返回 `"未成年"`
⑤ 把返回值赋给 status

**要点**:
- 语法:`值1 if 条件 else 值2`
- 不要嵌套三元表达式(可读性差)
- 复杂逻辑用普通 if-else

**常见错误**

1. **错误现象**:把 if-else 语句塞进一行
   ```python
   if age >= 18: print("A") else: print("B")
   # 语法上可以,但 PEP 8 不推荐
   ```

2. **错误现象**:嵌套三元表达式
   ```python
   # 可读性极差!
   level = "A" if s >= 90 else "B" if s >= 80 else "C"
   # 建议用 if-elif-else
   ```

#### in 运算符 —— 判断元素是否在序列中

> **痛点**

需要判断"某个值是否在列表中",
或"某个字符是否在字符串中",
用多个 `or` 写起来很繁琐。

> **类比**

点名:**"张三"在不在名单里?**
翻名单找一遍,找到了就说"在",
没找到就说"不在"。
`in` 就是帮你"翻名单"的运算符。

> **解释**

`in` 判断元素是否存在于
字符串、列表、元组等 **可迭代对象** 中,
返回 True / False。
`not in` 是反向判断(不在其中)。

````
# 示例 1:in 判断字符串中是否包含子串
# --- 执行过程 ---# 这行:计算右边 "zhangsan@example.com",把结果存入变量 email# 行:判断条件 "@" in email → 如果 True 执行下面缩进块    # 这行:计算 "看起来像邮箱",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "不是邮箱格式",输出结果到屏幕email = "zhangsan@example.com"

if "@" in email:
    print("看起来像邮箱")
else:
    print("不是邮箱格式")
````

````
# 示例 2:in 判断列表中是否有某个元素
# --- 执行过程 ---# 这行:计算右边 ["苹果", "香蕉", "橘子"],把结果存入变量 fruits# 行:判断条件 "苹果" in fruits → 如果 True 执行下面缩进块    # 这行:计算 "有苹果",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "没苹果",输出结果到屏幕fruits = ["苹果", "香蕉", "橘子"]

if "苹果" in fruits:
    print("有苹果")
else:
    print("没苹果")
````

````
# 示例 3:not in 判断"不在其中"
# --- 执行过程 ---# 这行:计算右边 ["黑客", "垃圾账号"],把结果存入变量 blacklist# 这行:计算右边 "正常用户",把结果存入变量 username# 行:判断条件 username not in blacklist → 如果 True 执行下面缩进块    # 这行:计算 f"{username} 可以注册",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 f"{username} 已被封禁",输出结果到屏幕blacklist = ["黑客", "垃圾账号"]

username = "正常用户"
if username not in blacklist:
    print(f"{username} 可以注册")
else:
    print(f"{username} 已被封禁")
````

**逐行解剖**

① `"@" in email` 判断 `@` 是否在 email 中
② `"苹果" in fruits` 判断列表中是否有该元素
③ `username not in blacklist` 判断 **不在** 黑名单
④ 返回值是 bool,可直接用于 if

**要点**:
- `in` 可用于字符串、列表、元组、字典(查键)
- `not in` 是 `in` 的反向操作
- 字符串的 `in` 是 **子串匹配**

**常见错误**

1. **错误现象**:对数字使用 in
   ```python
   if 3 in 123:  # 错!int 不可迭代
   if "3" in "123":  # 对!字符串可迭代
   ```

2. **错误现象**:混淆 in 和 ==
   ```python
   if "app" in "apple":  # 对!子串匹配
   if "app" == "apple":  # 错!完全相等
   ```

把今天学的 **比较运算符 + if-elif-else** 串成一个完整程序。

````
# 综合练习 1:成绩等级判断
# --- 执行过程 ---# 这行:计算右边 85,把结果存入变量 score# 行:判断条件 score < 0 or score > 100 → 如果 True 执行下面缩进块    # 这行:计算 "成绩无效",输出结果到屏幕# 行:判断条件 elscore >= 90 → 如果 True 执行下面缩进块    # 这行:计算 "A",输出结果到屏幕# 行:判断条件 elscore >= 80 → 如果 True 执行下面缩进块    # 这行:计算 "B",输出结果到屏幕# 行:判断条件 elscore >= 70 → 如果 True 执行下面缩进块    # 这行:计算 "C",输出结果到屏幕# 行:判断条件 elscore >= 60 → 如果 True 执行下面缩进块    # 这行:计算 "D",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "F",输出结果到屏幕score = 85

# 先判断是否合法
if score < 0 or score > 100:
    print("成绩无效")
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
````

#### 综合练习 2 —— BMI 计算器

把 **if-else + 比较运算符 + 逻辑运算** 综合运用。

**需求**:
- 输入身高(m)和体重(kg)
- 计算 BMI = 体重 / 身高^2
- 根据中国标准输出偏瘦/正常/偏胖/肥胖
- BMI < 18.5 偏瘦,18.5-24 正常,
  24-28 偏胖,>= 28 肥胖

````
# 综合练习 2:BMI 计算器
# --- 执行过程 ---# 这行:计算右边 1.75   # 身高(米),把结果存入变量 height# 这行:计算右边 70     # 体重(公斤),把结果存入变量 weight# 这行:计算右边 weight / (height ** 2),把结果存入变量 bmi# 这行:计算 f"BMI = {bmi:.1f}",输出结果到屏幕# 行:判断条件 bmi < 18.5 → 如果 True 执行下面缩进块    # 这行:计算 "偏瘦",输出结果到屏幕# 行:判断条件 elbmi < 24 → 如果 True 执行下面缩进块    # 这行:计算 "正常",输出结果到屏幕# 行:判断条件 elbmi < 28 → 如果 True 执行下面缩进块    # 这行:计算 "偏胖",输出结果到屏幕# 行:上面所有条件都不满足,执行这里    # 这行:计算 "肥胖",输出结果到屏幕height = 1.75   # 身高(米)
weight = 70     # 体重(公斤)

bmi = weight / (height ** 2)
print(f"BMI = {bmi:.1f}")

if bmi < 18.5:
    print("偏瘦")
elif bmi < 24:
    print("正常")
elif bmi < 28:
    print("偏胖")
else:
    print("肥胖")
````

**逐行解剖**

① `height ** 2` 计算身高的平方
② `weight / (height ** 2)` 计算 BMI
③ `f"{bmi:.1f}"` 保留 1 位小数
④ 按 BMI 区间判断体型

**要点**:
- `**` 是乘方运算符
- 区间判断用 `<` 而不是 `<=`
  (因为上一个分支已经排除了更小的情况)
