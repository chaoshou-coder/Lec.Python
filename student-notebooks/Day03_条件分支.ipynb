---

**第 1 讲:比较运算符**

比较运算符用来比较两个值的大小关系,结果为 **True** 或 **False**(布尔值)。

注意:**==** 是比较相等(双等号),**=** 是赋值(单等号),千万不要搞混。

```python
print(3 > 1)     # True
print(3 < 1)     # False
print(3 == 3)    # True —— 双等号才是比较
print(3 != 4)    # True —— 不等
print(3 >= 3)    # True —— 大于等于
print(3 <= 2)    # False
```

---

**第 2 讲:if 语句**

**if** 在条件成立时执行对应的代码块。Python 用 **缩进**(4 个空格)来区分代码块,不需要花括号。

冒号 `:` 不能忘;缩进必须统一,混用 Tab 和空格会报错。

```python
age = 18
if age >= 18:                    # 冒号不能忘
    print("你已经成年了")        # 缩进 4 格,属于 if
    print("可以考取驾照")
print("这行始终执行")             # 没有缩进,不属于 if
```

---

**第 3 讲:if-else 二选一**

条件成立走 `if`,不成立走 `else`,两条路必走其一。

```python
age = 16
if age >= 18:
    print("成年人")
else:                            # else 也带冒号
    print("未成年人")
```

---

**第 4 讲:if-elif-else 多选一**

当需要多个分支时,用 **elif**(= else if) 串起来。从上到下,第一个满足的分支执行,执行完就跳出整个结构。

顺序敏感:大范围条件(比如 >= 90)要写在小范围前面,否则会被提前拦截。

```python
score = 85
if score >= 90:
    print("A")
elif score >= 80:                # 隐含条件:90 > score >= 80
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

---

**第 5 讲:逻辑运算 and / or / not**

逻辑运算符把多个条件 **组合** 在一起:**and** 两边都真才真;**or** 一边为真就真;**not** 取反。

```python
# and:两边都满足才成立 —— 判断是否在劳动年龄
age = 25
if age >= 18 and age <= 60:
    print("劳动年龄")
```

```python
# or:任意一个满足就成立 —— 判断无效成绩
score = 105
if score < 0 or score > 100:
    print("成绩无效")
```

```python
# not:取反 —— 判断不在下雨
is_raining = False
if not is_raining:
    print("今天不下雨")
```

---

**第 6 讲:嵌套条件 —— 条件里再挖条件**

在一个 `if` 的内部再写一个 `if`,就是嵌套。两层以内可读性好,超过两层建议用 `elif` + `and`  **扁平化**。

```python
age = 25
if age >= 18:
    if age >= 60:                # 第二层缩进
        print("老年人")
    else:
        print("成年人")
else:
    print("未成年人")
```

---

**第 7 讲:三元表达式 —— 一行搞定简单二选一**

语法:**值1 if 条件 else 值2**。条件成立返回值1,不成立返回值2。只适合简单场景,复杂逻辑不要强塞一行。

```python
age = 18
status = "成年" if age >= 18 else "未成年"
print(status)      # 成年
```

---

**第 8 讲:in 运算符 —— 判断是否包含 / 是否在其中**

**in** 判断某个元素是否存在于字符串、列表、字典等容器中,返回 True / False。

```python
# in 判断字符串中是否包含某个子串
email = "zhangsan@163.com"
if "@" in email:
    print("看起来像邮箱")
```

```python
# in 判断列表里是否有某个元素
fruits = ["苹果", "香蕉", "橘子"]
if "苹果" in fruits:
    print("有苹果")
else:
    print("没苹果")
```

---

**今日小结**

今天你给代码加上了 **判断力**:
- **比较运算符** > < >= <= == != 返回布尔值
- **if / elif / else** 分岔执行,第一个满足就跳出
- **and / or / not** 组合多个条件
- **嵌套** 条件里再挖条件,但尽量扁平化
- **三元表达式** `a if cond else b` 一行搞定简单二选一
- **in** 判断成员是否在容器中

> 易错点:**=** 是赋值 **==** 是比较,if 里写 `if x = 5:` 必报错;**elif** 顺序敏感,大范围条件写前面;`or` / `and` 两边必须是完整表达式。
