**1. 比较运算符与 if 基础**

比较运算符两边的表达式计算后返回 **`True` / `False`**(布尔值)。口诀:**一个"等于"装盒子,两个"等于"比谁大**。`if` 语句:条件成立才执行,冒号不能忘,缩进 
4 格决定归属。


```python
# 比较运算符
print(3 > 1)    # True
print(3 < 1)    # False
print(3 == 3)   # True   ← 注意是双等号!(单等号是赋值)
print(3 != 4)   # True   ← 不等
print(3 >= 3)   # True
print(3 <= 2)   # False

# if 语句:只有条件成立才执行
age = 18
if age >= 18:                    # 冒号不能忘!
    print("你已经成年了")        # 缩进 4 格
    print("可以考取驾照")
print("这行始终执行")             # 没有缩进,不属于 if

# if-else:二选一
age = 16
if age >= 18:
    print("成年人")
else:                            # else 也始终带冒号
    print("未成年人")

# if-elif-else:多选一
score = 85
if score >= 90:
    print("A")
elif score >= 80:                # elif = else if,可写多个
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

**✏️ 练一练 ⭐⭐**

**2. 逻辑运算 —— 组合多个条件**

`and` 两边都真才真,`or` 一边为真就真,`not` 取反。注意:`or`/`and` 两边必须是**完整表达式**,`x == 1 or 2` 是错误的,要写 `x == 1 
or x == 2`。


```python
# and:两边都真才真
age = 25
if age >= 18 and age <= 60:
    print("劳动年龄")

# or:一边为真就真
score = -5
if score < 0 or score > 100:
    print("成绩不合法")

# not:取反
is_raining = False
if not is_raining:
    print("不用带伞")

# 实战:闰年判断(四年一闰,百年不闰,四百年又闰)
year = 2000
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(is_leap)  # True
```

**3. 嵌套条件与三元表达式**

条件里再挖条件就是嵌套。两层嵌套以内可用,超过两层建议用 `elif` + `and` 扁平化。三元表达式:一行搞定简单二选一,语法:`值1 if 条件 else 值2`。

```python
# 嵌套条件:条件里再挖条件
age = 25
if age >= 18:
    if age >= 60:
        print("老年人")
    else:
        print("成年人")
else:
    print("未成年人")

# 三元表达式:一行搞定简单二选一
age = 18
status = "成年" if age >= 18 else "未成年"
print(status)  # 成年

# 三元表达式嵌套(可读性下降,了解即可)
score = 85
level = "优秀" if score >= 90 else "良好" if score >= 80 else "一般"
print(level)  # 良好
```

**✏️ 练一练 ⭐⭐⭐**

输入三个整数,用 `if` 嵌套找出最大值(不用 `max()`)。

示例输入:`3 7 5` → 输出:`最大值是 7`

```python
# ======================
# 学员代码区
# ======================
```

```python
# 参考答案
a = int(input("请输入第一个整数:"))
b = int(input("请输入第二个整数:"))
c = int(input("请输入第三个整数:"))
if a >= b:
    if a >= c:
        print(f"最大值是 {a}")
    else:
        print(f"最大值是 {c}")
else:
    if b >= c:
        print(f"最大值是 {b}")
    else:
        print(f"最大值是 {c}")
```

**4. in 运算符 —— 判断包含**

`in` 判断一个元素是否包含在字符串/列表/字典中,返回 `True` / `False`。常和 `if` 搭配做成员判断。

```python
# 字符串包含判断
email = "zhangsan@163.com"
if "@" in email:
    print("看起来像邮箱")

# 成员判断(列表)
fruits = ["苹果", "香蕉"]
if "橘子" in fruits:
    print("有橘子")
else:
    print("没橘子")

# 实战:BMI 计算器
h = float(input("请输入身高(米):"))
w = float(input("请输入体重(千克):"))
bmi = w / (h ** 2)
if bmi < 18.5:
    result = "偏瘦"
elif bmi < 24:            # 隐含 >= 18.5
    result = "正常"
elif bmi < 28:            # 隐含 >= 24
    result = "偏胖"
else:
    result = "肥胖"
print(f"BMI: {bmi:.2f},评级: {result}")
```

**✏️ 练一练 ⭐⭐⭐**

简易计算器:输入两个整数和一个运算符(`+` `-` `*` `/`),输出计算结果。除法时除数为 0 输出 `除数不能为 0`。

示例输入:`10 5 /` → 输出:`2.0`
示例输入:`10 0 /` → 输出:`除数不能为 0`

```python
# ======================
# 学员代码区
# ======================
```

```python
# 参考答案
a = int(input("请输入第一个整数:"))
b = int(input("请输入第二个整数:"))
op = input("请输入运算符(+-*/):")
if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    if b == 0:
        print("除数不能为 0")
    else:
        print(a / b)
else:
    print("未知运算符")
```

**今日小结**

本节让代码拥有判断力。对比:

| 结构 | 适用场景 |
|---|---|
| `if` | 单条件,成立就执行 |
| `if-else` | 二选一 |
| `if-elif-else` | 多选一(从上到下第一个满足的分支) |
| 嵌套 `if` | 条件里再挖条件(≤2 层) |
| 三元表达式 | 一行简单二选一 |

> 🔴 **易错点**:
> 1. `=` 是赋值 `==` 是比较,`if x = 5:` 必报错
> 2. 缩进必须统一,混用 Tab 和空格会報 `IndentationError`
> 3. `elif` 顺序敏感,范围宽的(高分/高金额)条件要写前面
> 4. `or`/`and` 两边必须是完整表达式,`x == 1 or 2` 是错误的
