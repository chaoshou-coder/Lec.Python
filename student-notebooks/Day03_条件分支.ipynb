**1. 比较运算符 — 让程序比大小**

比较运算符是条件判断的基础。它把两个值拿来"比一比"，
结果只可能是 **True** 或 **False**
（注意大小写，不能写成 true/false）。

常用的比较运算符有：**>**（大于）、**<**（小于）、
**==**（等于）、**!=**（不等于）、
**>=**（大于等于）、**<=**（小于等于）。
特别注意 **==** 表示"比较是否相等"，
而单个 **=** 是赋值，这是初学者最容易搞混的一对。

```python
# 比较运算符 demo
a = 10
b = 20

# 大于 / 小于
print(a > b)    # False，10 不大于 20
print(a < b)    # True，10 小于 20

# 等于 / 不等于
print(a == b)   # False，10 和 20 不相等
print(a != b)   # True，10 不等于 20

# 大于等于 / 小于等于
print(a >= 10)  # True，10 等于 10
print(b <= 15)  # False，20 不小于等于 15

# 易错：== 与 = 的区别
# a = b   → 把 b 的值赋给 a
# a == b  → 判断 a 是否等于 b，返回 True 或 False
```

**2. if / else — 二选一的分岔口**

**if** 语句让程序在满足条件时执行某段代码。语法要点：条件末尾加冒号 **:**，下面要执行的代码必须缩进 4 个空格。

当"条件不满足时要做另一件事"，就用 **else** 来兜底。if 和 else 像一条分岔路，程序会且只会走进其中一条。

```python
# if / else 示例
score = 75

# 单独一个 if：满足条件就执行
if score >= 60:
    print("及格了，继续加油！")

# if-else：二选一，必定执行一个分支
if score >= 60:
    print("考试成绩：及格")
else:
    print("考试成绩：不及格，需要补考")

# 字符串也可以比较
password = "abc123"
if password == "abc123":
    print("密码正确，欢迎登录")
else:
    print("密码错误，拒绝访问")
```

**3. if / elif / else — 多选一**

当有两种以上的可能时，用 **elif**（else if 的缩写）列出所有分支。程序从上到下依次判断，遇到第一个满足的条件就进入对应分支，然后立刻退出整个判断结构。

顺序很重要！把最严格、最特殊的条件写在前面。比如判断成绩等级时，先判断 >=90 再判断 >=80；如果反过来写，一个 95 分的人会永远停留在第一个分支里。

```python
# 成绩等级评定：A / B / C / D / F
score = 85

# 从高到低判断，第一个满足的等级即为结果
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"分数：{score}，等级：{grade}")

# 演示顺序的重要性：先写范围小的
age = 25
if age < 18:
    print("未成年人")
elif age < 60:
    print("成年人")
else:
    print("老年人")
```

**4. 逻辑运算 and / or / not**

有时候一个条件不够用。**and** 表示"两边都成立才算成立"，**or** 表示"至少一个成立就算成立"，**not** 表示"取反"。

注意：and/or 两边必须是完整的表达式。不能偷懒写 `x==1 or 2`，要老老实实写 `x==1 or x==2`。

```python
# 逻辑运算示例
a = 10
b = 20

# and：两边都成立才返回 True
print(a > 5 and b > 15)   # True and True → True
print(a > 5 and b > 50)   # True and False → False

# or：至少一个成立就返回 True
print(a > 5 or b > 50)    # True or False → True
print(a < 5 or b > 50)    # False or False → False

# not：取反
print(not a > 5)          # not True → False
print(not False)          # True

# 综合应用：判断闰年
# 规则：能被 4 整除但不能被 100 整除，或者能被 400 整除
year = 2024
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(f"{year} 年是闰年吗？{is_leap}")

year = 1900
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(f"{year} 年是闰年吗？{is_leap}")
```

**5. 三元表达式 — 一行搞定简单二选一**

当 if/else 每个分支只有一行代码时，可以用三元表达式压缩成一行：`结果A if 条件 else 结果B`。条件成立取结果A，否则取结果B。

三元表达式只推荐用在简单场景。如果逻辑复杂，老老实实写多行 if/else 更好读。

```python
# 三元表达式示例
age = 20

# 基本用法：条件成立取前半段，否则取后半段
status = "成年" if age >= 18 else "未成年"
print(f"年龄 {age}，状态：{status}")

# 也可以直接用在 print 里
print("偶数" if age % 2 == 0 else "奇数")

# 两个数取较大值
x = 15
y = 8
max_val = x if x > y else y
print(f"{x} 和 {y} 中较大的是 {max_val}")
```

**6. in 运算符 — 判断"包含"**

**in** 用于判断某个元素是否"在"一个序列（字符串、列表等）中，返回 True/False。写作 `元素 in 序列`。

实际场景：检查邮箱里有没有 @ 符号、判断用户输入的水果是否在菜单里……这类"有没有包含"的问题用 in 最方便。

```python
# in 运算符示例

# 在字符串中查找子串
email = "hello@example.com"
if "@" in email:
    print("邮箱格式正确（包含 @）")
else:
    print("邮箱格式错误（缺少 @）")

# 判断水果是否在菜单里
fruits = ["苹果", "香蕉", "橘子", "葡萄"]
target = "橘子"
if target in fruits:
    print(f"{target} 在菜单中")
else:
    print(f"{target} 不在菜单中")

target = "榴莲"
if target in fruits:
    print(f"{target} 在菜单中")
else:
    print(f"{target} 不在菜单中")

print("-" * 20)
# not in 表示"不包含"
if "榴莲" not in fruits:
    print("菜单中找不到榴莲")
```

**7. 综合实战 — BMI 计算器**

BMI（身体质量指数）= 体重(kg) ÷ 身高(m)²。中国成人标准：偏瘦(<18.5)、正常(18.5~23.9)、偏胖(24~27.9)、肥胖(≥28)。

我们来写一个完整的 BMI 计算器：输入身高体重，用 if/elif/else 判断等级，用 f-string 输出整洁的结果。

```python
# BMI 计算器
# 输入：身高(米)、体重(千克)
height = 1.75
weight = 70

# 计算 BMI
bmi = weight / (height ** 2)

# 用 if/elif/else 判断等级
if bmi < 18.5:
    level = "偏瘦"
elif bmi < 24:
    level = "正常"
elif bmi < 28:
    level = "偏胖"
else:
    level = "肥胖"

# f-string 输出结果，保留一位小数
print(f"身高 {height}m，体重 {weight}kg")
print(f"BMI = {bmi:.1f}，判定：{level}")
```

**8. 易错点与今日小结**

今天我们让程序学会了"判断"，核心要点有：

- **比较运算符** `> < == != >= <=` 返回 **True/False**
- **if** 依靠冒号和缩进来划分代码块
- **if-else** 二选一；**if-elif-else** 多选一，第一个满足即退出
- **and/or/not** 组合条件，两边都要写完整表达式
- **三元表达式** 只适合简单场景，复杂逻辑别强行一行
- **in** 判断"包含"关系，可用于字符串和列表

易错点速记：

- `=` 是赋值，`==` 是比较
- elif 顺序要从高到低，最严格的放前面
- or/and 两边要写完整：`x==1 or x==2`，不能偷懒写 `x==1 or 2`
- 三元表达式只用于简单场景
- 缩进要统一，不要用 tab 和空格混着来
