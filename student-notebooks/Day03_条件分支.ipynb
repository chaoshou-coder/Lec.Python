## 第一讲(15 分钟)—— 比较运算符与 `if` 基础

### 比较运算符:两个值比大小

比较运算符返回 **`True` / `False`**(布尔值):

| 运算符 | 含义 |
|---|---|
| `>` | 大于 |
| `<` | 小于 |
| `==` | 相等(双等号!) |
| `!=` | 不等 |
| `>=` | 大于等于 |
| `<=` | 小于等于 |

````python
# 示例代码:比较运算符
print(3 > 1)    # True
print(3 < 1)    # False
print(3 == 3)   # True   ← 注意是双等号!(单等号是赋值)
print(3 != 4)   # True   ← 不等
print(3 >= 3)   # True
print(3 <= 2)   # False
````

> 🔴 **教学红线**: `=` 是赋值,`==` 是比较。`if x = 5:` 会报 `SyntaxError`!

**口诀**: 一个"等于"装盒子,两个"等于"比谁大。

### `if` 语句:只有条件成立才执行

Python 用**缩进**(4 个空格)决定哪些语句属于哪个分支。

````python
# 示例代码:if 语句
age = 18
if age >= 18:                    # 冒号不能忘!
    print("你已经成年了")        # 缩进 4 格
    print("可以考取驾照")
print("这行始终执行")             # 没有缩进,不属于 if
````

### `if-else`:二选一

````python
# 示例代码:if-else
age = 16
if age >= 18:
    print("成年人")
else:                            # else 也始终带冒号
    print("未成年人")
````

### `if-elif-else`:多选一

**执行路径**:从上到下第一个满足的分支执行,执行完就跳出整个结构。

````python
# 示例代码:if-elif-else
score = 85
if score >= 90:
    print("A")
elif score >= 80:                # elif = else if,可写多个
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
````

> 🔴 **易错点**: `elif` 顺序很重要!要从高分往低分写,否则第一名条件会吃掉所有人。

## 第二讲(15 分钟)—— 逻辑运算与嵌套

### 逻辑运算 `and` `or `not`:组合多个条件

| 运算符 | 含义 | 例子 |
|---|---|---|
| `and` | 两边都真才真 | `age >= 18 and age < 60` |
| `or`  | 一边为真就真 | `score < 0 or score > 100` |
| `not` | 取反 | `not is_raining` |

````python
# 示例代码:逻辑运算
age = 25
if age >= 18 and age <= 60:
    print("劳动年龄")

# 判断闰年:能被4整除但不能被100整除,或者能被400整除
year = 2024
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(f"{year} 是闰年? {is_leap}")
````

> 🔴 **易错点**: `if x == 1 or 2:` 是错误的!要写 `if x == 1 or x == 2:`

### 嵌套条件:条件里再挖条件

````python
# 示例代码:嵌套条件
age = 25
if age >= 18:
    if age >= 60:
        print("老年人")
    else:
        print("成年人")
else:
    print("未成年人")
````

### 三元表达式:一行搞定简单二选一

语法: `值1 if 条件 else 值2`

````python
# 示例代码:三元表达式
age = 18
status = "成年" if age >= 18 else "未成年"
print(status)  # 成年

# 再举一例
score = 75
result = "及格" if score >= 60 else "不及格"
print(result)  # 及格
````

````python
# 参考答案
year = int(input("请输入年份:"))
month = int(input("请输入月份:"))

if month in [1, 3, 5, 7, 8, 10, 12]:
    days = 31
elif month in [4, 6, 9, 11]:
    days = 30
elif month == 2:
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        days = 29
    else:
        days = 28
else:
    days = -1  # 非法月份

if days > 0:
    print(f"{year} 年 {month} 月有 {days} 天")
else:
    print("月份输入错误!")
````

````python
# 参考答案
a = int(input("请输入第一个数:"))
b = int(input("请输入第二个数:"))
c = int(input("请输入第三个数:"))

if a >= b:
    if a >= c:
        max_val = a
    else:
        max_val = c
else:
    if b >= c:
        max_val = b
    else:
        max_val = c

print(f"最大值: {max_val}")
````

## 第三讲(15 分钟)—— `in` 运算符与综合实战

### `in` 运算符:判断包含

````python
# 示例代码:in 运算符
# 字符串包含判断
email = "zhangsan@163.com"
if "@" in email:
    print("看起来像邮箱")

# 列表成员判断
fruits = ["苹果", "香蕉", "橘子"]
if "橘子" in fruits:
    print("有橘子")
else:
    print("没橘子")
````

### 综合实战 1:BMI 计算器

````python
# 示例代码:BMI 计算器
h = float(input("请输入身高(米):"))
w = float(input("请输入体重(千克):"))
bmi = w / (h ** 2)

if bmi < 18.5:
    result = "偏瘦"
elif bmi < 24:            # 隐含 >= 18.5
    result = "正常"
elif bmi < 28:            # 隐含 >= 24
    result = "偏胖"
else:
    result = "肥胖"

print(f"BMI: {bmi:.2f}, 评级: {result}")
````

### 综合实战 2:闰年判断

**闰年口诀**: 四年一闰,百年不闰,四百年又闰。

````python
# 示例代码:闰年判断
year = int(input("请输入年份:"))
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

if is_leap:
    print(f"{year} 是闰年")
else:
    print(f"{year} 不是闰年")
````
