> 鸭子类型:Duck Typing 多态

**痛点**:Day06 用 isinstance 判断类型,但每加类型都要改分发函数。

**类比**:'像鸭子就是鸭子' -- 走路像、叫声像,它就是鸭子,不管它继承谁。

**解释**:Python 不关注对象是哪个类,只关心 '有没有这个方法'。不同类实现同名方法,调用方无需 if-elif。

```python

class Alipay:
    def execute(self, amount):
        return f"支付宝支付 {amount} 元"

class WeChatPay:
    def execute(self, amount):
        return f"微信支付 {amount} 元"

class ApplePay:
    def execute(self, amount):
        return f"Apple Pay {amount} 元"

def checkout(cart_total, payment):
    # 不管 payment 是什么类,只要它有 execute 方法
    return payment.execute(cart_total)

for p in [Alipay(), WeChatPay(), ApplePay()]:
    print(checkout(99.0, p))
```

> 逐行解剖

- 三个类没有继承同一个父类
- 但它们都有 execute 方法 -> 鸭子类型
- checkout 不关心类型,只关心有没有 execute
- 完全去掉 if-elif!

> NCDL Break It #1:鸭子类型的代价

鸭子类型的致命弱点:**漏写方法不会提前报错**。

你写了 Alipay,但漏掉了 execute 方法。
测试时这行代码可能好几天后才跑到 ...

```python

# ============ BREAK IT 演示 ============
class BrokenAlipay:
    # 注意:execute 方法被漏写了!
    pass

def checkout(total, payment):
    return payment.execute(total)

# 创建对象时不报错!(鸭子类型不强制接口)
alipay = BrokenAlipay()
print("创建成功,一切看似正常...")

# 直到运行到这里,才报 AttributeError!
try:
    checkout(99.0, alipay)
except AttributeError as e:
    print(f"报错: {e}")
    print("错误在运行中才暴露,难以追溯!")
# ============ END BREAK IT ============
```

> 诊断:鸭子类型的弱点 -- bug 难追溯。

**结论**:个人脚本用鸭子类型 OK,但团队项目需要更强的保护。
下一节:用 abc.ABC 把'必须实现的方法'写进契约,
让错误提前到实例化时就报错。

> abc.ABC 抽象基类 + abstractmethod

**痛点**:鸭子类型弱在漏写方法不报错,怎么让'漏写'变成'必现'?

**类比**:国家发布 USB-C 接口规范,每厂商必须有 type-c 口。

**解释**:

- abc.ABC 作为父类 -> 类成为抽象基类
- @abstractmethod 装饰的方法 -> 子类必须实现
- 漏写 -> 实例化时报 TypeError

```python

import abc

class Payment(abc.ABC):
    @abc.abstractmethod
    def execute(self, amount):
        ...

# 尝试实例化抽象类 -> 立刻报错!
try:
    p = Payment()
except TypeError as e:
    print(f"报错: {e}")

class Alipay(Payment):
    def execute(self, amount):
        return f"支付宝支付 {amount} 元"

alipay = Alipay()  # OK
print(alipay.execute(99.0))
```

> 逐行解剖

- class Payment(abc.ABC) 成为抽象基类
- @abc.abstractmethod 标记契约
- Payment() -> TypeError
- Alipay(Payment) 实现 execute -> 可实例化

> NCDL Break It #2:忘记继承 abc.ABC

场景:你用了 @abstractmethod,但忘了继承 abc.ABC。
'能跑',但漏写 execute 不报错 -> 鸭子类型的弱点又回来了!

```python

# ============ BREAK IT 演示 ============
import abc

class BrokenPayment:  # 没有 abc.ABC! BREAK IT!
    @abc.abstractmethod
    def execute(self, amount):
        ...

class Alipay(BrokenPayment):
    # 漏写 execute!
    pass

# 竟然不报错!(因为没有 abc.ABC,abstractmethod 是空操作)
alipay = Alipay()
print("没报错!但 execute 不存在,运行时才爆炸")

try:
    alipay.execute(99)
except AttributeError as e:
    print(f"运行时报错: {e}")
# ============ END BREAK IT ============
```

> 结论:@abstractmethod + abc.ABC 必须一起用,
否则装饰器是空操作,等于没有契约。

> 接口概念(团队协作契约)

**痛点**:多人协作写插件,各人写各人,没有统一规范,集成时方法名不一致。

**解释**:

- 接口 = 只包含抽象方法的抽象类
- Python 用'全抽象方法的 ABC'模拟
- 架构师定义接口 -> 开发者各自实现

```python

import abc

class Notifier(abc.ABC):
    @abc.abstractmethod
    def send(self, msg): ...
    @abc.abstractmethod
    def channel(self): ...

class EmailNotifier(Notifier):
    def send(self, msg): return f"邮件: {msg}"
    def channel(self): return "email"

class SMSNotifier(Notifier):
    def send(self, msg): return f"短信: {msg}"
    def channel(self): return "sms"

notifiers = [EmailNotifier(), SMSNotifier()]
for n in notifiers:
    print(f"[{n.channel()}] {n.send('订单已发货')}")
```

**练习**:Serializer 接口

定义接口 Serializer,抽象方法 serialize(obj) 和 deserialize(text)。
子类 JsonSerializer 简单模拟格式。

问自己:

- 接口中可以有普通方法吗?
- 子类只实现一个抽象方法能实例化吗?

```python

# ============ 学员代码区 ============
import abc

class Serializer(abc.ABC):
    # 请定义 serialize 抽象方法
    pass

# class JsonSerializer(Serializer): ...
pass

# ============ 参考答案 ============
import abc

class Serializer(abc.ABC):
    @abc.abstractmethod
    def serialize(self, obj): ...
    @abc.abstractmethod
    def deserialize(self, text): ...

class JsonSerializer(Serializer):
    def serialize(self, obj):
        return "{data: " + str(obj) + "}"
    def deserialize(self, text):
        return text

s = JsonSerializer()
print(s.serialize("hello"))
print(s.deserialize("hi"))
```

> 综合练习:Payment 支付系统(L3 完整落地)

目标:把 Day06 末尾那个有 if-elif 味的系统,重构为多态版本。

要求:

- Payment(abc.ABC) 定义契约
- checkout(cart_total, payment) 不使用 if/elif/isinstance/type
- 漏写 execute 时在实例化阶段就报 TypeError
- 新增支付方式只需添加一个文件

```python

# ============ 学员代码区 ============
import abc

class Payment(abc.ABC):
    # 定义 execute 抽象方法
    pass

class Alipay(Payment):
    pass

class WeChatPay(Payment):
    pass

class ApplePay(Payment):
    pass

def checkout(cart_total, payment):
    # 不超过 4 行,不使用 if/elif/isinstance/type
    pass

payments = [Alipay(), WeChatPay(), ApplePay()]
for p in payments:
    print(checkout(99.0, p))
pass

# ============ 参考答案 ============
import abc

class Payment(abc.ABC):
    @abc.abstractmethod
    def execute(self, amount: float) -> bool:
        ...

class Alipay(Payment):
    def execute(self, amount):
        print(f"支付宝支付 {amount} 元")
        return True

class WeChatPay(Payment):
    def execute(self, amount):
        print(f"微信支付 {amount} 元")
        return True

class ApplePay(Payment):
    def execute(self, amount):
        print(f"Apple Pay {amount} 元")
        return True

def checkout(cart_total, payment):
    if cart_total <= 0:
        print("购物车为空")
        return False
    return payment.execute(cart_total)

# 验证漏写
try:
    class BadPay(Payment):
        pass
    BadPay()
except TypeError:
    print("漏写 execute 会立刻报错!")

for p in [Alipay(), WeChatPay(), ApplePay()]:
    checkout(99.0, p)
```
