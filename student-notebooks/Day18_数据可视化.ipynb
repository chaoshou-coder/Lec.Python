#### 基本图形 —— 四种图形,四种用途

> **痛点**:1000 行数据堆在眼前,看不出趋势、对比、关系、分布。
>
> **类比**:数据可视化 = 给数据"拍照"。
> **折线图**拍"时间流逝",**柱状图**拍"谁高谁低",
> **散点图**拍"两个变量是否相关",**直方图**拍"数据分布"。
>
> **解释**:Matplotlib 是 Python 绘图基础库。
> `plt.plot.bar.scatter.hist` = 四种基本图形。

> **ASCII 用途速查**
> ```
> 折线图 plot            柱状图 bar
>   *  *                   |
>  *    *                 | |
> *      *                | | |
> *        *              | | | |
> 趋势/时间序列           比大小/各类别
>
> 散点图 scatter          直方图 hist
>   *  *                   |
>  *    *                *|*
>   *  *                * | *  *
>    * *                *  |  *  *
> 两个变量的关系          数据分布形态
> ```

````python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5, 6]
sales  = [120, 135, 148, 162, 155, 170]

plt.plot(months, sales)
plt.show()

# --- 执行过程 ---
# 第 1 行 import matplotlib.pyplot as plt:
#   1) 把 pyplot 模块绑定到别名 plt,后续调用 plt.xxx
# 第 3-4 行 months / sales:
#   1) 两个列表长度必须一致,否则 plt 报错
# 第 6 行 plt.plot(months, sales):
#   1) 在内存中的"画布"上画线
#   2) 按 (1,120) -> (2,135) -> ... 顺序连线
#   3) 默认蓝色实线,默认线宽
# 第 7 行 plt.show():
#   1) 把画布渲染到屏幕(弹出窗口)
#   2) 显示完毕后清空画布
````

> **逐行解剖**
> - `plt.plot(x, y)`:x=横坐标列表,y=纵坐标列表,画折线
> - `plt.show()`:必须调用,否则图表不显示
> - Jupyter 可用 `%matplotlib inline` 省略 show

> **常见错误**
> 1. **错误现象**:运行后没有图表弹出
>    **原因**:忘记调用 `plt.show()`
> 2. **错误现象**:`ValueError: x and y must have same ...`
>    **原因**:两个列表长度不一致
> 3. **错误现象**:`NameError: name 'plt' is not defined`
>    **原因**:跳过前面的 import 直接跑后面的 cell

````python
import matplotlib.pyplot as plt
import numpy as np

fruits  = ["苹果", "香蕉", "橙子", "葡萄"]
tonnes  = [350, 280, 210, 160]
height  = np.array([160, 165, 170, 175, 180, 185])
weight  = np.array([52, 58, 65, 70, 76, 82])
scores  = np.random.normal(75, 10, 1000)

plt.bar(fruits, tonnes)         # 柱状图
plt.title("水果产量"); plt.show()

plt.scatter(height, weight)     # 散点图
plt.title("身高 vs 体重"); plt.show()

plt.hist(scores, bins=20)       # 直方图
plt.title("考试成绩分布"); plt.show()

# --- 执行过程 ---
# plt.bar(fruits, tonnes):
#   1) 每个类别(fruits)画一个柱子,高度=对应 tonnes
# plt.scatter(height, weight):
#   1) 每次取 (height[i], weight[i]) 画一个点,不连线
# plt.hist(scores, bins=20):
#   1) 把 scores 分成 20 个区间,统计每个区间的样本数
#   2) 每个区间画一个柱子,高度=样本数
````

#### 多图 subplots —— 一画布里放多张小图

> **痛点**:想看销量的趋势、分布、地区对比,
> 画 3 张图跳出 3 个窗口,来回切换眼花缭乱。
>
> **类比**:subplots = 在 Excel 里把多个图表
> 排版到同一张 sheet,一目了然,不用翻页。
>
> **解释**:`plt.subplots(nrows, ncols, figsize=)`
> 创建"画布 + 轴网格",返回 fig 和 axes 数组,
> 然后用 `axes[i,j].plot(...)` 往每个格子画。

> **ASCII subplots 布局**
> ```
> fig, axes = plt.subplots(2, 2, figsize=(10, 8))
> +------------------+------------------+
> | axes[0, 0]       | axes[0, 1]       |
> | 左上角:折线       | 右上角:柱状       |
> +------------------+------------------+
> | axes[1, 0]       | axes[1, 1]       |
> | 左下角:散点       | 右下角:直方       |
> +------------------+------------------+
> 每个格子用 set_title / set_xlabel 加标注
> ```

````python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.arange(1, 7)
y = [120, 135, 148, 162, 155, 170]

axes[0, 0].plot(x, y, marker="o")
axes[0, 0].set_title("月度销量趋势")

axes[0, 1].bar(x, y, color="orange")
axes[0, 1].set_title("月度销量对比")

axes[1, 0].scatter(x, y, color="red", s=80)
axes[1, 0].set_title("销量散点")

data = np.random.normal(75, 10, 500)
axes[1, 1].hist(data, bins=15, color="green")
axes[1, 1].set_title("成绩分布")

plt.tight_layout()
plt.show()

# --- 执行过程 ---
# plt.subplots(2, 2, figsize=(10, 8)):
#   1) 创建 Figure 对象(画布),绑定到 fig
#   2) 创建 2*2 共 4 个 Axes 对象,绑定到 axes (2D 数组)
#   3) figsize=(10, 8) 指画布宽 10 英寸,高 8 英寸
# axes[0, 0].plot(x, y, marker="o"):
#   1) 在左上角子图里画折线 + 圆点标记
#   2) 注意:子图里用 set_title,不用 plt.title
# plt.tight_layout():
#   1) 自动调整子图间距,避免标题重叠
````

> **逐行解剖**
> - `fig, axes = plt.subplots(r, c)`:返回画布 fig 和轴数组 axes
> - `axes[i, j].plot(...)`:在对应格子里画图
> - `axes[i, j].set_title("...")`:子图专属标题(注意 set_ 前缀)
> - `plt.tight_layout()`:调整间距,放在 show 之前

> **常见错误**
> 1. **错误现象**:子图里 plt.title 加错位
>    **原因**:子图应使用 `axes[i,j].set_title`,不是 plt.title
> 2. **错误现象**:子图挤在一起,标题重叠
>    **原因**:忘记调用 `plt.tight_layout()`
> 3. **错误现象**:axes 索引报 IndexError
>    **原因**:subplots(1, 3) 只有一行,应写作 axes[0]

#### 样式 —— 颜色 / 标记 / 线型

> **痛点**:一张图里画 3 条线,全是默认蓝实线,
> 分不清哪条是"预测",哪条是"实际"。
>
> **类比**:样式 = 给数据"穿上不同颜色的衣服",
> 一条线一个颜色一个标记,legend 一标注,清清楚楚。
>
> **解释**:plt.plot 支持三个样式参数:
> - `color`(或 c):颜色,"r"/"g"/"b"/"k" 或 "#FF5733"
> - `linestyle`(或 ls):"-" 实线 / "--" 虚线 / ":" 点线
> - `marker`:"o" 圆 / "s" 方块 / "*" 星 / "D" 菱形

> **ASCII 样式速查**
> ```
> color:        linestyle:       marker:
>   r 红          - 实线           o 圆点
>   g 绿          -- 虚线          s 方块
>   b 蓝          -. 点划线         * 星形
>   k 黑          : 点线           D 菱形
>   或十六进制    或 "solid"       或 "^" 上三角
> ```

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y_actual  = [2, 4, 6, 8, 10]
y_predict = [2.1, 3.8, 6.5, 7.9, 10.2]

plt.plot(x, y_actual,  color="blue",
         linestyle="-",  marker="o",
         label="实际值")
plt.plot(x, y_predict, color="red",
         linestyle="--", marker="s",
         label="预测值")
plt.legend()
plt.title("实际 vs 预测")
plt.show()

# --- 执行过程 ---
# 第 4 行 plt.plot(..., label="实际值"):
#   1) color="blue" 指定线条蓝色
#   2) linestyle="-" 实线,marker="o" 每个数据点画圆
#   3) label="实际值":记录名字,等下 legend 要用
# 第 8 行 plt.plot(..., label="预测值"):
#   1) 红色虚线方块标记,画在同一个画布上
#   2) 同图多线:plt 自动叠加,直到 show 才一次渲染
# 第 12 行 plt.legend():
#   1) 扫描本画布所有 plot 调用,收集各自的 label
#   2) 在图上画出图例,标注每条线的含义
````

> **逐行解剖**
> - `color="blue"` 或 `c="b"`:线条和标记的颜色
> - `linestyle="--"` 或 `ls="--"`:实线 / 虚线 / 点划线
> - `marker="o"`:在数据点上画圆 / 方块 / 星形
> - `label="..."`:记录标签名,配合 `plt.legend()` 显示

> **常见错误**
> 1. **错误现象**:legend 只显示其中一条线
>    **原因**:某次 plot 忘写 label 参数
> 2. **错误现象**:图例挤在数据上方,挡住了线
>    **原因**:可改 loc="lower right" 固定位置
> 3. **错误现象**:颜色没变
>    **原因**:`color` 写成 `colour`(英式拼写),Python 不认

#### 标签和标题 —— 让图表"自己说话"

> **痛点**:画了一张柱状图,同事看不懂 ——
> 横轴是啥?纵轴单位是什么?
> 没有标注的图表 = 无字天书。
>
> **类比**:图表 = 一本书,title 是书名,
> xlabel / ylabel 是章节标题,legend 是图例注解。
>
> **解释**:plt 提供 4 个函数加文字:
> - `plt.title("标题", fontsize=14)`
> - `plt.xlabel("X 轴含义")`
> - `plt.ylabel("Y 轴含义")`
> - `plt.legend()` 显示 plot 里的 label
> - `plt.grid(True)` 画参考网格

> **ASCII 标注位置**
> ```
>          plt.title("月度销量")
>                  |(居中)
>   y   |         +--------------+
>   p   | 纵  |                   |
>   l   | 轴  |    图表区域       |
>   o   | 标  |                   |
>   t   | 签  +-------------------+
>                  x axis label
> ```

````python
import matplotlib.pyplot as plt

month    = ["Jan", "Feb", "Mar", "Apr", "May"]
beijing  = [2.1, 2.3, 2.5, 2.4, 2.6]
shanghai = [2.4, 2.5, 2.6, 2.7, 2.8]

plt.plot(month, beijing,  marker="o", label="北京")
plt.plot(month, shanghai, marker="s", label="上海")

plt.title("2024 年上半年 GDP 对比", fontsize=16)
plt.xlabel("月份", fontsize=12)
plt.ylabel("GDP(万亿元)", fontsize=12)
plt.grid(True)
plt.legend(loc="lower right")
plt.show()

# --- 执行过程 ---
# plt.title("2024 年上半年 GDP 对比", fontsize=16):
#   1) 在画布顶部居中显示标题,字号 16 磅
# plt.xlabel / ylabel:
#   1) 在横轴 / 纵轴下方 / 左侧显示文字
# plt.grid(True):
#   1) 在图表区域画水平/垂直参考线,方便读数
# plt.legend(loc="lower right"):
#   1) 在右下角画图例,区分北京线和上海线
````

> **逐行解剖**
> - `plt.title(s, fontsize=n)`:标题文字 + 字号(磅)
> - `plt.xlabel(s)` / `plt.ylabel(s)`:横 / 纵轴说明
> - `plt.grid(True)`:显示网格(False 则关闭)
> - `plt.legend(loc=...)`:loc 取 "best"/"center"/"lower right"

#### 保存图片 —— 从屏幕到文件

> **痛点**:图表弹出来了,想插入 PPT / 报告,
> 只能用截图,模糊且麻烦。
>
> **类比**:savefig = 给图表"冲印照片"。
> 可以指定分辨率 dpi 决定清晰度。
>
> **解释**:`fig.savefig(path, dpi=300, bbox_inches="tight")`
> - `dpi`:每英寸点数,300 = 印刷级清晰度
> - `bbox_inches="tight"`:裁剪空白边框
> - 顺序:先 savefig,再 show(show 会清空画布)

> **ASCII savefig 参数**
> ```
> fig.savefig(
>     "output.png",        # 文件名,支持 png / jpg / svg / pdf
>     dpi=300,             # 分辨率:印刷 300,屏幕 100
>     bbox_inches="tight", # 裁剪空白边
> )
> 顺序:先 savefig → 再 show
> ```

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, color="purple")
ax.set_title("sin(x)")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")

# 若想存盘,取消下一行注释后运行
# fig.savefig("sin.png", dpi=150, bbox_inches="tight")
print("取消上一行注释即可保存 sin.png")
plt.show()

# --- 执行过程 ---
# fig.savefig("sin.png", dpi=150, bbox_inches="tight"):
#   1) 把当前 fig 渲染为 PNG 文件,存到当前工作目录
#   2) dpi=150:每英寸 150 像素点,屏幕展示足够清晰
#   3) bbox_inches="tight":裁剪空白边框,不留白边
# 注意:savefig 必须放在 show 之前
#   因为 plt.show() 执行后会清空画布,
#   之后 savefig 只会保存一张空白图
````

> **逐行解剖**
> - `fig.savefig(path, dpi=N)`:fig 是 plt.subplots 返回值

#### Seaborn 集成 —— 更美观的高级图表

> **痛点**:Matplotlib 默认样式朴素,
> 画箱线图 / 热力图 / 多变量散点矩阵,

> **类比**:Seaborn = Matplotlib 的"美颜滤镜",
> 更少的代码,更美观的默认色彩,
> 且与 Pandas DataFrame 无缝衔接。
>
> **解释**:`import seaborn as sns`,常用函数:
> - `sns.boxplot(x=, y=, data=df)` 箱线图
> - `sns.heatmap(data, annot=True)` 热力图
> - `sns.pairplot(df, hue=)` 多变量两两散点矩阵

> **ASCII Seaborn vs Matplotlib**
> ```
> Matplotlib:              Seaborn:
> plt.bar(x, y)            sns.barplot(x=, y=, data=df)
> plt.hist(...)            sns.boxplot(x=, y=, data=df)
> 手动调样式 20 行          sns.heatmap(corr, annot=True)

#### Pandas 直接 .plot() —— DataFrame 原生可视化

> **痛点**:Pandas 里处理好的 DataFrame,
> 还要手动抽列丢给 plt.plot,来回转换麻烦。
>
> **类比**:Pandas .plot = DataFrame 自己画笔 ,
> 不用搬数据给 Matplotlib,直接调方法出图。
>
> **解释**:DataFrame / Series 内置 plot 方法:
> - `df.plot()` 默认折线图
> - `df.plot.bar()` 柱状图
> - `df.plot.scatter(x=, y=)` 散点图
> - `df.plot.hist()` 直方图
> 底层调用 Matplotlib,所以 plt.show 仍需要。

> **ASCII df.plot() 家族**
> ```
> df.plot()            折线图(df.plot.line 同)
> df.plot.bar()        柱状图
> df.plot.barh()       横向柱状图
> df.plot.scatter(x, y)散点图(必须指定 x, y)
> df.plot.hist()       直方图
> df.plot.box()        箱线图
> df.plot.kde()        核密度估计曲线
> ```

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Q1": [120, 200, 90],
    "Q2": [135, 210, 95],
    "Q3": [148, 220, 100],
    "Q4": [162, 240, 110],
}, index=["苹果", "香蕉", "橙子"])

# DataFrame 直接画折线图(x 轴 = 列名 Q1~Q4)
df.plot()
plt.title("季度销量趋势")
plt.ylabel("销量(吨)")
plt.show()

# --- 执行过程 ---
# df.plot():
#   1) Pandas 检测到 DataFrame,默认画折线图
#   2) 行索引(["苹果","香蕉","橙子"]) → 图例
#   3) 列名(Q1~Q4) → x 轴刻度
#   4) 每行一条线,自动分配颜色
# plt.title / ylabel:
#   1) Pandas 的 plot 底层就是 Matplotlib,可以直接用
#   2) plt 给当前画布加标题和轴标签
````

> **逐行解剖**
> - `df.plot()`:默认折线图,x 轴 = 列名,每行一条线
> - `df.plot.bar()`:柱状图,每行一组柱子
> - `df.plot.scatter(x, y)`:散点图,必须指定两列
> - `df["col"].hist()`:单列画直方图
> - 所有 Pandas plot 都返回 matplotlib.axes.Axes

> **常见错误**
> 1. **错误现象**:`ValueError: scatter requires x and y`
>    **原因**:scatter 忘写 x= 和 y= 参数
> 2. **错误现象**:柱子被遮挡看不清
>    **原因**:列数太多,应加 figsize=(10, 5) 放大画布
> 3. **错误现象**:df.plot 后没显示
>    **原因**:Jupyter 里 Pandas 自动 show,脚本里需 plt.show()
