#### matplotlib 基础(折线 / 柱状 / 散点 / 直方图)

痛点:数据处理完了,一堆数字摆眼前,看不出趋势、比不了大小。

类比:同样是一份销售报表,老板看表格 3 分钟看不出来,看折线图一眼就能说"这个月下滑了"。

解释:Matplotlib 的固定套路是 **"准备数据 -> 调用绘图函数 -> 添加标注 -> plt.show() 显示"**。四种基本图形各有适用场景:折线看趋势、柱状比大小、散点看关系、直方看分布。


```python
# 导入 matplotlib 的 pyplot 模块,约定别名 plt
import matplotlib.pyplot as plt

# 固定数据,不要读文件,方便教学
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# --- 演练折线图 ---
# plot(x, y):逐点连线
plt.plot(x, y)
# title / xlabel / ylabel:图表必备三标
plt.title("折线图示例")
plt.xlabel("X 轴")
plt.ylabel("Y 轴")
# show:非 Jupyter 环境下必须调这起显示,否则啥也不出现
plt.show()
```

# --- 执行过程 ---
# ① plt.plot(x, y):把 (1,2) (2,4) ... (5,10) 逐点用线段连接
# ② plt.title / xlabel / ylabel:给图形加上文字说明
# ③ plt.show():把画布渲染出来,弹出窗口(或在 Jupyter 中直接显示)


**逐行解剖**

- `plt.plot(x, y)`:最基本的一笔画,参数 x 为横坐标、y 为纵坐标,长度必须一致。
- `plt.title("折线图示例")`:图表的"大字号",告诉读者这张图在讲什么。
- `plt.xlabel` / `plt.ylabel`:给两根轴命名,"是什么"、"带不带单位"靠这两行。
- `plt.show()`:在脚本中必须调用;在 Jupyter 里配合 `%matplotlib inline` 会自动显示,但显式调用更保险。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# 数据准备
categories = ["苹果", "香蕉", "橙子", "葡萄"]
values = [30, 25, 40, 35]

# --- 演练柱状图 ---
# bar(x, height):每个柱子占据一个类别
plt.bar(categories, values)
plt.title("水果销量")
plt.xlabel("种类")
plt.ylabel("销量(吨)")
plt.show()
```

# --- 执行过程 ---
# ① plt.bar(categories, values):x 是类别,y 是值,每个柱子一个刻度
# ② 柱状图天然适合"类别 vs 数量"的直观对比


**逐行解剖**

- `plt.bar()`:柱状图,默认蓝柱子。适合"不同类别对比",比折线图更适合离散类别。
- `plt.barh()`:水平柱状图,类别名称很长时用它更美观。
- color 参数可以指定颜色,例如 `plt.bar(categories, values, color="orange")`。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# 数据:学习时长 vs 考试分数
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
scores = [50, 55, 65, 70, 75, 80, 88, 92]

# --- 演练散点图 ---
# scatter(x, y):只画点不连线,观察相关性
plt.scatter(study_hours, scores)
plt.title("学习时长与分数关系")
plt.xlabel("学习时长(小时)")
plt.ylabel("考试分数")
plt.show()
```

# --- 执行过程 ---
# ① plt.scatter(study_hours, scores):每组 (x, y) 画一个散点
# ② 散点图上点从左下到右上分布,说明两者正相关


**逐行解剖**

- `plt.scatter()`:散点图,用于观察"两个连续变量之间是不是有关系"。点的分布往往能直观看出正相关、负相关或无相关。
- `s` 参数控制点的大小,`c` 控制颜色,`alpha` 控制透明度。例如 `plt.scatter(x, y, s=100, c="red", alpha=0.5)`。
- 散点图 + 趋势线 = 回归分析的经典可视化入门。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# 固定一组成绩数据
grades = [
    55, 61, 64, 70, 72, 75, 78, 80, 81, 83,
    84, 85, 85, 88, 88, 90, 91, 92, 93, 95
]

# --- 演练直方图 ---
# hist(x, bins):把数据切成若干区间,统计落在每个区间的数量
plt.hist(grades, bins=5, edgecolor="black")
plt.title("成绩分布")
plt.xlabel("分数区间")
plt.ylabel("人数")
plt.show()
```

# --- 执行过程 ---
# ① plt.hist(grades, bins=5):把 55~95 均分成 5 个区间
# ② 自动统计每个区间有多少个成绩,形成"分布"概览
# ③ edgecolor="black":给每个柱子描边,视觉更清晰


**逐行解剖**

- `plt.hist(x, bins)`:直方图,专用于观察"一个连续变量的分布"。`bins` 是区间数量,越大柱子越多、越细。
- 直方图 不等于 柱状图:柱状图是每个类别一个柱子,直方图是把数值分桶后画柱。
- `edgecolor` 建议设成 "black" 或 "white",否则相邻柱子粘在一起。

#### subplots(2x2 + tight_layout + savefig)

痛点:报告要放好几张图,每次都弹一个窗口,汇报时频繁切换太麻烦。

类比:subplots 把画布切成 N 个格子,每个格子放一张小图,就像"九宫格朋友圈"。

解释: **plt.subplots(nrows, ncols)** 一次性创建多个子图,**axes[i, j]** 用来定位每个格子,**tight_layout()** 自动调整子图间距,**savefig()** 把整个画布保存成图片。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# --- 演练 2x2 子图 ---
# fig:代表整张画布;axes:二维数组,对应每个子图
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# 第 1 个子图(左上):折线图
axes[0, 0].plot([1, 2, 3], [1, 4, 9])
axes[0, 0].set_title("折线图")

# 第 2 个子图(右上):柱状图
axes[0, 1].bar(["A", "B", "C"], [3, 7, 2])
axes[0, 1].set_title("柱状图")

# 第 3 个子图(左下):散点图
axes[1, 0].scatter([1, 2, 3, 4], [2, 3, 5, 7])
axes[1, 0].set_title("散点图")

# 第 4 个子图(右下):直方图
axes[1, 1].hist([1, 1, 2, 2, 2, 3], bins=3)
axes[1, 1].set_title("直方图")

# tight_layout:自动调整子图间距,防止重叠
plt.tight_layout()
# 保存图片:savefig 必须在 show 之前,否则会存白图
plt.savefig("subplots_demo.png", dpi=150, bbox_inches="tight")
plt.show()
```

# --- 执行过程 ---
# ① plt.subplots(2, 2, figsize=(8, 6)):创建 2x2 网格,画布 8x6 英寸
# ② axes[0, 0]:取第 0 行第 0 列的子图,在这个画布上 draw
# ③ axes[0, 0].set_title("折线图"):子图模式下必须用 set_title,不能再用 plt.title
# ④ plt.tight_layout():自动压缩边距,让 4 张图不互相压
# ⑤ plt.savefig:画布落盘,必须在 show 前面(show 后会清空画布)


**逐行解剖**

- `figsize=(宽, 高)`:单位英寸,常用 (8, 6)、(10, 8)。屏幕展示 dpi=100 够用,报告/印刷 dpi=300。
- `axes[i, j]`:子图对象。**在这上面画图用 axes[i, j].plot,不能用 plt.plot**。
- `set_title` / `set_xlabel` / `set_ylabel`:子图模式下"plt."全部换成"axes[i,j]."。
- `plt.tight_layout()`:必加,不加容易子图标题和子图重叠。
- `plt.savefig(path, dpi, bbox_inches="tight")`:**务必在 show 之前**,否则保存空白图。`bbox_inches="tight"` 裁掉多余白边。

#### seaborn(箱线图 / 热力图)

痛点:Matplotlib 默认样式朴素,画箱线图和热力图需要手工调,代码又长又丑。

类比:Seaborn 相当于 Matplotlib 的"美颜相机",几行代码就能出好看的统计图表,而且和 Pandas DataFrame 完美配合。

解释:**sns.boxplot** 画箱线图看分布和异常值,**sns.heatmap** 画热力图看相关性矩阵。Seaborn 画图后仍可以用 Matplotlib 加标题、保存图片。


```python
# 导入 seaborn,别名 sns
import seaborn as sns
# 导入 pyplot 用于加标题和显示
import matplotlib.pyplot as plt
# 导入 pandas 构造数据
import pandas as pd

# 构造三班的成绩数据,每班 5 人
data = pd.DataFrame({
    "班级": ["一班"] * 5 + ["二班"] * 5 + ["三班"] * 5,
    "分数": [80, 85, 90, 75, 88,
             70, 78, 82, 65, 75,
             90, 92, 95, 88, 93]
})

# --- 演练箱线图 ---
# boxplot:箱内是 Q1~Q3,中线是中位数,触须外的点是异常候选值
sns.boxplot(x="班级", y="分数", data=data)
plt.title("各班级成绩箱线图")
plt.show()
```

# --- 执行过程 ---
# ① pd.DataFrame 构造班级+分数长表(Seaborn 喜欢长表+列名)
# ② sns.boxplot(x="班级", y="分数", data=data):x 是类别,y 是数值
# ③ 每个箱体表示一个班级,可直观看出三班整体分数更高


**逐行解剖**

- `sns.boxplot(x=, y=, data=)`:Seaborn 推荐全程用列名 + DataFrame,直观且避免错传。
- 箱体解读:下边缘 Q1,中线 median,上边缘 Q3;上触须=1.5xIQR 范围内最大值,之外是离群点。
- `sns.violinplot`:箱线的升级版,叠加了概率密度,数据量大时更好看。


```python
# 导入 seaborn 和 pyplot
import seaborn as sns
import matplotlib.pyplot as plt
# 导入 pandas
import pandas as pd

# 构造相关性矩阵:3 科之间的两两相关系数
df_corr = pd.DataFrame({
    "语文": [1.0, 0.8, 0.3],
    "数学": [0.8, 1.0, 0.5],
    "英语": [0.3, 0.5, 1.0]
}, index=["语文", "数学", "英语"])

# --- 演练热力图 ---
# annot=True:在格子里面写上具体数字
# cmap:颜色映射,数值越大颜色越深
sns.heatmap(df_corr, annot=True, cmap="Blues")
plt.title("学科相关性热力图")
plt.show()
```

# --- 执行过程 ---
# ① df_corr 是对称矩阵,对角线恒为 1
# ② sns.heatmap:把数值映射成颜色深浅
# ③ annot=True:叠加数字,让热力图兼顾"看图"和"看数" 


**逐行解剖**

- `annot=True`:每格里写数值,论文/报告必备;数据量大时可关掉避免拥挤。
- `cmap="Blues"`:蓝由浅到深,相关系数越高越深。相关性矩阵常用 `cmap="coolwarm"`(负蓝正红)。
- `linewidths=0.5`:在格子之间描细线,矩阵更大时更清晰。
- Seaborn 画完之后仍可 plt.title / plt.xlabel / plt.savefig,两种库可以混着用。

#### 中文乱码处理

痛点:图表的标题、轴标签一旦写中文,默认会变成小方块,根本没法看。

类比:Matplotlib 默认只认识英文,给中文它就像看到外星文,必须"装中文字体"才能显示。

解释:通过 **plt.rcParams** 全局指定中文字体和负号处理。Mac 用 "Arial Unicode MS",Windows 用 "SimHei"。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# --- 演练:先出问题 ---
# 没有设置字体,中文会变成方框
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("没有设置字体的中文标题")
plt.show()
```

# --- 执行过程 ---
# ① Matplotlib 默认字体列表里没有中文字体
# ② 渲染时找不到对应字形,就用"豆腐块"(方框)占位


**逐行解剖**

- 乱码起因:Matplotlib 的默认字体 (DejaVu Sans) 不包含中文。
- 两类错误:一是中文变方块,二是负号 "-" 也变方块(影响数字可读性)。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# --- 演练:Mac 系统中文设置 ---
# font.sans-serif:英文字体列表,按顺序挑一个本机已安装的
plt.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
# axes.unicode_minus:不再用 Unicode 的减号,避免负号也变框
plt.rcParams["axes.unicode_minus"] = False

# 含中文标题的折线图
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("折线图:每月销量")
plt.xlabel("月份")
plt.ylabel("销量(万元)")
plt.show()
```

# --- 执行过程 ---
# ① plt.rcParams["font.sans-serif"] = ["Arial Unicode MS"]:全局指定 Mac 中文字体
# ② plt.rcParams["axes.unicode_minus"] = False:负号显示为拉丁短横,不再依赖 Unicode 减号
# ③ 接下来所有标题、标签、图例中的中文都能正常显示


**逐行解剖**

- `plt.rcParams["font.sans-serif"] = ["Arial Unicode MS"]`:Mac 自带该字体,跨平台发布时注意 Windows 没有该字体,Windows 要改成 `["SimHei"]` 或 `["Microsoft YaHei"]`。
- `plt.rcParams["axes.unicode_minus"] = False`:防止 "-10" 里的负号显示为方块。
- 这些设置是**全局的**,在 notebook 顶部设置一次即可,后续所有 cell 都受益。
- 最佳实践:把它们和 `import matplotlib.pyplot as plt` 写在 notebook 第一个代码 cell。


```python
# 导入 pyplot
import matplotlib.pyplot as plt

# 全局中文字体设置
plt.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
plt.rcParams["axes.unicode_minus"] = False

# --- 演练 seaborn + 中文标题 ---
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    "科目": ["数学", "语文", "英语", "物理"],
    "平均分": [78, 82, 75, 80]
})

# seaborn 画图,matplotlib 加标题
sns.barplot(x="科目", y="平均分", data=data)
plt.title("高一各班平均成绩")
plt.xlabel("科目")
plt.ylabel("平均分")
plt.show()
```

# --- 执行过程 ---
# ① 顶部 rcParams 设置一次, seaborn 画图时自动继承
# ② sns.barplot 画柱状图
# ③ plt.title 加中文标题,不会乱码


**逐行解剖**

- Seaborn + Matplotlib 可混用:Seaborn 画主体,Matplotlib 加后续装饰(title、label、savefig)。
- 注意:**plt.title() 只对当前子图有效**,多子图时要用 `axes[i, j].set_title()`。
- 跨平台部署:把 `["Arial Unicode MS"]` 改成 `["SimHei"]` 或改用 `FontProperties` 局部指定,避免中文字体不存在报错。

```python
# 参考答案
import matplotlib.pyplot as plt

# 1. 顶部设置中文字体
plt.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
plt.rcParams["axes.unicode_minus"] = False

# 2. 硬编码数据
categories = ["苹果", "香蕉", "橙子", "葡萄"]
values = [30, 25, 40, 35]

# 3. 画柱状图
plt.bar(categories, values, color="steelblue")
plt.title("各水果销量")
plt.xlabel("水果种类")
plt.ylabel("销量(吨)")

# 4. 保存图(必须在 show 之前)
plt.savefig("fruit_sales.png", dpi=150, bbox_inches="tight")
plt.show()
```
