#### 为什么需要函数(DRY 原则)

> **痛点**
代码里出现 5 段长得一样的代码,改 bug 要改 5 处。复制粘贴越多,越痛。

> **类比**
函数 = 咖啡机:放进去原料(参数),按按钮(调用),出来咖啡(返回值)。不用每次从头磨豆子。

> **解释**
函数把重复逻辑**打包成一个名字**,定义一次,调用无数次。核心原则:DRY = Dont Repeat Yourself。

```python
# 反例:复制粘贴
s = "  hello  "
s = s.strip()
s = s.upper()
print(s)

s = "  world  "
s = s.strip()
s = s.upper()
print(s)

s = "  python  "
s = s.strip()
s = s.upper()
print(s)
```

```python
# --- 执行过程 ---
# 第 1-4 行:处理 "  hello  ":
#   ① s = "  hello  " 赋值
#   ② s.strip() 删除两端空白 -> "hello",重新赋值给 s
#   ③ s.upper() 转大写 -> "HELLO",重新赋值给 s
#   ④ print(s) 输出: HELLO
#
# 第 5-8 行:处理 "  world  ":
#   ① 同样的 3 步又要重写一遍
#   ② 如果需求变更(如要加 .replace),得改 5 处
#
# 改进:把重复逻辑打包成函数,改一处就行
```

> **逐行解剖**
- 3 段逻辑完全一样,只是字符串不同
- 这就是"重复代码",应该用函数封装

> **常见错误**
1. **觉得复制粘贴方便**:短期方便,长期维护难
2. **函数写太长**:一个函数应该只做一件事

> 问自己:

> - 如果逻辑更复杂(10 步),改一处 vs 改十处?
> - 函数能解决所有重复问题吗?

```python
# ============ 学员代码区 ============
# 把上面的 3 段重复代码改成函数,只需改函数内部一次

pass

# ============ 参考答案 ============
def format(s):
    s = s.strip()
    s = s.upper()
    return s

print(format("  hello  "))
print(format("  world  "))
print(format("  python  "))
```

#### def 定义和调用 - 给一段代码起个名字

> **解释**
`def 函数名(参数):` 定义函数,不会立刻执行。
`函数名(参数)` 触发调用。

```python
def say_hi():
    print("Hello!")
    print("欢迎学习 Python")

say_hi()
say_hi()
say_hi()
```

```python
# --- 执行过程 ---
# 第 1-3 行 def say_hi():
#   ① Python 把函数体打包,挂到名字 say_hi 上
#   ② 这 3 行并不立刻执行
#   ③ 只是"注册"这个函数
#
# 第 5 行 say_hi():
#   ① Python 看到函数名 + 括号,触发调用
#   ② 跳到 def 函数体,逐行执行
#   ③ 打印 Hello! 和 欢迎学习 Python
#   ④ 函数体执行完,回到调用处
#
# 第 6-7 行:同理,调用几次执行几次
def say_hi():
    print("Hello!")

say_hi()
```

> **逐行解剖**
- `def` 关键字 + 函数名 + 括号 + 冒号
- 函数体缩进(4 空格)
- 调用:函数名 + 括号,触发执行

> **常见错误**
1. **忘调用**:写了 def 不等于"跑一次",必须调用
2. **忘冒号**:`def say_hi()` → SyntaxError
3. **缩进错**:函数体不缩进 → IndentationError

#### 函数四种形式

| 形式 | 参数 | 返回值 | 使用频率 |
| --- | --- | --- | --- |
| 无参无返回 | 无 | 无 | 中(打印菜单) |
| 有参无返回 | 有 | 无 | 中(打印定制文案) |
| 无参有返回 | 无 | 有 | 低 |
| **有参有返回** | 有 | 有 | **最高** |

> **解释**
第 4 种 (有参有返回) 最常用!其他都是它的特例。

```python
def greet():
    print("Hello!")

def greet_name(name):
    print(f"你好,{name}")

def get_pi():
    return 3.14159

def add(a, b):
    return a + b

greet()
greet_name("小明")
pi = get_pi()
result = add(3, 5)
print(pi, result)
```

```python
# --- 执行过程 ---
# 第 9 行 pi = get_pi():
#   ① 调用 get_pi() → 函数体执行
#   ② return 3.14159 → 把 3.14159"送回来"
#   ③ 3.14159 赋值给 pi
#
# 第 10 行 result = add(3, 5):
#   ① 调用 add(3, 5)=Python 调用 -> a=3, b=5
#   ② return a + b -> return 8
#   ③ result = 8
def greet(): pass
def greet_name(name): pass
def get_pi(): return 3.14159
def add(a, b): return a + b
pi = get_pi()
result = add(3, 5)
print(pi, result)
```

#### 形参/实参/默认参数/关键字参数

> **痛点**
- 形参是"坑位",实参是"值"
- 默认参数让某些参数"可选填"
- 关键字参数让顺序无关

```python
def introduce(name, age, city):
    print(f"{name},{age} 岁,来自 {city}")

# 位置传参(按位置)
introduce("小明", 18, "北京")

# 关键字参数(按名字,顺序无关)
introduce(age=20, city="上海", name="小红")

# 默认参数
def greet(name="学员"):
    print(f"你好,{name}")

greet()
greet("张三")
```

```python
# --- 执行过程 ---
# 第 4 行 introduce("小明", 18, "北京"):
#   ① "小明" 赋给形参 name
#   ② 18 赋给形参 age
#   ③ "北京" 赋给形参 city
#   ④ 执行函数体,打印"小明,18 岁,来自 北京"
#
# 第 7 行 introduce(age=20, city="上海", name="小红"):
#   ① 关键字参数按名字传,不靠顺序
#   ② name="小红",age=20,city="上海"
#   ④ 打印"小红,20 岁,来自 上海"
#
# 第 11 行 greet():
#   ① name 没有传参,使用默认值 "学员"
#   ② 打印"你好,学员"
def introduce(name, age, city):
    print(f"{name},{age} 岁,来自 {city}")
introduce("小明", 18, "北京")
introduce(age=20, city="上海", name="小红")

def greet(name="学员"):
    print(f"你好,{name}")
greet()
greet("张三")
```

> **常见错误**
1. **默认参数放前面**:`def f(a=1, b)` SyntaxError
2. **位置参数在关键字参数后**:`f(a=1, 2)` SyntaxError
3. **mutable 默认参数**:None

#### return - 带结果回家 + 立刻退出

> **解释**
`return` 把结果送回调用处,同时立刻退出函数(后面代码不执行)。

```python
def check_age(age):
    if age < 0:
        return "年龄不能为负"
    if age < 18:
        return "未成年"
    return "成年"

print(check_age(20))
print(check_age(15))
print(check_age(-5))
```

```python
# --- 执行过程 ---
# check_age(20):
#   ① age=20
#   ② 20 < 0? → False,跳过
#   ③ 20 < 18? → False,跳过
#   ④ return "成年" → 立刻退出(函数结束)
#
# check_age(15):
#   ① age=15
#   ② 15 < 0? → False
#   ③ 15 < 18? → True → return"未成年"立刻退出
#   ④ "return "成年""永远不会执行

def check_age(age):
    if age < 0: return "年龄不能为负"
    if age < 18: return "未成年"
    return "成年"
print(check_age(20))
print(check_age(15))
```

> **常见错误**
1. **以为 return 后还执行**:return 立刻退出,后面是死代码
2. **return 无值**:return None,函数无返回值
3. **忘记 print 返回值**:需要 `print(f(x))` 才能看到结果

#### 多值返回 - 元组解包

> **解释**
`return a, b, c` 本质返回元组,用逗号变量接收叫"解包"。

```python
def calculate(a, b):
    return a + b, a - b, a * b

sum_, diff, prod = calculate(10, 3)
print(sum_, diff, prod)
```

```python
# --- 执行过程 ---
# 第 4 行 sum_, diff, prod = calculate(10, 3):
#   ① 先调用 calculate(10, 3)
#   ② 函数返回 (13, 7, 30) 元组
#   ③ 元组解包:sum_=13, diff=7, prod=30
#
# 本质:return 13, 7, 30 == return (13, 7, 30)
def calculate(a, b):
    return a + b, a - b, a * b
sum_, diff, prod = calculate(10, 3)
print(sum_, diff, prod)
```

#### 作用域 - 全局 vs 局部

> **解释**
- 全局变量:函数外定义,函数内可读
- 局部变量:函数内定义,函数外不可见
- 函数内不能直接修改全局变量(除非 global)

```python
x = 10

def foo():
    y = 20
    print(x)
    print(y)

foo()
# print(y) 报错!y 是局部变量
```

```python
# --- 执行过程 ---
# 第 1 行 x = 10:
#   ① 创建全局变量 x = 10
#
# 第 4 行 y = 20:
#   ① 创建局部变量 y = 20(只在函数内有效)
#
# 第 5 行 print(x):
#   ① 函数内没有局部 x
#   ② 沿局部 -> 全局查找,找到全局 x = 10
#   ③ 输出: 10
#
# 第 8 行 foo() 执行完:
#   ① 局部变量 y 被销毁
#   ② print(y) 会报 NameError
x = 10
def foo():
    y = 20
    print(x)
    print(y)
foo()
```

> **常见错误**
1. **函数内改全局变量**:`x += 1` 报 UnboundLocalError
2. **以为局部变量函数外能用**:函数结束局部变量销毁

#### 综合练习 - 工具函数库

定义 5 个工具函数:
1. `is_valid_phone(phone)` -> 返回 bool(11 位且全数字)
2. `mask_phone(phone)` -> 返回 "138****8000"
3. `bmi_report(h, w)` -> 返回评级字符串
4. `is_leap(year)` -> 返回 bool
5. `grade(score)` -> 返回 "A/B/C/D/F"

```python
# ============ 学员代码区 ============

pass

# ============ 参考答案 ============
def is_valid_phone(phone):
    return len(phone) == 11 and phone.isdigit()

def mask_phone(phone):
    return phone[:3] + "****" + phone[-4:]

def bmi_report(h, w):
    bmi = w / (h ** 2)
    if bmi < 18.5: return "偏瘦"
    if bmi < 24: return "正常"
    if bmi < 28: return "偏胖"
    return "肥胖"

def is_leap(year):
    return (year % 4 == 0 and year % 100 != 0) or year % 400 == 0

def grade(score):
    if score >= 90: return "A"
    if score >= 80: return "B"
    if score >= 70: return "C"
    if score >= 60: return "D"
    return "F"

print(is_valid_phone("13812345678"))
print(mask_phone("13812345678"))
print(bmi_report(1.75, 70))
print(is_leap(2024))
print(grade(85))
```
