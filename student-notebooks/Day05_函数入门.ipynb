## 第一讲(15 分钟)—— `def` 定义与四种形式

### 函数是什么:给一段代码起个名字

**函数 = 工具 + 说明书**。定义时写一次,调用时用无数次。

```
def 函数名(参数):
    函数体
    return 返回值
```

**口诀**: 定义时打包,调用时拆包。没有调用,函数里面的代码不会执行。

````python
# 示例代码:定义和调用函数
def greet():                    # def + 函数名 + 括号 + 冒号
    print("Hello!")             # 函数体缩进
    print("欢迎学习 Python")

greet()                         # 调用:函数名 + 括号
greet()                         # 再调用一次
````

### 函数四种形式

| 形式 | 参数 | 返回值 | 例子 |
|---|---|---|---|
| ① 无参无返回 | 无 | 无 | `print_hello()` |
| ② 有参无返回 | 有 | 无 | `print_name("小明")` |
| ③ 无参有返回 | 无 | 有 | `get_today()` |
| ④ 有参有返回 | 有 | 有 | `add(1, 2)` |

**重点掌握 ④**(最高频),其他都会退化到 ④ 的组合。

````python
# 示例代码:四种形式
def say_hi():                   # ① 无参无返回
    print("Hi!")

def greet(name):                # ② 有参无返回
    print(f"你好,{name}")

def get_pi():                   # ③ 无参有返回
    return 3.14159

def add(a, b):                  # ④ 有参有返回
    return a + b

# 调用
say_hi()                        # Hi!
greet("小明")                   # 你好,小明
pi = get_pi()
c = add(3, 5)
print(c)                        # 8
````

## 第二讲(15 分钟)—— 形参/实参/默认参数

### 形参 vs 实参

- **形参**: 函数定义时写在括号里的"坑位"(占位符)
- **实参**: 函数调用时塞进去的"值"

````python
# 示例代码:形参和实参
def greet(name):        # name 是"形参"(形式参数)
    print(f"你好,{name}")

greet("小明")           # "小明" 是"实参"(实际参数)
greet("小红")           # 不同实参,同一段逻辑
````

### 默认参数:有些参数"可选填"

````python
# 示例代码:默认参数
def greet(name="学员"):     # name 默认值为 "学员"
    print(f"你好,{name}")

greet()                      # 你好,学员(用默认值)
greet("小明")                # 你好,小明(覆盖默认值)
````

> 🔴 **教学红线**: 默认参数后面不能再跟非默认参数,否则 `SyntaxError`!

```python
def greet(name="学员", age):  # ❌ 错误!
def greet(age, name="学员"):  # ✅ 正确!
```

### 关键字参数:按名字传参

````python
# 示例代码:关键字参数
def introduce(name, age, city):
    print(f"{name},{age} 岁,来自 {city}")

introduce(age=18, city="成都", name="小明")  # 顺序无关
````

## 第三讲(15 分钟)—— `return` 与作用域

### `return`:带一个结果回家 + 立刻退出

`return` 像"放学铃" —— 听到就停,后面代码不执行。

````python
# 示例代码:return 提前退出
def check_age(age):
    if age < 0:
        return "年龄不能为负"    # 立即退出,后面语句不再执行
    if age < 18:
        return "未成年"
    return "成年"

print(check_age(20))            # 成年
print(check_age(-1))            # 年龄不能为负
````

### 返回多个值:元组解包

Python 的"返回多个值"本质是**返回一个元组**,用逗号分隔的变量"解包"。

````python
# 示例代码:多值返回
def calculate(a, b):
    return a + b, a - b, a * b  # 返回元组 (和, 差, 积)

sum_, diff, prod = calculate(10, 3)
print(sum_, diff, prod)         # 13 7 30
````

### 作用域:变量在函数里/外

**口诀**: 函数里能读外面的,不能直接改外面。

````python
# 示例代码:作用域
x = 10              # 全局变量:函数外都能用

def foo():
    y = 20          # 局部变量:只在函数内有效
    print(x)        # ✅ 能访问全局

foo()
# print(y)          # ❌ 报错:函数外访问局部

def bar():
    x = 99          # 这个 x 是局部变量,和全局的 x 没关系
    print(x)

bar()               # 99
print(x)            # 10 ← 全局 x 没被改
````

## 综合练习:工具函数库

### ✏️ 练习 7(小项目) ⭐⭐⭐⭐

**题目**: 封装 5 个工具函数,并在主程序里调用验证:

1. `is_valid_phone(phone)` → 返回 `bool`(11 位且全数字)
2. `mask_phone(phone)` → 返回 `"138****8000"`
3. `bmi_report(h, w)` → 返回评级字符串
4. `is_leap(year)` → 返回 `bool`
5. `grade(score)` → 返回 `"A/B/C/D/F"`

**要求**: 每个函数**只负责一件事**,不要在一个函数里又算又打印。

示例:
```
print(is_valid_phone("13800138000"))  # True
print(mask_phone("13800138000"))      # 138****8000
print(bmi_report(1.75, 70))           # 正常
print(is_leap(2024))                  # True
print(grade(85))                      # B
```

````python
# ======================
# 学员代码区
# ======================
pass
````

<details>
<summary>💡 点击查看参考答案</summary>

````python
# 参考答案
def is_valid_phone(phone):
    return len(phone) == 11 and phone.isdigit()

def mask_phone(phone):
    return phone[:3] + "****" + phone[7:]

def bmi_report(h, w):
    bmi = w / (h ** 2)
    if bmi < 18.5:
        return "偏瘦"
    elif bmi < 24:
        return "正常"
    elif bmi < 28:
        return "偏胖"
    else:
        return "肥胖"

def is_leap(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# 测试
print(is_valid_phone("13800138000"))  # True
print(mask_phone("13800138000"))      # 138****8000
print(bmi_report(1.75, 70))           # 正常
print(is_leap(2024))                  # True
print(grade(85))                      # B
````

</details>
