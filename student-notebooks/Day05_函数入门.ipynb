---

**第 1 讲:def 定义与调用**

定义函数用关键字 **def**,后跟函数名、括号(放参数)、冒号。函数体需要缩进。

定义 ≠ 执行。函数 **必须调用** 才会执行里面的代码,否则写了白写。

```python
# 定义一个最简单的函数 —— 无参数,无返回值
def greet():
    print("Hello!")
    print("欢迎学习 Python")

greet()                     # 调用:函数名 + 括号
greet()                     # 再调用一次
```

---

**第 2 讲:函数的四种形式**

函数可以根据 **有没有参数** 和 **有没有返回值** 组合成四种形式。最常见、最通用的就是第 4 种:**有参有返回**。

```python
# 形式 1:无参无返回 —— 只负责打印
def say_hi():
    print("Hi!")

# 形式 2:有参无返回 —— 接受参数,打印结果
def greet(name):
    print(f"你好,{name}")

# 形式 3:无参有返回 —— 不接受参数,返回一个值
def get_pi():
    return 3.14159

# 形式 4:有参有返回 —— 最通用
def add(a, b):
    return a + b

# 调用
say_hi()
greet("小明")
pi = get_pi()
c = add(3, 5)
print(c)
```

---

**第 3 讲:形参 vs 实参**

**形参** 是定义时写在括号里的占位符;**实参** 是调用时塞进去的实际值。

类比:形参是 "展位",实参是 "展品"。展位只有一个,展品可以换着摆。

```python
def greet(name):              # name 是形参(占位符)
    print(f"你好,{name}")

greet("小明")                 # "小明" 是实参(实际值)
greet("小红")                 # 同一个函数,传入不同实参
```

---

**第 4 讲:默认参数 —— 有些参数可选填**

形参可以有 **默认值**,调用时如果不传对应参数,就用默认值。

注意:**默认参数必须放在非默认参数后面**,否则会报 SyntaxError。

```python
def greet(name="学员"):        # name 默认值为 "学员"
    print(f"你好,{name}")

greet()                        # 不传参,用默认值 → 你好,学员
greet("小明")                  # 传参,覆盖默认值 → 你好,小明
```

---

**第 5 讲:关键字参数 —— 按名字传参**

调用函数时用 **`形参名=值`** 的方式传参,顺序可以打乱,提高可读性。

```python
def introduce(name, age, city):
    print(f"{name},{age} 岁,来自 {city}")

# 关键字参数:按名字传值,顺序无关
introduce(age=18, city="成都", name="小明")
```

---

**第 6 讲:return —— 带结果回家 + 立刻退出**

**return** 把函数内部的结果返回给调用者。遇到 return,函数**立刻退出**,后面代码不执行 —— 像放学铃。

一个函数可以有多个 return,但只执行**遇到的第一个**。

```python
def check_age(age):
    if age < 0:
        return "年龄不能为负"    # 立刻退出
    if age < 18:
        return "未成年"          # 前面的没命中才到这里
    return "成年"

print(check_age(20))            # 成年
print(check_age(-1))            # 年龄不能为负
```

---

**第 7 讲:多值返回 —— 元组解包**

Python 可以同时 "返回多个值",本质是返回一个 **元组**。用逗号分隔的变量接收,称为**解包**。

```python
def calculate(a, b):
    return a + b, a - b, a * b   # 返回一个元组(和, 差, 积)

sum_, diff, prod = calculate(10, 3)   # 元组解包给三个变量
print(sum_, diff, prod)
```

---

**第 8 讲:作用域 —— 变量在函数里/外**

**全局变量** 定义在函数外,整个文件都能访问。**局部变量** 定义在函数内,只在函数里有效。

口诀:**函数里能读外面的,不能直接改外面**(除非用 global,本节不提倡)。

```python
x = 10                          # 全局变量

def foo():
    y = 20                      # 局部变量
    print(x)                    # 能访问全局变量 x

foo()
```

```python
x = 10                          # 全局变量

def bar():
    x = 99                      # 这个 x 是局部变量,和全局 x 无关
    print(x)                    # 99

bar()
print(x)                        # 10 —— 全局 x 没被改
```

---

**今日小结**

今天你进入 **函数层**:
- **def** 定义函数,定义了必须调用才执行
- 四种形式:无参无返回 / 有参无返回 / 无参有返回 / **有参有返回**(最常用)
- **形参** 是占位符,**实参** 是实际值
- **默认参数** 让参数可选填,必须放最后
- **关键字参数** 按名字传值,顺序无关
- **return** 返回结果 + 立刻退出
- 多值返回本质是**元组解包**
- **作用域**:函数里能读外面的,不能直接改外面的

> 易错点:函数定义后忘了调用等于白写;return 后面的代码不执行;默认参数必须放最后;函数内改全局变量要加 global。
