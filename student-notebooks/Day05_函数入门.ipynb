**1. def 定义与四种形式**

类比:**函数 = 工具 + 说明书**。定义时写一次,调用时用无数次。口诀:**定义时打包,调用时拆包**。没有调用,函数里面的代码不会执行。四种形式:① 无参无返回 ② 有参无返回 ③ 
无参有返回 ④ 有参有返回(最高频)。


```python
# ① 无参无返回
def say_hi():
    print("Hi!")

# ② 有参无返回
def greet(name):
    print(f"你好,{name}")

# ③ 无参有返回
def get_pi():
    return 3.14159

# ④ 有参有返回(最高频)
def add(a, b):
    return a + b

# 调用
say_hi()                        # Hi!
greet("小明")                   # 你好,小明
pi = get_pi()
c = add(3, 5)
print(c)                        # 8
```

**2. 形参、实参与默认参数**

形参是函数**定义时**写在括号里的"坑位",实参是函数**调用时**塞进去的"值"。默认参数让某些参数"可选填",但**默认参数必须放最后**,否则 `SyntaxError`。

```python
# 形参 vs 实参
def greet(name):        # name 是"形参"(形式参数),相当于占位符
    print(f"你好,{name}")

greet("小明")           # "小明" 是"实参"(实际参数),调用时传进去
greet("小红")           # 不同实参,同一段逻辑

# 默认参数:有些参数"可选填"
def greet(name="学员"):     # name 默认值为 "学员"
    print(f"你好,{name}")

greet()                      # 你好,学员(用默认值)
greet("小明")                # 你好,小明(覆盖默认值)

# 关键字参数:按名字传参,顺序无关
def introduce(name, age, city):
    print(f"{name},{age} 岁,来自 {city}")

introduce(age=18, city="成都", name="小明")  # 顺序无关
```

**3. return —— 带一个结果回家 + 立刻退出**

`return` 像"放学铃" —— 听到就停,后面代码不执行。一个函数可以有多个 `return`,但**只执行遇到的第一个**。`return a, b` 实际返回元组,用逗号分隔的变量"解包"。

```python
# return:带一个结果回家 + 立刻退出
def check_age(age):
    if age < 0:
        return "年龄不能为负"    # 立即退出,后面语句不再执行
    if age < 18:
        return "未成年"
    return "成年"

print(check_age(20))            # 成年
print(check_age(-1))            # 年龄不能为负

# 返回多个值:元组解包
def calculate(a, b):
    return a + b, a - b, a * b  # 返回元组 (和, 差, 积)

sum_, diff, prod = calculate(10, 3)
print(sum_, diff, prod)         # 13 7 30
```

**4. 作用域 —— 变量在函数里/外**

口诀:**函数里能读外面的,不能直接改外面**(除非加 `global`,本节不提倡,了解即可)。函数内定义的变量叫**局部变量**,只在函数内有效;函数外定义的是**全局变量**,函数内外都能读。

```python
x = 10              # 全局变量:函数外都能用

def foo():
    y = 20          # 局部变量:只在函数内有效
    print(x)        # ✅ 能访问全局

foo()
# print(y)          # ❌ 报错:函数外访问局部

def bar():
    x = 99          # 这个 x 是局部变量,和全局的 x 没关系
    print(x)

bar()               # 99
print(x)            # 10 ← 全局 x 没被改
```

**5. 综合实战 —— 工具函数库**

每个函数**只负责一件事**,不要在一个函数里又算又打印。"**计算和输出分开**"是良好函数设计的第一课。

```python
# 工具函数库
def is_valid_phone(phone):
    """判断是否为合法手机号(11 位且全数字)"""
    return len(phone) == 11 and phone.isdigit()

def mask_phone(phone):
    """手机号脱敏:138****8000"""
    return phone[:3] + "****" + phone[7:]

def grade(score):
    """成绩转等级"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# 主程序调用
print(is_valid_phone("13800138000"))  # True
print(mask_phone("13800138000"))      # 138****8000
print(grade(85))                       # B
```

**✏️ 练一练 ⭐⭐⭐⭐**

定义一个函数 `bmi_report(h, w)`,计算 BMI 并返回评级字符串(`"偏瘦"` / `"正常"` / `"偏胖"` / `"肥胖"`)。在主程序里调用它测试身高 
1.75、体重 70。

输出:`正常`(BMI = 22.86)

```python
# ======================
# 学员代码区
# ======================
```

```python
# 参考答案
def bmi_report(h, w):
    bmi = w / (h ** 2)
    if bmi < 18.5:
        return "偏瘦"
    elif bmi < 24:
        return "正常"
    elif bmi < 28:
        return "偏胖"
    else:
        return "肥胖"

print(bmi_report(1.75, 70))
```

**今日小结**

本节进入"函数层"。对比:

| 概念 | 说明 |
|---|---|
| 形参 | 定义时的占位符 |
| 实参 | 调用时塞进去的值 |
| 默认参数 | 可选填,必须放最后 |
| `return` | 带结果回家 + 立刻退出 |
| 作用域 | 函数里能读外面的,不能直接改外面 |

> 🔴 **易错点**:
> 1. 函数定义后不调用,代码不会执行 —— 写了 `def` 不等于"跑一次"
> 2. `return` 语句后面的代码不会执行 —— 像放学铃,听到就走
> 3. 默认参数必须放在非默认参数后面 —— `SyntaxError`
> 4. 函数内部不能直接修改全局变量 —— 要用 `global` 显式声明(本节了解)
> 5. `return a, b, c` 实际返回元组 —— 多值返回本质是单值(元组)
