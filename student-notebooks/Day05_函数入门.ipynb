# Day 05 · 函数入门> 前置知识：变量、输入输出、算术运算、列表、for 循环> 关键问题：为什么需要函数？怎样给一段代码起个名字反复调用？写代码时你一定会遇到这种情况：同一段逻辑在好几个地方重复出现，比如每次打印菜单都要写一串 `print`，每次算总价都要写一遍加法。复制粘贴当然能用，但一旦逻辑要改，你就得改所有粘贴过的地方，漏改一处就是一颗定时炸弹。**函数**就是给一段代码起一个名字，以后只要喊名字就能执行，改一处即改全部。今天我们从最基本的定义开始。

**1. 函数是什么 — 给一段代码起个名字**函数是一段有名字的、可重复执行的代码块。用 `def` 关键字定义，写一次就能调用无数次。定义只是"登记"下来，代码并不会立刻运行，只有调用的时候才会真正执行——就像菜谱写好了你不会自动吃到饭，得亲自按菜谱做一遍才行。下面定义一个最简单的打招呼函数并调用它。

```python
# 定义一个最简函数：无参数，只负责打印一句话def greet():    # 函数体：缩进的四行代码    print("你好，欢迎来到 Python 课堂！")    print("今天学习函数入门，加油！")# 下面这一行才是调用，写了函数名加括号才算真正执行greet()# 可以再调用一次，说明函数能重复使用greet()
```

**2. 函数的四种形式**根据「有没有参数」和「有没有返回值」两个维度，函数可以分成四类：① 无参无返回：只负责执行动作，比如打印菜单；② 有参无返回：接收数据执行动作，比如按名字打招呼；③ 无参有返回：不调外部数据，但要返回一个结果，比如读当前时间；④ 有参有返回：最常见、也最实用——收数据、算结果、返回给你。日常开发中 ④ 用得最多，下面把四种都写一遍，体会它们的区别。

```python
# ===== ① 无参无返回：只执行动作 =====def show_menu():    print("=== 菜单 ===")    print("1. 点餐")    print("2. 结账")    print("3. 退出")show_menu()  # 直接调用即可# ===== ② 有参无返回：接收数据再执行 =====def greet_by_name(name):    # name 是形参，调用时由外部传入    print(f"你好，{name}！欢迎加入学习。")greet_by_name("小明")   # 实参为 "小明"greet_by_name("小红")   # 实参为 "小红"# ===== ③ 无参有返回：返回一个结果 =====def get_pi():    # 虽然没有参数，但把结果交回给调用处    return 3.14159pi = get_pi()   # 返回值赋给变量print(f"圆周率约等于 {pi}")# ===== ④ 有参有返回：最常用 =====def add(a, b):    # 接收两个参数，返回它们的和    return a + bresult = add(3, 5)   # 调后拿到返回值print(f"3 + 5 = {result}")
```

**3. 形参与实参**定义函数时括号里写的变量叫**形式参数**（形参），是占位用的；调用时传进去的具体值叫**实际参数**（实参）。就像快递柜的格子编号是形参，你放进去的包裹才是实参。形参和实参按位置一一对应。

```python
# 定义：left 和 right 是形参，仅起到"占位"作用def print_sum(left, right):    print(f"{left} + {right} = {left + right}")# 调用：10 和 20 是实参，按照位置对应赋给 left 和 rightprint_sum(10, 20)     # left=10, right=20# 换个数据再调一次，体现实参可变print_sum(100, 200)   # left=100, right=200# 形参是"复印件"：函数内改形参不影响外面的变量def try_change(x):    x = 999                # 这里改的只是副本    print(f"函数内部 x = {x}")num = 1try_change(num)print(f"函数外部 num = {num}")   # 仍然是 1
```

**4. 默认参数与关键字参数**给参数预设一个值，调用时就可以不传——这就是**默认参数**。但它必须写在非默认参数之后，否则会报错。而**关键字参数**让你在调用时显式指定"这个值给哪个形参"，顺序就不重要了。

```python
# 默认参数：name 不传时自动使用 "学员"def greet(name="学员"):    print(f"你好，{name}！")greet()              # 走默认值，输出"学员"greet("小明")        # 传了就走传入的值# 关键字参数：调用时显式指定形参名，顺序可打乱def introduce(name, age):    print(f"{name}今年{age}岁")# 按位置传introduce("小明", 18)# 按关键字传，顺序颠倒也没关系introduce(age=20, name="小红")# 混合使用：位置参数必须写在关键字参数前面introduce("小刚", age=22)
```

**5. return — 带一个结果回家**`return` 会把结果送回调用处，并且**立即退出函数**。`return` 之后的任何代码都不会被执行。一个函数可以写多个`return`，但只会执行遇到的第一个。如果需要返回多个值，用逗号隔开——Python 会把它们打包成元组，调用处可以解包接收。

```python
# return 会立刻退出，后面的代码不执行def check_age(age):    if age >= 18:        return "成年"      # 命中这里就直接返回，下面的不跑了        print("这行永远不会打印")    return "未成年"print(check_age(20))   # 成年print(check_age(15))   # 未成年# 多值返回：逗号隔开，实际返回一个元组def get_min_max(numbers):    return min(numbers), max(numbers)# 直接打印返回值，能看到它是一个元组print(get_min_max([3, 1, 4, 1, 5]))# 解包接收：把元组分给多个变量lo, hi = get_min_max([3, 1, 4, 1, 5])print(f"最小值 {lo}, 最大值 {hi}")
```

**6. 作用域 — 变量在哪里能用**变量不是在哪儿都能用的。在函数外定义的叫**全局变量**，整个文件都能用；在函数内定义的叫**局部变量**，只在函数里有效。函数可以**读取**外层的变量，但不能直接**修改**它；想改的话必须用 `global` 关键字声明——不过初学者尽量少用 `global`，传参 + return 是更安全的选择。

```python
# 全局变量：定义在函数外面，整个文件都能读到count = 0# 函数可以读取外层变量def show_count():    print(f"当前 count = {count}")   # 只读，没问题show_count()# 函数内创建局部变量，出了函数就失效def make_local():    inner = 100            # 这是局部变量    print(f"函数内 inner = {inner}")make_local()# print(inner)   # 打开会报错 NameError# 想让函数修改全局变量，需要 global 关键字def reset_count():    global count           # 声明要改的是外面的 count    count = 0reset_count()print(f"复位后 count = {count}")
```

**7. 易错点与今日小结**今天我们学完了函数最重要的几个概念，把核心串一遍：- 定义用 `def`，调用要加 `()`，只定义不调用代码不会执行。- 参数是输入，`return` 是输出，二者可以组合成四种形式。- 形参是占位符，实参是真正传进来的值。- 默认参数写在后面，关键字参数能打乱顺序。- `return` 会立即退出函数，下面不再执行。- 多值返回其实是返回一个元组，可以解包接收。- 函数可以读外层变量，改外层变量要 `global`。下面用两个小练习巩固今天的内容。先自己写一遍，写完再对答案。

**今日小结**函数就是给一段代码起名字。以后每次想复用这段逻辑，只需呼喊名字，不用重复抄写。更关键的是：如果业务逻辑需要调整，你只需改动函数内部，所有调用的地方自动跟着变化——这是写"可维护代码"的第一步。记牢容易踩坑的五颗雷：1. 定义后不调用，代码不会执行；2. `return` 后面的代码不会被执行；3. 默认参数必须放在非默认参数后面；4. 函数内不能直接改全局变量，要改得用 `global`；5. `return a, b, c` 其实返回的是元组，不是三个独立的值。课后练习建议：- 写一个 `fahr_to_c(c)`，把摄氏温度转成华氏温度；  公式：`F = C × 9 / 5 + 32`。- 写一个 `stats(nums)`，返回一个列表的最小值、最大值、平均值，  并用解包语法把三个结果分别打印出来。
