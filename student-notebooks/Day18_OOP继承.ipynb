#### 为什么需要继承（代码复用 / is-a 关系）

**痛点：为什么需要它**  
写几个类发现彼此有大量重复代码，比如 Student 和 Teacher 都有 name、age、introduce()，复制粘贴越多，改 BUG 越痛。

**类比：生活例子**  
动物 -> 猫 / 狗：猫'is-a'动物，继承了动物共有的呼吸、进食，再扩展自己的'叫'。

**解释：它是什么**  
继承让子类**自动拥有**父类的属性和方法，再在这个基础上添加或修改，实现**代码复用**。

```python
# 父类（基类）
class Animal:
    def __init__(self, name):
        self.name = name

    def breathe(self):
        print(f"{self.name} 在呼吸")

# 子类：自动继承 Animal 的所有内容
class Dog(Animal):
    # 子类自己新增方法
    def bark(self):
        print(f"{self.name} 汪汪叫")

dog = Dog("旺财")
dog.breathe()   # 继承自 Animal
dog.bark()      # 子类自己的方法
```

```python
# --- 执行过程 ---
# ① class Animal: 创建父类
# ② def __init__(self, name): 绑定构造函数
# ③ def breathe(self): 绑定父类方法
# ④ class Dog(Animal): Dog 继承 Animal 的所有成员
# ⑤ def bark(self): 在 Dog 上额外添加 bark
# ⑥ dog = Dog('旺财')
#    -> Dog 自己没有 __init__，沿着'子类 → 父类'找到 Animal.__init__
#    -> 把 '旺财' 绑到 dog.name
# ⑦ dog.breathe()
#    -> Dog 自己没有 breathe，找到 Animal.breathe
#    -> self = dog，所以 name 是 '旺财'
# ⑧ dog.bark()
#    -> Dog 自己就有 bark，直接调用
```

**逐行解剖**

- `class Dog(Animal):` 括号里写父类名，表示'is-a'关系。
- 子类没有 `__init__` 时，自动用父类的构造函数。
- 子类可以**扩展**自己的方法（如 bark）。
- 查找顺序：**子类 -> 父类 -> 父类的父类**。

#### 单继承语法 class Child(Parent) + super().__init__()

**痛点：为什么需要它**  
子类想扩展父类属性，又不想重复写父类的绑定逻辑；直接写 父类.__init__(self, ...) 看起来能用，但层级一深就出错。

**类比：生活例子**  
儿子继承了父亲的房子（父类属性），再加盖一层（扩展属性），但必须**先**把父亲的房子盖好。

**解释：它是什么**  
super() 返回父类对象，用来**安全地调用父类方法**；写子类构造函数时，第一行就要 super().__init__(父类参数)。

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        # 先让父类把 name 绑好
        super().__init__(name)
        # 再扩展子类自己的属性
        self.breed = breed

    def bark(self):
        print(f"{self.name}({self.breed}) 汪汪叫")

dog = Dog("旺财", "柯基")
print(dog.name)    # 旺财
print(dog.breed)   # 柯基
dog.bark()
```

```python
# --- 执行过程 ---
# ① class Animal: 父类，__init__ 绑定 name
# ② class Dog(Animal): 子类
# ③ Dog.__init__(self, name, breed):
#    super().__init__(name) -> 调用 Animal.__init__
#    -> self.name = name 绑好
#    -> self.breed = breed 再绑子类自己的
# ④ dog = Dog('旺财', '柯基')
#    -> Dog.__init__(dog, '旺财', '柯基')
#    -> super().__init__('旺财') -> Animal.__init__(dog, '旺财')
#    -> dog.name = '旺财'
#    -> dog.breed = '柯基'
# ⑤ dog.bark() -> 输出格式化字符串
```

**逐行解剖**

- `super().__init__(name)` 等价于 `Animal.__init__(self, name)`，但**更规范**。
- 必须先调 `super().__init__()`，再绑定子类自己的属性。
- 忘记写 `super().__init__()` 会导致**父类属性未初始化**。

#### 方法重写（override）

**痛点：为什么需要它**  
父类的方法不能满足子类需要，比如父类 Animal.speak() 输出 '发声'，但狗要 '汪汪'、猫要 '喵喵'。

**类比：生活例子**  
法律是父类，地方法规是子类：国家规定最低工资为 A，但地方可**重写**为更高的 B。

**解释：它是什么**  
子类重新定义父类的同名方法，调用时**优先执行子类版本**，这就是**方法重写**。

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} 发出声音")

class Dog(Animal):
    # 重写父类的 speak
    def speak(self):
        print(f"{self.name} 汪汪叫")

class Cat(Animal):
    # 重写父类的 speak
    def speak(self):
        print(f"{self.name} 喵喵叫")

a = Animal("某动物")
d = Dog("旺财")
c = Cat("咪咪")

a.speak()   # 某动物 发出声音
d.speak()   # 旺财 汪汪叫
c.speak()   # 咪咪 喵喵叫
```

```python
# --- 执行过程 ---
# ① class Animal: 定义父类 speak
# ② class Dog(Animal):
#    def speak(self): 注册到 Dog，覆盖了 Animal.speak
# ③ class Cat(Animal):
#    def speak(self): 注册到 Cat，覆盖了 Animal.speak
# ④ a = Animal('某动物') -> a 是 Animal 实例
# ⑤ d = Dog('旺财') -> d 是 Dog 实例
# ⑥ c = Cat('咪咪') -> c 是 Cat 实例
# ⑦ a.speak()
#    -> Animal 实例，找到 Animal.speak -> 输出 '发出声音'
# ⑧ d.speak()
#    -> Dog 实例，Dog 自己有 speak -> 输出 '汪汪叫'
# ⑨ c.speak()
#    -> Cat 实例，Cat 自己有 speak -> 输出 '喵喵叫'
```

**逐行解剖**

- 重写只需在子类里**同名定义**即可，完全替换。
- 如果想保留父类版本并扩展，可以在子类里用 `super().speak()` 先调父类。
- 重写时方法签名（参数）**不必和父类完全一致**，但建议保持。

#### 多态（同一接口，不同实现）

**痛点：为什么需要它**  
一个函数要处理多种动物，如果每个动物单独写 if-elif，代码越加越长，新增类型还得改函数。

**类比：生活例子**  
USB 接口 = 同一个接口，U 盘、键盘、鼠标 = 不同实现。插上去就能用，主机不需要知道具体是哪个。

**解释：它是什么**  
不同类的对象，对**同一方法名**给出**不同行为**；调用方只关心 '有没有这个方法'，不关心具体类型。

```python
class Dog:
    def speak(self):
        return "汪汪"

class Cat:
    def speak(self):
        return "喵喵"

class Cow:
    def speak(self):
        return "哞哞"

def animal_sound(obj):
    # 不管传入哪个对象，只要它有 speak 方法
    print(f"{type(obj).__name__} 叫: {obj.speak()}")

# 同一函数，不同实现
animal_sound(Dog())   # Dog 叫: 汪汪
animal_sound(Cat())   # Cat 叫: 喵喵
animal_sound(Cow())   # Cow 叫: 哞哞
```

```python
# --- 执行过程 ---
# ① 定义 Dog、Cat、Cow 三个类，各自实现 speak
# ② 定义 animal_sound(obj)，只关心 obj 有没有 speak
# ③ animal_sound(Dog())
#    -> 实例化 Dog
#    -> obj.speak() 调用 Dog.speak -> 返回 '汪汪'
#    -> type(obj).__name__ 得到 'Dog'
# ④ animal_sound(Cat())
#    -> 调用 Cat.speak -> 返回 '喵喵'
# ⑤ animal_sound(Cow())
#    -> 调用 Cow.speak -> 返回 '哞哞'
```

**逐行解剖**

- 多态的前提：不同类实现了**同名方法**。
- Python 是动态类型，不要求继承同一父类（鸭子类型）。
- 调用方只写一次逻辑，新增类型只要符合接口，**无需修改原函数**。

#### isinstance() / issubclass()

**痛点：为什么需要它**  
多态场景下想知道一个对象到底属于哪个类，或者一个类是不是另一个类的子类，靠猜容易出错。

**类比：生活例子**  
身份证检查：你是不是中国人（issubclass）？你是不是某个具体人（isinstance）？

**解释：它是什么**  
- isinstance(obj, cls) 判断对象是否是某类（或其子类）的实例。
- issubclass(A, B) 判断 A 是否是 B 的子类。

```python
class Animal:
    pass

class Dog(Animal):
    pass

class Cat(Animal):
    pass

dog = Dog()

# isinstance: 判断对象
print(isinstance(dog, Dog))     # True
print(isinstance(dog, Animal))  # True  (继承链上也算)
print(isinstance(dog, Cat))     # False

# issubclass: 判断类
print(issubclass(Dog, Animal))  # True
print(issubclass(Cat, Animal))  # True
print(issubclass(Dog, Cat))     # False
```

```python
# --- 执行过程 ---
# ① class Animal: 父类
# ② class Dog(Animal): Dog 继承 Animal
# ③ class Cat(Animal): Cat 继承 Animal
# ④ dog = Dog() -> 创建 Dog 实例
# ⑤ isinstance(dog, Dog) -> True（精确匹配）
# ⑥ isinstance(dog, Animal)
#    -> Dog 是 Animal 子类，沿着继承链找，找到 -> True
# ⑦ isinstance(dog, Cat) -> False（Dog 与 Cat 无关）
# ⑧ issubclass(Dog, Animal) -> True（Dog 继承 Animal）
# ⑨ issubclass(Cat, Animal) -> True（Cat 继承 Animal）
# ⑩ issubclass(Dog, Cat) -> False（两者互为独立子类）
```

**逐行解剖**

- isinstance(obj, cls) 会把**整条继承链**都算进去。
- issubclass(A, B) 中 A 也算自己的子类：issubclass(Dog, Dog) 为 True。
- 两者都支持**元组**作第二参数，例如 isinstance(obj, (Dog, Cat))。

```python
# ============ 参考答案 ============
class Vehicle:
    pass

class Car(Vehicle):
    pass

class Bike(Vehicle):
    pass

car = Car()
print(isinstance(car, Car))      # True
print(isinstance(car, Vehicle))  # True  (继承链)
print(isinstance(car, Bike))     # False

print(issubclass(Bike, Vehicle)) # True
print(issubclass(Bike, Car))     # False
```

#### 综合练习：把继承五个知识点串起来

设计一个员工薪资系统：
1. 父类 Employee：属性 name、base_salary，方法 pay() 返回 base_salary
2. 子类 Sales（销售）：新增属性 commission（提成），**重写** pay() 返回 base_salary + commission
3. 子类 Manager（经理）：新增属性 bonus（奖金），**重写** pay() 返回 base_salary + bonus
4. 用 super().__init__() 调用父类构造函数
5. 写函数 show_pay(emp) 传入任意员工，打印 pay() 返回的薪资
6. 用 isinstance 判断传入的是不是 Employee

> 问自己:
> - 重写 pay 时如何保留父类的计算逻辑?
> - show_pay 函数体现了什么特性（多态）?

```python
# ============ 学员代码区 ============
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def pay(self):
        return self.base_salary

# class Sales(Employee): ...
# class Manager(Employee): ...
# def show_pay(emp): ...
# emps = [Employee(...), Sales(...)]
pass
```

```python
# ============ 参考答案 ============
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def pay(self):
        return self.base_salary

class Sales(Employee):
    def __init__(self, name, base_salary, commission):
        super().__init__(name, base_salary)
        self.commission = commission

    def pay(self):
        # 保留父类逻辑 + 提成
        return super().pay() + self.commission

class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def pay(self):
        return super().pay() + self.bonus

def show_pay(emp):
    if isinstance(emp, Employee):
        print(f"{emp.name} 薪资: {emp.pay()}")
    else:
        print('不是员工类型')

emps = [
    Employee('张三', 5000),
    Sales('李四', 5000, 2000),
    Manager('王五', 8000, 5000),
]

for e in emps:
    show_pay(e)
```
