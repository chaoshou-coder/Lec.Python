**1. 索引与切片 —— 给每个字符编上门牌号**

类比:字符串 = 一排格子,每个字符住一格,从 0 开始编号(index)。**索引**取单个字符,**切片**取一段。口诀:**从零开始,负数倒着数**。

```python
name = "孙悟空"
print(name[0])   # 孙  ← 第 1 个,索引 0
print(name[1])   # 悟
print(name[2])   # 空
print(name[-1])  # 空  ← 倒数第 1,负数倒着数
print(name[-2])  # 悟

# 切片 s[start:end]:从 start 切到 end 之前(左闭右开)
s = "hello world"
print(s[0:5])    # hello
print(s[6:11])   # world
print(s[:5])     # hello  ← 省略 start 默认从 0 开始
print(s[6:])     # world  ← 省略 end 默认切到末尾
print(s[:])      # hello world  ← 全切,相当于复制

# 步长 s[start:end:step]
s = "abcdefghijk"
print(s[::2])    # acegik  ← 每隔一个取一个
print(s[::-1])   # kjihgfedcba  ← 逆序(常用!)
```

**2. 字符串判断方法 —— 是不是数字?是不是字母?**

这三个方法都返回 `True` / `False`,常和 `if` 搭配做输入校验。口诀:**digit = 数字键,isalpha = 字母表,alnum = 字母+数字**。

```python
print("12345".isdigit())    # True  ← 全是数字
print("12.3".isdigit())     # False ← 小数点不是数字
print("abc".isalpha())      # True  ← 全是字母
print("abc123".isalpha())   # False
print("abc123".isalnum())   # True  ← 字母或数字(只要不是符号就行)
print("abc-123".isalnum())  # False ← 横杠是符号

# 实战:校验手机号合法性(11 位且全数字)
phone = input("请输入手机号:")
if len(phone) == 11 and phone.isdigit():
    print("手机号格式正确")
else:
    print("手机号格式错误")
```

**3. find 与 rfind —— 正着找 vs 倒着找**

`find(sub)` 返回子串 `sub` **第一次**出现的索引;`rfind(sub)` 返回**最后一次**出现的索引。找不到都返回 `-1`。为什么需要两个?例:文件路径 
`/tmp/data/a.txt` 要取最后一个 `/` 之后的内容,只能倒着找。


```python
s = "hello hello hello"
print(s.find("hello"))   # 0   ← 第一次
print(s.rfind("hello"))  # 12  ← 最后一次
print(s.find("python"))  # -1  ← 找不到

# 实战:从文件路径提取文件名
path = "/tmp/data/a.txt"
last_slash = path.rfind("/")
filename = path[last_slash + 1:]
print(filename)  # a.txt
```

**✏️ 练一练 ⭐⭐**

输入一个文件路径(如 `/home/user/doc/report.pdf`),用 `rfind` 定位最后一个 `/` 和最后一个 `.`,切片提取文件名(不含路径和后缀)。

示例输入:`/home/user/doc/report.pdf` → 输出:`report`

**4. replace 与 f-string —— 替换 & 格式化输出**

`replace(old, new)` 把字符串中**所有** `old` 子串替换成 `new`,返回新字符串(原串不变 —— 字符串是不可变的)。f-string = 在字符串前面加 
`f`,大括号里直接写变量,最常用最推荐。


```python
# replace:批量替换(原串不变!)
text = "xxx 是个好网站,xxx 很好用"
clean = text.replace("xxx", "***")
print(clean)  # *** 是个好网站,*** 很好用
print(text)   # xxx 是个好网站,xxx 很好用(原串没变)

# f-string:把变量塞进字符串
name = "小明"
age = 18
print(f"我是 {name},今年 {age} 岁")

# f-string 保留两位小数
pi = 3.1415926
print(f"圆周率: {pi:.2f}")  # 3.14
```

**今日小结**

本节把字符串变成可查询、可切分、可校验的"小型数据集"。对比:

| 操作 | 方法 | 返回值 |
|---|---|---|
| 取单个字符 | `s[i]` | 字符 |
| 取一段 | `s[start:end]` | 子串(左闭右开) |
| 正着找 | `s.find(sub)` | 索引或 -1 |
| 倒着找 | `s.rfind(sub)` | 索引或 -1 |
| 判断数字 | `s.isdigit()` | True/False |
| 替换 | `s.replace(old, new)` | 新字符串 |
| 格式化 | `f"...{变量}..."` | 字符串 |

> 🔴 **易错点**:
> 1. 索引差一:第 n 个字符的索引是 `n-1`;切片 `s[a:b]` 取到 `b` 之前,不含 `b`
> 2. `find()` 找不到返回 `-1`,不是 `None` 也不是 `0`
> 3. 字符串不可变:`replace()` 返回新字符串,原串不变,必须用变量接住返回值
> 4. `isdigit()` 只认纯数字:负数、小数、全角数字都返回 `False`
