**1. 字符串索引 — 给字符编门牌号**字符串里的每个字符都像一间间连续的房间,Python 从 0 开始编号。正着数用 0、1、2...,倒着数用 -1、-2...,非常直观。

```python
# 定义一个示例字符串s = "Hello, Python!"# 正索引: 第一个字符从 0 开始print("第一个字符:", s[0])   # H# 取第 8 个字符(索引为 7)print("第 8 个字符:", s[7])  # P# 负索引: -1 是最后一个字符print("最后一个字符:", s[-1])  # !# 负索引: -2 是倒数第二个字符print("倒数第二个字符:", s[-2])  # n# 字符串长度(字符个数)print("字符串长度:", len(s))  # 14
```

可以这样记: **第 n 个字符的索引是 n-1**, 因为 Python 从 0 开始数。索引超出范围会抛出 IndexError,初学者最常在这里翻车。

**2. 切片 — 一次切出一段**切片语法是 s[start:end], 遵循"左闭右开"原则:包含 start, 不包含 end。省略 start 默认从头开始,省略 end 默认到末尾结束,第三个参数是**步长**。

```python
# 示例字符串s = "Python编程真有趣"# 取索引 0~5 的字符(不含 5),即前 5 个字符print("s[0:5]:", s[0:5])  # Pytho# 省略 start, 从开头切到索引 5(不含)print("s[:5]:", s[:5])    # Pytho# 从索引 6 切到末尾print("s[6:]:", s[6:])    # 有趣# 省略 start 和 end, 复制整串print("s[:]:", s[:])      # Python编程真有趣# 步长为 2: 每隔一个字符取一个print("s[::2]:", s[::2])  # Pto编真趣# 步长为 -1: 经典逆序(字符串反转)print("s[::-1]:", s[::-1])  # 趣有真程编nohtyP
```

**3. 字符串查找方法: find / rfind**find(sub) 从左往右找,返回子串第一次出现的索引;rfind(sub) 从右往左找,返回最后一次出现的索引。找不到时**返回 -1**, 而不是 None 或 0, 这点要牢记。

```python
# 示例字符串s = "香蕉味的香蕉牛奶"# 查找 "香蕉" 第一次出现的位置pos1 = s.find("香蕉")print("find(香蕉):", pos1)  # 0# 查找 "香蕉" 最后一次出现的位置pos2 = s.rfind("香蕉")print("rfind(香蕉):", pos2)  # 4# 查找不存在的子串, 返回 -1pos3 = s.find("西瓜")print("find(西瓜):", pos3)  # -1# 从指定位置开始查找pos4 = s.find("香蕉", 1)  # 从索引 1 开始往后找print("find(香蕉, 1):", pos4)  # 4
```

**4. 字符串判断方法**Python 提供了三个常用的"是否纯"判断方法:isdigit() 判断是否全是数字, isalpha() 判断是否全是字母,isalnum() 判断是否只包含字母和数字。注意: 负号、小数点都不算数字。

```python
# isdigit: 判断是否纯数字组成print("12345".isdigit())   # Trueprint("-12345".isdigit())  # False, 含负号print("1.5".isdigit())     # False, 含小数点print("".isdigit())        # False, 空字符串print("---")# isalpha: 判断是否纯字母print("Python".isalpha())   # Trueprint("Python3".isalpha())  # False, 含数字print("你好".isalpha())     # True, 中文也算字母print("---")# isalnum: 判断是否只含字母或数字print("Python3".isalnum())  # Trueprint("P3 版本".isalnum())  # False, 含空格
```

**5. replace — 批量替换**replace(old, new) 会把字符串中**所有** old 子串替换成 new,但字符串是**不可变类型**, 所以 replace 返回的是**一个新的字符串**,原字符串保持不变, 必须用变量接收返回值。

```python
# 示例字符串s = "猫吃鱼, 狗吃肉, 熊猫吃竹子"# 替换所有 "吃" 为 "爱"s2 = s.replace("吃", "爱")print("替换后:", s2)  # 猫爱鱼, 狗爱肉, 熊猫爱竹子# 原字符串未被修改(不可变性)print("原字符串:", s)  # 猫吃鱼, 狗吃肉, 熊猫吃竹子# 限制替换次数: 最多替换 1 次s3 = s.replace("吃", "爱", 1)print("最多换 1 次:", s3)  # 猫爱鱼, 狗吃肉, 熊猫吃竹子
```

**6. f-string — 把变量塞进字符串**f-string 是 Python 3.6+ 主推的格式化方式,在字符串前加 f, 用大括号 {} 包裹表达式即可嵌入。大括号里可以是变量、运算、函数调用, 甚至是格式控制(如 :.2f)。

```python
# 示例变量name = "小明"age = 20height = 1.753# f-string: 最推荐的写法print(f"姓名: {name}, 年龄: {age}")# f-string 里直接做运算print(f"明年 {name} {age + 1} 岁")# 保留两位小数: :.2fprint(f"身高: {height:.2f} 米")  # 1.75 米print("---")# % 格式化(老式写法, 了解即可)print("姓名: %s, 年龄: %d" % (name, age))print("---")# .format 写法(Python 2/3 通用)print("姓名: {}, 年龄: {}".format(name, age))# .format 带位置索引print("年龄: {1}, 姓名: {0}".format(name, age))
```

```python
# =====================# 学员代码区# =====================phone = "13812345678"  # 示例输入# TODO: 判断 phone 是否全为数字且长度为 11# TODO: 用切片取前 3 位和后 4 位, 中间用 **** 填充# TODO: 用 f-string 输出掩码后的结果pass
```

```python
# =====================# 参考答案# =====================phone = "13812345678"# 1. 校验: 必须全是数字且长度恰好 11if phone.isdigit() and len(phone) == 11:    # 2. 用切片取前 3 位和后 4 位    head = phone[:3]    tail = phone[-4:]    # 3. 用 f-string 拼接掩码结果    masked = f"{head}****{tail}"    print("掩码结果:", masked)else:    print("请输入合法的 11 位手机号")print("---")# 测试不合法输入phone2 = "138123abc78"if phone2.isdigit() and len(phone2) == 11:    print(f"{phone2[:3]}****{phone2[-4:]}")else:    print("请输入合法的 11 位手机号")
```
