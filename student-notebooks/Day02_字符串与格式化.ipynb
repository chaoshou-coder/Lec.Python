## 第一讲(15 分钟)—— 索引、切片与长度

### 概念解释

类比: 字符串像一排**带门牌号的房子**:

```
字符串:  孙  悟  空
正索引:   0   1   2
负索引:  -3  -2  -1
```

````python
# 示例代码:索引
name = "孙悟空"
print(name[0])   # 孙  ← 第 1 个,索引 0
print(name[1])   # 悟
print(name[2])   # 空
print(name[-1])  # 空  ← 倒数第 1,负数倒着数
print(name[-2])  # 悟
````

**口诀**: 从零开始,负数倒着数。

> 🔴 **易错点**: 学员常把"第几个"和"索引"搞混。第 1 家住门牌号 0,第 2 家住门牌号 1。

### 切片:一次切出一段

`s[start:end]` 从 `start` 切到 `end` **之前**(左闭右开),就像切蛋糕 —— 刀口落在 `end`,但 `end` 那块不拿。

````python
# 示例代码:切片
s = "hello world"
print(s[0:5])    # hello
print(s[6:11])   # world
print(s[:5])     # hello  ← 省略 start 默认从 0 开始
print(s[6:])     # world  ← 省略 end 默认切到末尾
print(s[:])      # hello world  ← 全切,相当于复制
````

### 步长 `s[start:end:step]`

````python
# 示例代码:步长
s = "abcdefghijk"
print(s[::2])    # acegik  ← 每隔一个取一个
print(s[::-1])   # kjihgfedcba  ← 逆序

# 趣味应用:判断回文
word = "level"
if word == word[::-1]:
    print(f"{word} 是回文!")
````

### `len()`:求长度

````python
# 示例:len() 求字符串长度
s = "hello world"
print(len(s))       # 11
print(len("孙悟空"))  # 3
````

## 第二讲(15 分钟)—— 查找与判断

### `find()` 与 `rfind()`:正着找 vs 倒着找

- `find(sub)` 返回子串**第一次**出现的索引
- `rfind(sub)` 返回子串**最后一次**出现的索引
- 找不到都返回 `-1`

````python
# 示例代码:find 和 rfind
s = "hello hello hello"
print(s.find("hello"))   # 0   ← 第一次
print(s.rfind("hello"))  # 12  ← 最后一次
print(s.find("python"))  # -1  ← 找不到

# 实用场景:文件路径取文件名
path = "/tmp/data/a.txt"
last_slash = path.rfind("/")
filename = path[last_slash + 1:]
print(filename)  # a.txt
````

### 字符串判断方法:是不是数字?是不是字母?

| 方法 | 含义 | 示例 |
|---|---|---|
| `.isdigit()` | 全是数字 | `"123".isdigit()` → True |
| `.isalpha()` | 全是字母 | `"abc".isalpha()` → True |
| `.isalnum()` | 字母或数字 | `"abc123".isalnum()` → True |

**口诀**: digit = 数字键, isalpha = 字母表, alnum = 字母+数字

````python
# 示例代码:字符串判断
print("12345".isdigit())    # True  ← 全是数字
print("12.3".isdigit())     # False ← 小数点不是数字
print("abc".isalpha())      # True  ← 全是字母
print("abc123".isalpha())   # False
print("abc123".isalnum())   # True  ← 字母或数字
print("abc-123".isalnum())  # False ← 横杠是符号
````

## 第三讲(15 分钟)—— replace 与 f-string

### `replace()`:批量替换

`str.replace(old, new)` 把**所有** `old` 替换成 `new`,返回**新字符串**(原串不变 —— 字符串是不可变的)。

````python
# 示例代码:replace
text = "xxx 是个好网站,xxx 很好用"
clean = text.replace("xxx", "***")
print(clean)  # *** 是个好网站,*** 很好用

# ⚠️ 字符串不可变! replace 不修改原串
s = "abc"
s.replace("a", "x")
print(s)  # abc ← 还是 abc,必须用变量接住!

# 正确写法
s = s.replace("a", "x")
print(s)  # xbc
````

> 🔴 **教学红线**: `replace()` 返回新字符串,原串不变,必须用变量接住返回值!

### f-string:最推荐的格式化方式

把变量"塞"进字符串,推荐用 **f-string**(Python 3.6+):

````python
# 示例代码:f-string(重点!)
name = "小明"
age = 18

# 方式 1:% 格式化(古老,了解即可)
print("我是 %s,今年 %d 岁" % (name, age))

# 方式 2:str.format()(了解)
print("我是 {},今年 {} 岁".format(name, age))

# 方式 3:f-string(最推荐!)
print(f"我是 {name},今年 {age} 岁")

# f-string 还可以放表达式
print(f"{name}十年后 {age + 10} 岁")

# 保留两位小数
pi = 3.1415926
print(f"圆周率: {pi:.2f}")  # 3.14
````
