#### 1. 字符串创建 —— 三种引号

> **痛点**: 文字里也有引号怎么办?
> **类比**: 单/双引号就像左右手,完全等价,选一种即可。
> **解释**: 用 **单引号** **双引号** **三引号** 三种方式创建字符串。
> 当文字本身含引号,用另一种引号框住即可。

**ASCII 内存图**:

```
s = "悟空" → s ──→ 连续内存块:字符 '悟'(0) '空'(1)
```

**关键**:字符按顺序排,每人编号 0、1、2……

````python
# 示例:三种引号
s1 = '单引号'; s2 = "双引号"; s3 = '''三引号'''

# --- 执行过程 ---
# s1, s2, s3 分别赋值引号内的内容
# 单引号(')与双引号(")完全等价

print(s1); print(s2); print(s3)

# 引号嵌套
line1 = '悟空说:"俺老孙来也"'
line2 = "她说:'你好'"

# --- 执行过程 ---
# 外层用单/双引号,内部用另一种作为普通字符

print(line1); print(line2)
````

**逐行解剖**:
- `'..'` 与 `".."` 完全等价,选一种统一用即可
- `'''...'''` `"""..."""` 为三引号,支持跨行
- 引号是边界标记,**不会**出现在字符串里

**常见错误**:
1. `'悟空说:'你好''` → SyntaxError(同类引号不能嵌套同类)
2. 文字中 `\` 若要输出,写成 `\\`

#### 2. 索引 —— 给每个字符编门牌号

> **痛点**: "孙悟空" 要取 "悟"(第二个字),怎么做?
> **类比**: 队伍同学编号 — 正数 0 起,负数 -1 倒数。
> **解释**: `s[0]` = 第一个字符; `s[-1]` = 最后一个字符。

**ASCII 内存图**:

```
s = "孙悟空"
正数索引:  0('孙')  1('悟')  2('空')
负数索引:    -3       -2       -1
```

````python
# 示例:正数索引 + 负数索引
name = "孙悟空"

# --- 执行过程 ---
# name[0] → 索引 0,即 '孙'
# name[1] → 索引 1,即 '悟'
# name[2] → 索引 2,即 '空'
# name[-1] → 倒数第一,即 '空'
# name[-2] → 倒数第二,即 '悟'
# name[-3] → 倒数第三,即 '孙'

print(name[0])
print(name[1])
print(name[2])
print(name[-1])
print(name[-2])
print(name[-3])
````

**逐行解剖**:
- `name[0]` 取出索引 0,即第一个字符
- 索引只能是整数,小数报 TypeError
**常见错误**:
1. `name[3]` → IndexError(越界)
2. 第 N 个字符索引是 N - 1(新手常错)

#### 3. 切片 —— 一次截取一串

> **痛点**: 从邮箱 `sunwu@163.com` 取用户名/域名,怎么批量?
> **类比**: 切香肠 — 刀起 start,刀落 end,**包含 start, 不包含 end**。
> **解释**: `s[start:end]`,省略 start 默认 0,省略 end 默认末。

**ASCII 内存图**:

```
s = "0123456789"
s[2:6] 范围:2 3 4 5 (6 不包含)
结果: "2345"  (长度 6-2 = 4)

口诀:刀口落在 end,那块不拿。
```

````python
# 示例:完整切片 + 省略 start/end
s = "abcdefghij"

# --- 执行过程 ---
# s[2:6] → 索引 2、3、4、5 → "cdef"
# s[:3]  → 起默认 0,到 3 之前 → "abc"
# s[7:]  → 从 7 到末 → "hij"
# s[:]   → 复制整串 → "abcdefghij"

print(s[2:6])
print(s[:3])
print(s[7:])
print(s[:])

# 实际应用:从邮箱取用户名 + 域名
email = "sunwu@163.com"
print(email[:5])
print(email[6:])
````

**逐行解剖**:
- `s[2:6]` 2 包含,6 不包含,切片长度 = end - start
- `s[:3]` 等价 `s[0:3]`

**常见错误**:
1. 把 end 理解包含 → `s[1:4]` 取 1、2、3,**不含 4**
2. 起始 > 结束 → 返回空串(不报错)
3. end 超过串长 → 自动截断

#### 4. 步长 s[::step] —— 跳着取、倒着取

> **痛点**: 如何每隔一个取一个,或把字符串倒过来?
> **类比**: 每步跨 2 格(step=2);倒退着走(step=-1)。
> **解释**: `s[::2]` 跳取;`s[::-1]` 反转,是反转的经典写法。

**ASCII 内存图**:

```
s = "abcdefghijk" (索引 0~10)
s[::2]  → acegik (正步长跳取)
s[::-1] → kjihgfedcba (反转)
```

````python
# 示例:正步长 + 负步长
s = "abcdefghijk"

# --- 执行过程 ---
# s[::2]  → 0、2、4、6、8、10 → "acegik"
# s[::-1] → 从右往左逐个取 → "kjihgfedcba"
# s[8:1:-2] → 从 8 往左到 1 前,跨 2 格
#           → 8、6、4、2 → "igec"

print(s[::2])
print(s[::-1])
print(s[8:1:-2])
````

**逐行解剖**:
- `s[::2]` start、end 省略表整串
- `s[::-1]` 反转串,Python 3.6+ 标准
**常见错误**:
1. step=0 → ValueError
2. 倒序时 start<end → 空串
3. 负步长省略 start 默认最后一个

> **问自己**:

#### 7. find / rfind —— 查找子串位置

> **痛点**: 知道子串存在,还要知道它在 **哪里**。
> **类比**: find 正着翻书找第一次;rfind 倒着翻找最后一次。
> **解释**: `find(sub)` 返回第一次索引,`rfind(sub)` 最后一次,
> 找不到返回 -1。

**ASCII 内存图**:

```
s = "hello hello hello"
s.find("hello")  → 0
s.rfind("hello") → 12
s.find("world")  → -1
```

````python
# 示例:find / rfind 基本 + 应用
s = "hello hello hello"

# --- 执行过程 ---
# s.find("hello") → 0
# s.rfind("hello") → 12
# s.find("world") → -1

pos1 = s.find("hello")
pos2 = s.rfind("hello")
pos3 = s.find("world")
print(pos1); print(pos2); print(pos3)

# 应用:提取文件名
path = "/tmp/data/python/demo.txt"
last = path.rfind("/")
filename = path[last + 1:]
print(filename)
````

**逐行解剖**:
- `find(sub)` 返回整数,代表 sub 第一个字符位置
- 找不到返回 -1,不报错
- `"X" in s` 等价于 `s.find("X") != -1`
**常见错误**:
1. `if s.find("X"):` 在找不到时 -1 是真值,判断会反过来
2. 应写 `if s.find("X") == -1` 或 `if "X" not in s`

#### 8. replace / strip / join / split —— 字符串四件套

> **痛点**: 敏感词替换、清理头尾空格、拼合/切开字符串。
> **类比**: replace=换词,strip=削皮,join=粘胶,split=切开。
> **解释**: 最常用的字符串处理工具,配合使用能解决 80% 文字场景。

**ASCII 内存图**:

```
text = "xxx 是个好网站"
fresh = text.replace("xxx", "***")

text ──→ │xxx 是好网站│(旧串,未变)
fresh ──→│*** 是个好网站│(新串,replace 返回)

口诀:字符串不可变,replace 必须接返回值。
```

````python
# 示例:replace + strip + split + join
text = "xxx 是个好网站,xxx 很好用"

# --- 执行过程 ---
# replace 找到所有 "xxx",换为 "***"
# 原文本不变

fresh = text.replace("xxx", "***")
print(text); print(fresh)

s = "   hello   "
print(s.strip())

words = "a b c d"
print(words.split(" "))

date = "2026-03-15"
print(date.split("-"))

joined = "、".join(["悟空", "八戒"])
print(joined)
````

**逐行解剖**:
- `replace(old, new)` 把所有 old 换成 new,原串不变
- `strip()` 去头尾空白,中间不动
- `split(sep)` 按 sep 切,返回列表
- `sep.join(list)` 用 sep 拼合
**常见错误**:
1. 不接 replace 返回值 → 实际什么都没改
2. strip 只能去头尾,去中间空格用 replace

#### 9. 字符串不可变 —— 不能通过索引修改

> **痛点**: 想把 "hello" 第 0 个字符改成 "H",为什么报错?
> **类比**: 字符串像 **刻在石头上的字**,想改只能重刻一块。
> **解释**: 字符串是 **不可变**(immutable)类型,创建后不能通过索引修改。

**ASCII 内存图**:

```
s = "hello" → s ──→ │ h  e  l  l  o │ (不可修改)
s[0] = "H" → TypeError
改法:s = "H" + s[1:] → 新串 │ H  e  l  l  o │
```

````python
# 示例:修改字符串的正确做法
s = "hello"

# 取消注释会报错 TypeError:
# s[0] = "H"

# --- 执行过程 ---
# s[0] = "H" 触发 TypeError,字符串不允许索引赋值
# 正确做法:构造新串,赋给 s

s = "H" + s[1:]
print(s)

# replace 本质也是返回新串
t = "hello"
u = t.replace("h", "H")
print(u); print(t)
````

**逐行解剖**:
- `s[0] = "X"` → TypeError
- 通过切片+拼接构造新串
**常见错误**:
1. 把字符串当成可变 → 尝试 `s[0] = "x"` 报错
2. 错误以为 replace 修改了原串

#### 10. f-string 格式化输出 —— 把变量嵌入字符串

> **痛点**: 拼接 "我的名字是 X,今年 Y 岁" 这种句式,
> 手写 `+` 容易漏空格、忘转类型。
> **类比**: 像 **占位符** — "姓名:____ 年龄:____",运行时自动填充。
> **解释**: 在字符串前加 **f**,花括号 **{}** 内放变量或表达式。

**ASCII 内存图**:

```
name = "悟空"; age = 18
print(f"我是{name},今年{age}岁")
                 ↓       ↓
              替换为"悟空"  替换为18
结果:"我是悟空,今年18岁"
```

````python
# 示例:f-string 基本 + 表达式
name = "悟空"
age = 18

# --- 执行过程 ---
# {name} → "悟空",{age} → 18

print(f"我是{name},今年{age}岁")

# 花括号内可写表达式
a = 10; b = 3
print(f"{a} + {b} = {a + b}")
print(f"{a} × {b} = {a * b}")
````

**逐行解剖**:
- 前缀 **f** 表示这是 f-string
- `{变量名}` 会被替换为变量值
- 花括号内可写 **表达式**,如 `{age + 1}`
**常见错误**:
1. 漏前缀 f → 花括号原样输出
2. 花括号不配对 → `{{name}}` 输出字面 {name}

> **问自己**:

#### 11. f-string 进阶 —— 对齐 / 小数

> **痛点**: 默认 f-string 对不齐,小数位数不确定时怎么格式化?
> **类比**: 像 Excel 单元格格式:居左/右、保留 N 位小数。
> **解释**: 花括号内写 `变量:格式`,`{pi:.2f}` 保留两位,
> `{n:>10}` 右对齐宽度 10。

````python
# 示例:保留 N 位小数 + 对齐
pi = 3.14159265

# --- 执行过程 ---
# {pi:.2f} .2 → 保留 2 位小数,f → 浮点数

print(f"圆周率约 {pi:.2f}")
print(f"圆周率约 {pi:.4f}")

# 对齐
h1 = "书名"; h2 = "价格"
print(f"{h1:>8} {h2:>8}")
t = "Python 入门"; p = 59.8
print(f"{t:>8} {p:>8}")
````

**逐行解剖**:
- `:.N f` → 保留 N 位小数(四舍五入)
- `:>N` → 右对齐,占 N 字符宽
**常见错误**:
1. 忘前缀 f → 格式语法不会被解析
2. `{:2f}` 少写点 → 被当宽度 2,应是 `{:.2f}`

#### 15. 综合练习 4 —— 解析 URL

从 `url = "https://www.example.com/search?q=python"` 中提取:

- 协议 (`https`)
- 域名 (`www.example.com`)

````python
# 示例
url = "https://www.example.com/search?q=python"

# --- 执行过程 ---
# proto_end = url.find("://") → 5
# protocol = url[:5] → "https"

proto_end = url.find("://")
protocol  = url[:proto_end]
print(protocol)

# 提取域名
rest = url[proto_end + 3:]        # www.example.com/search?q=python
slash = rest.find("/")
host = rest[:slash]
path = rest[slash:]
print(host); print(path)
````

**逐行解剖**:
- `find("://")` 返回 "://" 起始位置

**今日小结**

| 概念 | 关键语法 |
| --- | --- |
| 三种引号 | `'..'` `".."` `'''..'''` |
| 索引 | 正数 0 起,负数倒数,s[-1] |
| 切片 | s[start:end],左闭右开 |
| 步长 | s[::2] 跳取,s[::-1] 反转 |
| len | 字符数,中文也算 1 |
| in | "X" in s → True/False |
| find/rfind | 找不到返回 -1 |
| 四件套 | replace / strip / join / split |
| 不可变 | 不能 s[i] = ...,必须构造新串 |
| f-string | f"{x}" 嵌入变量 |
| f-string 格式 | :.N f / :>N |
| 综合 | slicing + join + find + f-string |

**易错提醒**

- 切片 end **不包含**,长度 = end - start
- find 找不到返回 **-1**(不是 None)
- 字符串 **不可变**,replace 要接返回值
- f-string 必须带前缀 f,否则 {{}} 原样输出
