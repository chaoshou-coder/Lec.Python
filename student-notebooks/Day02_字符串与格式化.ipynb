#### 1. 字符串创建 —— 三种引号> **痛点**: 文字里也有引号怎么办?> **类比**: 单/双引号就像左右手,完全等价,选一种即可。> **解释**: 用 **单引号** **双引号** **三引号** 三种方式创建字符串。> 当文字本身含引号,用另一种引号框住即可。**ASCII 内存图**:```s = "悟空" → s ──→ 连续内存块:字符 '悟'(0) '空'(1)```**关键**:字符按顺序排,每人编号 0、1、2……

````python
# 示例:三种引号s1 = '单引号'; s2 = "双引号"; s3 = '''三引号'''# --- 执行过程 ---# s1, s2, s3 分别赋值引号内的内容# 单引号(')与双引号(")完全等价print(s1); print(s2); print(s3)# 引号嵌套line1 = '悟空说:"俺老孙来也"'line2 = "她说:'你好'"# --- 执行过程 ---# 外层用单/双引号,内部用另一种作为普通字符print(line1); print(line2)
````

#### 2. 索引 —— 给每个字符编门牌号> **痛点**: "孙悟空" 要取 "悟"(第二个字),怎么做?> **类比**: 队伍同学编号 — 正数 0 起,负数 -1 倒数。> **解释**: `s[0]` = 第一个字符; `s[-1]` = 最后一个字符。**ASCII 内存图**:```s = "孙悟空"正数索引:  0('孙')  1('悟')  2('空')负数索引:    -3       -2       -1```

````python
# 示例:正数索引 + 负数索引name = "孙悟空"# --- 执行过程 ---# name[0] → 索引 0,即 '孙'# name[1] → 索引 1,即 '悟'# name[2] → 索引 2,即 '空'# name[-1] → 倒数第一,即 '空'# name[-2] → 倒数第二,即 '悟'# name[-3] → 倒数第三,即 '孙'print(name[0])print(name[1])print(name[2])print(name[-1])print(name[-2])print(name[-3])
````

#### 3. 切片 —— 一次截取一串> **痛点**: 从邮箱 `sunwu@163.com` 取用户名/域名,怎么批量?> **类比**: 切香肠 — 刀起 start,刀落 end,**包含 start, 不包含 end**。> **解释**: `s[start:end]`,省略 start 默认 0,省略 end 默认末。**ASCII 内存图**:```s = "0123456789"s[2:6] 范围:2 3 4 5 (6 不包含)结果: "2345"  (长度 6-2 = 4)口诀:刀口落在 end,那块不拿。```

````python
# 示例:完整切片 + 省略 start/ends = "abcdefghij"# --- 执行过程 ---# s[2:6] → 索引 2、3、4、5 → "cdef"# s[:3]  → 起默认 0,到 3 之前 → "abc"# s[7:]  → 从 7 到末 → "hij"# s[:]   → 复制整串 → "abcdefghij"print(s[2:6])print(s[:3])print(s[7:])print(s[:])# 实际应用:从邮箱取用户名 + 域名email = "sunwu@163.com"print(email[:5])print(email[6:])
````

````python
# 示例:正步长 + 负步长s = "abcdefghijk"# --- 执行过程 ---# s[::2]  → 0、2、4、6、8、10 → "acegik"# s[::-1] → 从右往左逐个取 → "kjihgfedcba"# s[8:1:-2] → 从 8 往左到 1 前,跨 2 格#           → 8、6、4、2 → "igec"print(s[::2])print(s[::-1])print(s[8:1:-2])
````

#### 11. f-string 进阶 —— 对齐 / 小数> **痛点**: 默认 f-string 对不齐,小数位数不确定时怎么格式化?> **类比**: 像 Excel 单元格格式:居左/右、保留 N 位小数。> **解释**: 花括号内写 `变量:格式`,`{pi:.2f}` 保留两位,> `{n:>10}` 右对齐宽度 10。

````python
# 示例:保留 N 位小数 + 对齐pi = 3.14159265# --- 执行过程 ---# {pi:.2f} .2 → 保留 2 位小数,f → 浮点数print(f"圆周率约 {pi:.2f}")print(f"圆周率约 {pi:.4f}")# 对齐h1 = "书名"; h2 = "价格"print(f"{h1:>8} {h2:>8}")t = "Python 入门"; p = 59.8print(f"{t:>8} {p:>8}")
````

**逐行解剖**:- `:.N f` → 保留 N 位小数(四舍五入)- `:>N` → 右对齐,占 N 字符宽**常见错误**:1. 忘前缀 f → 格式语法不会被解析2. `{:2f}` 少写点 → 被当宽度 2,应是 `{:.2f}`> **问自己**:> - `{:.2f}` 和 `{.2f}` 哪个对?> - `{n:>10}` `<` `>` `^` 分别什么对齐?> - `a = 3.14159; print(f"{a:.2f}")` 输出什么?
