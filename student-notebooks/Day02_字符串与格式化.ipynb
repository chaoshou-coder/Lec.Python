**第 1 讲:字符串创建与转义字符**

字符串可以用 **双引号** 或 **单引号** 括起来,两种写法完全等价。当字符串里需要出现引号本身时,用反斜杠 **\** 进行 **转义**。

```python
# 双引号和单引号等价
s1 = "双引号字符串"
s2 = '单引号也行'

# 转义字符:用 \\ 让引号变成普通字符,不提前结束字符串
s3 = "He said \"Hi\""
s4 = 'It\'s ok'
```

**转义字符速查表**:**\n** = 换行;**\t** = 制表符;**\\** = 反斜杠本身;**\"** / **\'** = 引号。

---

**第 2 讲:索引 —— 给每个字符编上门牌号**

字符串里的每个字符都有一个 **索引**(门牌号),从 **0** 开始数。负数倒着数,`-1` 是最后一个字符。

口诀:**从零开始,负数倒着数**。

```python
name = "孙悟空"
# 正数索引:从左往右,从 0 开始
print(name[0])    # 孙
print(name[1])    # 悟
print(name[2])    # 空
# 负数索引:从右往左,-1 是最后一个
print(name[-1])   # 空
print(name[-2])   # 悟
```

---

**第 3 讲:切片 —— 一次切出一段**

**s[start:end]** 从 `start` 切到 `end` **之前**(左闭右开)。就像切蛋糕 —— 刀口落在 `end`,但 `end` 那块不拿。

省略 `start` 默认从 0 开始;省略 `end` 默认切到末尾。

```python
s = "hello world"
print(s[0:5])     # hello
print(s[6:11])    # world
print(s[:5])      # hello —— 省略 start,从 0 开始
print(s[6:])      # world —— 省略 end,切到末尾
print(s[:])       # hello world —— 全切,相当于复制
```

---

**第 4 讲:步长 s[i:j:step]**

在切片基础上加第三个参数 `step`,表示每隔几个字符取一个。`step` 为负数时,反向取值。

```python
s = "abcdefghijk"
# step=2:每隔一个取一个
print(s[::2])     # acegik

# step=-1:逆序 —— 整串倒过来
print(s[::-1])    # kjihgfedcba
```

---

**第 5 讲:len() 求长度**

**len()** 返回字符串里有多少个字符,也适用于列表、字典等其他容器。

```python
s = "hello world"
print(len(s))     # 11
```

---

**第 6 讲:find() 与 rfind() 查找子串**

**find(sub)** 返回子串第一次出现的索引;**rfind(sub)** 返回最后一次出现的索引。找不到都返回 **-1**。

类比:在文件路径 `/tmp/data/a.txt` 中,想取最后一个 `/` 之后的内容,只能倒着找 —— 用 **rfind("/")**。

```python
s = "hello hello hello"
print(s.find("hello"))    # 0 —— 第一次
print(s.rfind("hello"))   # 12 —— 最后一次
print(s.find("python"))   # -1 —— 找不到
```

---

**第 7 讲:字符串判断方法 —— 是不是数字?是不是字母?**

这些方法都返回 **True** / **False**,常与 `if` 搭配做输入校验。

```python
# isdigit:判断是否全是数字
print("12345".isdigit())     # True
print("12.3".isdigit())      # False —— 小数点不是数字

# isalpha:判断是否全是字母
print("abc".isalpha())       # True
print("abc123".isalpha())    # False

# isalnum:判断是否只有字母或数字(不含符号)
print("abc123".isalnum())    # True
print("abc-123".isalnum())   # False —— 横杠是符号
```

---

**第 8 讲:replace() 批量替换**

**str.replace(old, new)** 把字符串中 **所有** `old` 子串替换成 `new`,返回新字符串。

注意:字符串是 **不可变** 的 —— 原串不会被修改,必须用变量接住返回值。

```python
# 把所有 "xxx" 替换为 "***"
text = "xxx 是个好网站,xxx 很好用"
clean = text.replace("xxx", "***")
print(clean)              # *** 是个好网站,*** 很好用
print(text)               # 原串未变
```

```python
# 限制替换次数:第三个参数 count 表示最多替换几次
print("aaa aaa aaa".replace("aaa", "b", 2))   # b b aaa
```

---

**第 9 讲:f-string 格式化输出**

**f-string**(Python 3.6+)是最推荐的格式化方式。在字符串前加 `f`,用 **{}** 包裹变量或表达式,直接嵌入。

```python
name = "小明"
age = 18

# f-string:大括号里直接写变量名,最直观
print(f"我是 {name},今年 {age} 岁")
```

```python
# 大括号里可以写表达式
print(f"明年 {age + 1} 岁")
```

```python
# 保留两位小数 ::.2f
pi = 3.14159
print(f"圆周率约等于 {pi:.2f}")
```

**今日小结**

今天学会了字符串的核心工具箱:
- **索引** 从零开始,负数倒着数
- **切片** 左闭右开,省略首尾等价于从头/到尾
- **步长** `s[::2]` 跳着取,`s[::-1]` 倒序
- **find / rfind** 找第一次/最后一次出现的位置,找不到返回 -1
- **isdigit / isalpha / isalnum** 判断字符串构成
- **replace** 批量替换,原串不变,要用变量接返回值
- **f-string** 用大括号嵌入变量和表达式

> 易错点:切片索引容易差一(第 n 个字符的索引是 n-1);**find()** 找不到返回 -1 而不是 None;字符串不可变,**replace()** 要接返回值。
