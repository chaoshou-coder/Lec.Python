**第 1 讲:print() 的四种用法**

**print()** 是 Python 最常用的输出函数。把内容放进括号,电脑就会把它 **打印** 到屏幕上。

类比:**print()** 就像一个对话框,括号里写啥,屏幕上就显示啥。

```python
# 用法 1:基础输出 —— 打印一个字符串
print("Hello World")
```

```python
# 用法 2:输出中文 —— Python 3 原生支持,直接写
print("你好,我是 Python")
```

```python
# 用法 3:一次打印多个值 —— 逗号分隔,Python 自动加空格
print("刘备", "关羽", "张飞")
```

```python
# 用法 4:字符串拼接 —— + 号粘在一起,没有空格
print("刘" + "关" + "张")
```

---

**第 2 讲:print() 进阶 —— 换行与不换行**

**\n** 是换行符,相当于在字符串里按了一下回车。**end=""** 可以取消 print 默认的末尾换行,让下一次输出紧接在同一行。

```python
# \n 换行 —— 把一行拆成两行
print("第一行\n第二行")
```

```python
# end="" 取消默认换行 —— 下一次 print 继续在同一行输出
print("不换行", end="")
print("接续这一行")
```

**第 3 讲:input() 获取输入**

**input()** 让程序暂停,在屏幕上显示提示语,等你从键盘输入内容并按回车。

类比:**input()** 就像一个话筒 —— 你把说的话传给了程序。

```python
# input() 显示提示语,等你输入;按回车后把输入的内容作为字符串返回
name = input("请输入你的姓名:")
print("你好," + name)
```

**永远返回字符串**:**input()** 返回的永远是字符串类型。即使你输入数字 `18`,得到的也是字符串 `"18"`,而不是数字 `18`。

如果需要把输入的内容当作数字用,必须用 **int()** 或 **float()** 转换。

```python
# input() 返回的是字符串,字符串 + 整数会报错
# 正确做法:先用 int() 转换成整数再算
age = input("请输入年龄:")
age_num = int(age)    # 字符串 "18" → 整数 18
print(age_num + 5)     # 可以正常算数了
```

**第 4 讲:变量与赋值**

**变量** 就是贴了标签的盒子。你把数据装进盒子,以后用名字就能找到它。

**=** 不是数学里的 "等于",而是 **赋值**:把右边的值放进左边的盒子。`==" 才是比较。

```python
# 创建变量 —— 把数据装进贴了标签的盒子
name = "小明"       # 字符串
age = 18            # 整数
height = 1.75       # 浮点数
is_student = True   # 布尔值
```

```python
# 变量可以被多次赋值 —— 新值覆盖旧值
x = 10
print(x)       # 10
x = 20          # 用新值 20 覆盖旧值 10
print(x)       # 20
```

```python
# 把一个变量的值赋给另一个变量 —— 复制了值,不是共享
a = 10
b = a           # b 复制了 a 的值,此时 b 也是 10
a = 99          # 修改 a 不影响 b
print(b)        # 仍然是 10
```

**变量命名规则**:由 字母/数字/下划线 组成,不能以数字开头;区分大小写;不能是保留字(如 if/for/print)。

约定写法:**snake_case**(小写 + 下划线),例如 `user_name`。

```python
# 合法变量名
user_name = "张三"
_age = 18
name2 = "李四"

# 非法变量名(写了会报错)
# 2name = "张三"     # 数字开头
# my-name = "张三"   # 横杠不行
# if = 18            # 保留字
```
