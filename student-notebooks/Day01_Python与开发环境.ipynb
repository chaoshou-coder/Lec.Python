## 第一讲(15 分钟)—— `print` 与 Hello World

### 概念解释

类比: `print()` 就像一个**对话框**,你往括号里放什么,电脑就帮你显示什么。

```
print("内容")
  ↓       ↓
 函数   要显示的内容(字符串)
```

````python
# 示例代码:让电脑开口说话
print("Hello World")        # 让电脑在屏幕上打印一行字
print("你好,我是 Python")  # 中文也行,直接写
````

> 🔴 **易错点**: 请使用**英文输入法**! 中文括号 `（）` 和中文引号 `""` 会导致 `SyntaxError`

### `print()` 的高级用法

- 多个值用逗号分隔(自动空格)
- `+` 是字符串拼接(没有空格)
- `\n` 是换行符

````python
# 一次打印多个值
print("刘备", "关羽", "张飞")  # 刘关张(空格分隔)
print("刘备" + "关羽")         # 刘备关羽(字符串拼接)

# 换行
print("第一行\n第二行")          # \n = 换行

# 不换行
print("不换行", end="")         # end="" 取消默认换行
print("接续这一行")
````

**口诀**: `print` 就是对话框,括号里写啥出啥,多个值用逗号。

## 第二讲(15 分钟)—— `input` 获取输入

### 概念解释

类比: `input()` 就像**一只话筒**,按回车之前电脑会一直等你。

```
变量 = input("提示文字")
         ↓
    显示提示,等你输入
         ↓
    返回你输入的内容(永远是字符串!)
```

````python
# 示例代码:和小明对话
name = input("请输入你的姓名:")  # 屏幕显示提示,等你输入
print("你好," + name)
````

````python
# ⚠️ input 返回的是字符串!
age = input("请输入年龄:")       # age 现在是字符串 "18",不是数字 18
print(type(age))                 # <class 'str'>

# 想做数学运算,需要先转换
age_num = int(age)               # 把字符串 "18" 转成整数 18
print(age_num + 1)               # 19
````

> 🔴 **教学红线**: `input` 永远返回字符串,需要数字请 `int()` / `float()` 裹一层!

**记忆口诀**: `input` 永远是字符串,需要数字请 `int()` / `float()` 裹一层。

## 第三讲(15 分钟)—— 变量与命名

### 概念解释

类比: 变量 = **贴标签的盒子**。把数据装进盒子,以后用名字就能取出来。

```
变量名 = 数据
  ↓      ↓
 标签   盒子里装的东西
```

**四种基本数据**:
- `int` 整数: `18`, `-5`, `100`
- `float` 小数: `3.14`, `-0.5`
- `str` 字符串: `"hello"`, `'你好'`
- `bool` 布尔: `True`, `False`

````python
# 示例代码:变量的创建
name = "小明"       # 把字符串装进 label 叫 name 的盒子
age = 18            # 整数
height = 1.75       # 浮点数
is_student = True   # 布尔

print(type(name))       # <class 'str'>
print(type(age))        # <class 'int'>
print(type(height))     # <class 'float'>
print(type(is_student)) # <class 'bool'>
````

### 变量命名规则

✅ 合法:
- 字母、数字、下划线组成
- 不能以数字开头
- 区分大小写

❌ 非法:
- `2name` (数字开头)
- `my-name` (横杠不行)
- `if` (保留字)

````python
# 合法命名
user_name = "张三"     # snake_case 下划线风格
_age = 18
name2 = "李四"

# 以下会报错,取消注释看看
# 2name = "张三"     # 数字开头
# my-name = "张三"   # 横杠不行
# if = 18            # 保留字
````

### 赋值与重新赋值

```python
x = 10
x = 20          # x 被重写为 20

a = 10
b = a           # b 复制了 a 的值
a = 99
print(b)        # 10  ← 改 a 不影响 b
```

> 🔴 **数学等号 ≠ 赋值号**: 数学里 `x = x + 1` 永远不成立; Python 里表示"让 x 在原来基础上加 1"。

把 `=` 理解成"**把右边的结果装进左边的盒子**"。

````python
# 示例:赋值与重新赋值
a = 10
b = a           # b 复制了 a 的值 10
a = 99
print("a =", a)  # 99
print("b =", b)  # 10 ← 改 a 不影响 b

# x = x + 1 的含义
x = 10
x = x + 1       # 把 x+1 的结果装回 x
print("x =", x)  # 11
````

## 第四讲(15 分钟)—— `type()` 观察类型

### 概念解释

电脑需要知道一个数据是什么类型,才能决定能不能做运算。
`type()` 就像**放大镜**,可以看到数据的真实身份。

````python
# 示例:观察四种基本类型
print(type(42))         # <class 'int'>
print(type(3.14))       # <class 'float'>
print(type("hello"))    # <class 'str'>
print(type(True))       # <class 'bool'>

# 常见的类型转换
print(int("42"))        # 42 (字符串 → 整数)
print(float("3.14"))    # 3.14 (字符串 → 小数)
print(str(100))         # "100" (数字 → 字符串)
````
