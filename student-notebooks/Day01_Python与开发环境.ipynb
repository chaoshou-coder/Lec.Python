## 第一讲(15 分钟) —— print 与 Hello World

### 📖 核心概念

`print()` = 让电脑在屏幕上说话。

类比:对着麦克风喊,全场都听到。

````python
# 示例:第一句代码
# 让电脑打印一行字
print("Hello World")

# 中文也行,直接写
print("你好,我是 Python")
````

### 💡 一次打多个值

- 逗号分隔 → 自动加空格
- `+` 拼接 → 没有空格
- `\n` 换行

````python
# 多个值用逗号 → 空格分隔
print("刘备", "关羽", "张飞")

# + 号拼接 → 没有空格
print("刘备" + "关羽")

# \n = 换行
print("第一行\n第二行")

# end="" 取消默认换行
print("不换行", end="")
print("接续这一行")
````

> 🔴 **易错点**:括号、引号必须用**英文符号**!
>
> 中文 `（）` 或 `""` 会报 `SyntaxError`。

## 第二讲(15 分钟) —— input 获取输入

### 📖 核心概念

`input()` = 让电脑停下来等你打字。

类比:考试卷上的填空题,等你写答案。

````python
# 示例:让电脑问名字
# 屏幕显示提示,等你按回车
name = input("请输入你的姓名:")

# 把你输入的内容拼出来
print("你好," + name)
````

### 💡 input 返回的是字符串

想做数学运算,必须先转类型!

````python
# ⚠️ age 现在是字符串 "18",不是数字 18
age = input("请输入年龄:")

# 用 type() 查看类型
print(type(age))  # <class 'str'>

# int() 把字符串转成整数
age_num = int(age)
print(type(age_num))  # <class 'int'>
print(age_num + 1)    # 可以算数了
````

> 🔴 **易错点**:`input` 永远是字符串!
>
> 要算数必须先 `int()` 或 `float()` 转换。

## 第三讲(15 分钟) —— 变量与命名

### 📖 核心概念

变量 = 给数据贴标签,方便以后找。

类比:快递柜上的取件码。

````python
# 示例:把数据装进盒子
name = "小明"        # 字符串 str
age = 18             # 整数 int
height = 1.75        # 浮点数 float
is_student = True    # 布尔 bool

# 用 type() 查看每个盒子的类型
print(type(name))       # <class 'str'>
print(type(age))        # <class 'int'>
print(type(height))     # <class 'float'>
print(type(is_student)) # <class 'bool'>
````

### 📖 命名规则

- 由**字母、数字、下划线**组成
- 不能以数字开头
- 区分大小写
- **约定**:小写 + 下划线(`snake_case`)

````python
# ✅ 合法命名
user_name = "张三"
_age = 18
name2 = "李四"

# ❌ 非法命名(被注释掉了)
# 2name = "张三"    # 数字开头
# my-name = "张三"  # 横杠不行
# if = 18           # 保留字
````

### 💡 赋值与重新赋值

````python
# x 被重写为 20
x = 10
x = 20
print(x)  # 20

# b 复制了 a 的值
a = 10
b = a
a = 99
print(b)  # 10 ← 改 a 不影响 b

# x = x + 1 的含义
x = 10
x = x + 1  # 把 x+1 的结果装回 x
print(x)   # 11
````

> 🔴 **易错点**:`=` 是赋值,不是数学等号!
>
>`x = x + 1` 意思是"让 x 在原来基础上加 1"。

## 第四讲(15 分钟) —— type() 观察类型

### 📖 核心概念

电脑需要知道数据是什么类型,才能决定能不能做运算。

`type()` = 放大镜,看数据的真实身份。

````python
# 示例:观察四种基本类型
print(type(42))      # <class 'int'>
print(type(3.14))    # <class 'float'>
print(type("hello")) # <class 'str'>
print(type(True))    # <class 'bool'>

# 类型转换
print(int("42"))     # 42 (字符串 → 整数)
print(float("3.14")) # 3.14 (字符串 → 小数)
print(str(100))      # "100" (数字 → 字符串)
````
