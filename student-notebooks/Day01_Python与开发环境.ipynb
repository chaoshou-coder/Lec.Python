**1. print —— 让电脑开口说话**

类比:`print()` = 对着麦克风喊,全场都听到。括号里放你想让电脑显示的内容,运行后就会打印到屏幕上。

```python
# 最经典的 Hello World
print("Hello World")

# 中文也行,直接写
print("你好,我是 Python")

# 一次打印多个值,逗号分隔 → 自动加空格
print("刘备", "关羽", "张飞")  # 刘备 关羽 张飞

# + 号拼接 → 没有空格
"刘备" + "关羽"  # 刘备关羽
```

`\n` = 换行符,遇到它就跳到下一行。`end=""` 可以取消 print 默认的末尾换行,让两次打印接在同一行。

```python
# \n = 换行
print("第一行\n第二行")

# end="" 取消默认换行,两次打印接一起
print("不换行", end="")
print("接续这一行")  # 不换行接续这一行
```

**2. input —— 让电脑停下来等你打字**

类比:`input()` = 考试卷上的填空题,程序暂停,等你按回车才继续。括号里的字符串是提示语,不是结果。

```python
# input 会暂停程序,等你按回车
name = input("请输入你的姓名:")

# 把你输入的内容拼出来
print("你好," + name)

# ⚠️ input 返回的是字符串 "18",不是数字 18
age = input("请输入年龄:")
print(type(age))  # <class 'str'>

# int() 把字符串转成整数,才能做数学运算
age_num = int(age)
print(type(age_num))  # <class 'int'>
print(age_num + 1)    # 可以算数了
```

["**3. type 与数据类型**\n", "\n", "电脑需要知道数据是什么类型,才能决定能不能做运算。\n", "`type()` = 放大镜,看数据的真实身份。\n", "Python 四种基本类型:`int` `float` `str` `bool`。"]

```python
# 观察四种基本类型
print(type(42))       # <class 'int'>    整数
print(type(3.14))     # <class 'float'>  小数
print(type("hello"))  # <class 'str'>    字符串
print(type(True))     # <class 'bool'>   布尔

# 类型转换:字符串 ↔ 数字
print(int("42"))      # 42 (字符串 → 整数)
print(float("3.14"))  # 3.14 (字符串 → 小数)
print(str(100))       # "100" (数字 → 字符串)
```

**4. 变量 = 给数据起名字**

类比:变量 = 快递柜上的取件码。把数据装进盒子,以后用名字就能找到它。`=` 是赋值,不是数学等号 —— 意思是"把右边的值装进左边的盒子"。

```python
# 把数据装进盒子
name = "小明"        # 字符串 str
age = 18             # 整数 int
height = 1.75        # 浮点数 float
is_student = True    # 布尔 bool

# 查看每个盒子的类型
print(type(name))       # <class 'str'>
print(type(age))        # <class 'int'>
print(type(height))     # <class 'float'>
print(type(is_student)) # <class 'bool'>
```

命名规则:由**字母、数字、下划线**组成;不能以数字开头;区分大小写;不能用保留字(if/for/while 等)。约定俗成:**snake_case**(小写 + 下划线)。

```python
# ✅ 合法命名
user_name = "张三"
_age = 18
name2 = "李四"

# ❌ 非法命名(被注释掉了,取消注释会报错)
# 2name = "张三"    # 数字开头
# my-name = "张三"  # 横杠不行
# if = 18           # 保留字
```

赋值与重新赋值:`x = x + 1` 意思是"把 x 加 1 的结果装回 x",也就是自增。`b = a` 是把 a 的值**复制**一份给 b,改 a 不影响 b。

```python
# x 被重写为 20
x = 10
x = 20
print(x)  # 20

# b 复制了 a 的值,改 a 不影响 b
a = 10
b = a
a = 99
print(b)  # 10

# x = x + 1 的含义:自增
x = 10
x = x + 1  # 把 x+1 的结果装回 x
print(x)   # 11
```

**5. 易错点与今日小结**

今日核心三件套:`print` 输出、`input` 输入、变量存储。

```python
# ======================
# 学员代码区:自我介绍卡片
# ======================
# 题目:输入姓名、年龄、爱好,
#       输出一张自我介绍卡片。
#
# 输入示例:
#   请输入姓名:小明
#   请输入年龄:18
#   请输入爱好:打篮球
#
# 输出示例:
#   ==========
#   姓名:小明
#   年龄:18
#   爱好:打篮球
#   ==========

# 获取输入
name = input("请输入姓名:")
age = input("请输入年龄:")
hobby = input("请输入爱好:")

# 输出卡片
print("==========")
print("姓名:" + name)
print("年龄:" + age)
print("爱好:" + hobby)
print("==========")
```

```python
# 参考答案(用 f-string 更简洁)
name = input("请输入姓名:")
age = input("请输入年龄:")
hobby = input("请输入爱好:")

# f"...{变量}..." 自动把变量值嵌入字符串
print("==========")
print(f"姓名:{name}")
print(f"年龄:{age}")
print(f"爱好:{hobby}")
print("==========")
```

> 🔴 **易错点**:
> 1. 中英文符号混用 → `SyntaxError`,括号引号必须用**英文符号**
> 2. `input()` 返回字符串,直接算数报错 → 必须先 `int()` 或 `float()`
> 3. `=` 是赋值,`==` 才是比较
> 4. `print("a" + 1)` 报错 → 字符串和整数不能直接相加
