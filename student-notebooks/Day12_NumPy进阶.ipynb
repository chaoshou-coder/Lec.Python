**第 1 讲:逐元素乘法 vs 矩阵乘法**

口诀:**`*` 是"对位乘",`@` 是"行乘列"**。`np.dot(a, b)` 等价于 `a @ b`。矩阵乘法**不满足交换律**,`A @ B` ≠ `B @ A`,顺序不能换。常用线性代数函数:`a.T` 转置 / `np.linalg.det` 行列式 / `np.linalg.inv` 逆矩阵 / `np.linalg.eig` 特征值。


```python
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
b = np.array([[5, 6],
              [7, 8]])

# 逐元素乘法(对应位置相乘)
print(a * b)
# [[ 5 12]
#  [21 32]]

# 矩阵乘法(线性代数,行乘列)
print(a @ b)
# [[19 22]
#  [43 50]]

# 转置
print(a.T)
# 行列式
print(np.linalg.det(a))             # -2.0
# 验证 A @ inv(A) = 单位矩阵 I
inv = np.linalg.inv(a)
print(np.round(a @ inv, 5))         # [[1. 0.] [0. 1.]]
```

**聚合函数 + 轴(axis)概念**

对整个数组计算:`sum/mean/std/max/min/argmax`。`axis` 沿哪个轴聚合:**`axis=0` 跨行计算 → 剩列;`axis=1` 跨列计算 → 剩行**。记忆口诀:"axis=0 跨行,结果按列下来"。


```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])

# 不指定 axis,对整个数组聚合
print(a.sum())                      # 21
print(a.mean())                     # 3.5
print(a.std())                      # 1.707...

# axis=0:沿行方向(压扁行),结果长度 = 列数
print(a.sum(axis=0))                # [5 7 9] 每列求和
# axis=1:沿列方向(压扁列),结果长度 = 行数
print(a.sum(axis=1))                # [ 6 15] 每行求和
```

**第 2 讲:随机数生成 + 可复现种子**

机器学习实验需要**可复现**,设 `np.random.seed(42)` 后每次运行得到相同随机数。常用:`rand` 均匀 [0,1) / `randn` 标准正态 / `normal(均值, 标准差, size)` / `randint(low, high, size)`。


```python
import numpy as np

# 设种子,保证每次运行随机数一致
np.random.seed(42)

# 3 行 2 列均匀分布随机矩阵
print(np.random.rand(3, 2))
# 5 个标准正态分布 N(0,1) 随机数
print(np.random.randn(5))
# 5 个均值 10、标准差 2 的正态随机数
print(np.random.normal(10, 2, 5))
# 10 个 [1,7) 整数(模拟骰子)
print(np.random.randint(1, 7, 10))
```

**第 3 讲:文件读写 save / load / savetxt / loadtxt**

二进制 `.npy` 单数组 / `.npz` 多数组;文本 `.csv` 用 `savetxt` / `loadtxt`,注意 `delimiter`、`skiprows`、`comments=""`。


```python
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])

# 二进制 .npy 格式(保存单数组)
np.save("data.npy", a)
b = np.load("data.npy")
print(b)

# 二进制 .npz 格式(保存多个数组)
np.savez("data.npz", x=a, y=a * 2)
d = np.load("data.npz")
print(d["x"])
print(d["y"])

# 文本 .csv 格式
np.savetxt("output.csv", a, delimiter=",",
           header="col1,col2,col3", comments="")
loaded = np.loadtxt("output.csv", delimiter=",")
print(loaded)
```
