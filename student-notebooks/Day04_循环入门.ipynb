#### 为什么需要循环 —— 消灭重复代码

> **痛点**:同样语句复制粘贴 10 份,改需求要手动改 10 处,
>  一处漏改出 bug。
> **类比**:手写 100 份请帖 vs 打印机印 100 份,
>  循环 = 代码的打印机。
> **解释**:**DRY 原则** = Don't Repeat Yourself,
>  同一动作重复 N 次,用循环代劳。


````python
# 写法 A 复制粘贴 (反例)
print("打卡成功")
print("打卡成功")
print("打卡成功")
print("打卡成功")
print("打卡成功")
print("------------")

# 写法 B 循环 (正例)
for _ in range(5):
    print("打卡成功")

# --- 执行过程 ---
# 第 9 行 for _ in range(5):
#   ① range(5) 生成 0,1,2,3,4 共 5 个数
#   ② _ 第一次取值 0,进入循环体
#   ③ print 输出 "打卡成功" 第 1 次
#   ④ 回到 for,_ 取值 1 → 输出第 2 次
#   ⑤ 直到 5 个数全都取完,循环结束
````

> **ASCII 对比 + 逐行解剖 + 常见错误 + 引导**

```
复制粘贴:                    循环:
┌──────────────┐            ┌─────────────────────┐
│ 10 行同段代码  │            │ for _ in range(N):  │
│ 改 10 处      │            │     同一段代码       │
└──────────────┘            │     ← 只写一次       │
                            └─────────────────────┘
                              改 1 处,多处生效
```

> - `for _ in range(5):` —— `_` 占位变量,"不关心当前值"
> - `range(5)` —— 生成 0 到 4 共 5 个整数
> - 循环体 **缩进 4 格**,for 共跑 5 次
> - 1. **忘缩进**:`for i in range(3):print("x")`
>   → IndentationError
> - 2. **以为 range(5) 含 5**:实际不含,只到 4

#### while 循环 —— "当...时就继续"

> **痛点**:不知循环多少次,只知"满足条件就继续"。
> **类比**:跑步机 —— 心率低于目标值时继续,达标就停。
> **解释**:`while 条件:` 条件 **True** 执行循环体,
>  **False** 退出循环。


````python
# 不停询问密码,直到输入 123 才停
password = ""
while password != "123":
    password = input("请输入密码:")
print("密码正确,欢迎!")

# --- 执行过程 ---
# 第 1 行 password = "":
#   ① 创建 password, 初值空字符串
#
# 第 2 行 while password != "123":
#   ① "" != "123" → True → 进循环
#
# 第 3 行 password = input(...):
#   ① 显示提示,等待用户输入
#   ② 把输入赋给 password
#
# 回到第 2 行再判断:
#   ① 若输入 "abc" → True → 继续循环
#   ② 若输入 "123" → False → 退出循环
````

> **ASCII 判断流程 + 逐行解剖**

```
      ┌────────────┐
      │  判断条件    │
      └──────┬─────┘
         T  │      F
      ┌─────┘      └──┐
      ▼                ▼
┌──────────┐      ┌────────┐
│ 执行循环体 │      │ 跳出循环│
└────┬─────┘      └────────┘
     └──────────────┐
              (再判断)
```

> - `while 条件:` 末尾须有冒号,下方缩进为循环体
> - 循环体至少一行,空循环写 pass
> - 条件每次循环开始时重新判断
> - 1. **忘冒号**:`while password != "123"` → SyntaxError

#### while 三要素 —— 初始化 / 条件 / 更新

> **痛点**:while 易错 —— 忘初值会 NameError,
>  忘更新会死循环,条件写反也是死循环。
> **类比**:跑步机三要素 —— 开机(初始化) /
>  设目标心率(条件) / 逐渐提速(更新)。


````python
# 打印 1 到 5 (while 实现)
count = 1              # 要素 1:初始化
while count <= 5:      # 要素 2:条件
    print(count)
    count = count + 1  # 要素 3:更新
print("循环结束")

# --- 执行过程 ---
# 第 1 行 count = 1: ① 初始化 count = 1
# 第 2 行 while count <= 5:
#   ① 第 1 轮: 1 <= 5 → True → 进循环体
#   ② print 输出 1, count = 1+1 = 2
# 回到第 2 行再判断:
#   ① c=2→输出 2,c=3; c=3→输出 3,c=4
#   ② c=4→输出 4,c=5; c=5→输出 5,c=6
#   ③ 6 <= 5 → False → 退出循环
````

#### for + range —— 知道循环次数时首选

> **痛点**:while 必须手写三要素,简单计数太繁琐。
> **类比**:食堂窗口已知 5 个(for i in range(5)),
>  每个窗口打一次饭,打完就走。
> **解释**:`for 变量 in range(N):` 变量依次取
>  0,1,...,N-1,共 N 次。


````python
# 打印 0 到 4
for i in range(5):
    print(f"当前 i = {i}")
print("结束")

# --- 执行过程 ---
# 第 1 行 for i in range(5):
#   ① range(5) 生成序列 0,1,2,3,4
#   ② i 第一次取值 0
# 第 2 行 print(f"当前 i = {i}"):
#   ① 输出 "当前 i = 0"
# 回到第 1 行 i 自动取值 1:
#   ① 输出 "当前 i = 1"
# 依次到 i = 4 输出,循环结束
````

> **循环中修改 i 的陷阱**
> 在循环体里改 i,下一轮会被 range 的下一个值覆盖。
> - 问自己: `range(5)` 打印几行?分别是多少?
> - 问自己: `range(1)` 输出是什么?
> - 问自己: 循环体不打印 i 而打印 "hi" 会怎样?


````python
# 演示:循环内改 i,改完又被覆盖
for i in range(3):
    print(f"本轮开始 i = {i}")
    i = 999           # 改了也没用!
    print(f"本轮结束 i = {i}")
print("循环正常结束,range 不受影响")

# --- 执行过程 ---
# 第 1 轮: i=0 → print → i=999 → print → 回到 for
# 第 2 轮: i 被 range 重置为 1(不是 999)
# 第 3 轮: i 重置为 2
````

> **逐行解剖 + 常见错误**

> - `for i in range(5):` —— i 从 0 到 4,共 5 次
> - f-string `f"当前 i = {i}"` —— 把变量嵌入字符串
> - 循环体每次 i 取新值运行,循环体改 i 无效
> - 1. **以为 range(5) 含 5**:实际 0-4,左闭右开
> - 2. **在循环中改 i**:修改无效,被新值覆盖

#### range 的三种形式

> **痛点**:需指定起点、终点、步长时,记混参数含义。
> **类比**:跑道 —— **range(n)** 起点跑 n 步;
>  **range(a,b)** a 跑到 b;
>  **range(a,b,step)** 按步长跑。


````python
# 形式 1: range(stop) —— 0 到 stop-1
print("range(5):", end=" ")
for i in range(5):
    print(i, end=" ")
print()

# 形式 2: range(start,stop) —— 含 start,不含 stop
print("range(1,6):", end=" ")
for i in range(1, 6):
    print(i, end=" ")
print()

# 形式 3a: step 正数 —— 按步长跳
print("range(0,10,2):", end=" ")
for i in range(0, 10, 2):
    print(i, end=" ")
print()

# 形式 3b: step 负数 —— 倒序
print("range(5,0,-1):", end=" ")
for i in range(5, 0, -1):
    print(i, end=" ")
print()

# --- 执行过程(range(0,10,2)) ---
# 取 0 → +2 → 取 2 → +2 → 取 4 → ...
# 取 8 → +2=10 (≥ stop, 停止)
````

> **ASCII 三种形式对比 + 逐行解剖**

```
range(5)            → [0, 1, 2, 3, 4]
range(1, 6)         → [1, 2, 3, 4, 5]
range(0, 10, 2)     → [0, 2, 4, 6, 8]
range(5, 0, -1)     → [5, 4, 3, 2, 1]
```

> - `range(5)` —— 1 参数, 从 0, 不含 5
> - `range(1,6)` —— 2 参数, 从 1, 不含 6
> - `range(0,10,2)` —— 3 参数, 每次加 2
> - `range(5,0,-1)` —— step 负, 倒序, 不含 0
> - 1. **以为 stop 可含**:`range(1,10)` 最后是 9 不是 10
> - 2. **倒序 stop 写错**:想 5-1 却写 range(5,-1,-1) 多了 0
> - 问自己: `range(3,8)` 几次?哪些数?
> - 问自己: `range(10,0,-3)` 输出?
> - 问自己: 输出 5,4,3,2,1 参数?


````python
# 反向遍历: range(5,0,-1) 逐个打
for i in range(5, 0, -1):
    print(i)
print("发射!")

# --- 执行过程 ---
# i=5 → i=4 → i=3 → i=2 → i=1 → 结束
# 注意:0 不会被印出,因为 stop 不含
````

#### 嵌套循环 —— 循环里再写循环

> **痛点**:矩形、乘法表等"行列"结构,单层循环搞不定。
> **类比**:日历 —— 外层管"周",内层管"日"。
> **解释**:外层循环一次,**内层循环完整跑一圈**,
>  两层组合产生 **行 × 列** 效果。


````python
# 3 行 4 列的星号
for i in range(3):       # 外层:管行数
    for j in range(4):   # 内层:管列数
        print("*", end=" ")
    print()              # 每行结束换行

# --- 执行过程 ---
# 外层第 1 轮 i=0:
#   ① 内层 j=0,1,2,3 输出 "*"
#   ② 内层结束, print() 换行 → 第 1 行 "* * * *"
#
# 外层第 2 轮 i=1 → 内层完整跑 → 第 2 行
# 外层第 3 轮 i=2 → 内层完整跑 → 第 3 行
````

> **ASCII 嵌套执行轨迹**

```
外层 i=0 ──→ 内层 j=0 j=1 j=2 j=3 (内层完整跑完)
外层 i=1 ──→ 内层 j=0 j=1 j=2 j=3 (内层完整跑完)
外层 i=2 ──→ 内层 j=0 j=1 j=2 j=3 (内层完整跑完)
```

> 外层一次 = 内层一圈, 外层完再外, 直到全部跑完。


````python
# BREAK IT: 忘末尾 print() → 星号全挤一行
# 取消注释观察现象:
for i in range(3):
    for j in range(4):
        print("*", end=" ")
    # print()  # ← 注释掉, 再运行

# --- 执行过程(注释 print 后) ---
# print() 被注释, 内层输出后没换行
# 12 个星号全在一条横线
````

#### break —— 立刻退出整个循环

> **痛点**:满足条件后希望立刻结束,不再执行多余迭代。
> **类比**:翻书包找钥匙 —— 找到就停,不翻其余夹层。
> **解释**:`break` 终止 **离它最近的** 循环,
>  跳到循环后继续执行。


````python
# 在字符串中找第一个元音字母
word = "apple"
found = False
for ch in word:
    if ch in "aeiou":
        print(f"找到元音:{ch}")
        found = True
        break           # 找到就退出,不再继续
if not found:
    print("没找到元音")

# --- 执行过程 ---
# 第 1 轮 ch='a':
#   ① 'a' 在 "aeiou" → True
#   ② 输出 "找到元音:a", found=True
#   ③ break 立刻退出循环
# 后续 'p','p','l','e' 不再检查
````

> **BREAK IT 演示: 嵌套循环中 break 只退一层**

下面代码有一个嵌套循环,请观察 break 的影响范围。


````python
# break 只退内层, 不影响外层
for i in range(3):
    for j in range(3):
        if j == 1:
            break       # 只退内层, 外层继续
        print(f"i={i},j={j}")
    print("---外层一轮结束")

# --- 执行过程 ---
# i=0: 内层 j=0 输出 → j=1 break → 退出内层
# i=1: 内层重新取 j=0 输出 → j=1 break → 再退
# i=2: 同样行为, 外层共正常跑 3 轮, 输出 3 条
````

> **逐行解剖 + 常见错误**

> - `break` 只退当前层, 不影响外层
> - 常配合 `if` 使用, 满足条件才 break
> - 嵌套层 break 只退内层那层
> - 1. **以为 break 退所有循环**:只退一层, 外层继续
> - 2. **break 后同层代码永远不执行**

#### continue —— 跳过本次, 继续下一轮

> **痛点**:想跳过某些元素, 不终止整个循环。
> **类比**:点名时跳过请假的人, 其他人照样点名。
> **解释**:`continue` 跳过本轮循环体剩余代码,
>  直接进入下一轮判断, 循环本身不退出。


````python
# 只打印奇数, 偶数全部跳过
for i in range(1, 11):
    if i % 2 == 0:   # 偶数
        continue     # 跳过本轮
    print(i)

# --- 执行过程 ---
# i=1: 奇数 → 不 continue → 输出 1
# i=2: 偶数 → continue → 跳过 print
# i=3: 奇数 → 输出 3
# ...i=9 输出, i=10 跳过, 结束
````

> **break vs continue 对比**

下面同一个逻辑,分别用 break 和 continue 实现,
观察差异。


````python
# break vs continue 对比
print("break:", end=" ")
for i in range(1, 8):
    if i == 4:
        break         # i == 4 直接退出
    print(i, end=" ")
print()

print("continue:", end=" ")
for i in range(1, 8):
    if i == 4:
        continue      # i == 4 只跳过一轮
    print(i, end=" ")
print()

# break: 1 2 3  (到 4 就停)
# continue: 1 2 3 5 6 7  (跳过 4 继续)
````

> **逐行解剖 + 常见错误 + 口诀**

> - `continue` 配合 `if` 使用, 后面同层代码本轮跳过
> - 循环本身不退出, 下一轮照常判断
> - 1. **把 continue 当 break**:后续仍继续, 只跳当前
> - 2. **continue 后写了重要代码**:永远不执行
>
> **break vs continue 口诀**
>
> `break = "辞职" → 离开公司, 不再来`

#### 累加器与累乘器 —— 循环中累积结果

> **痛点**:求和、求乘积,单行 print 做不到。
> **类比**:存钱罐 —— 每轮投一枚硬币到 total,
>  不断累积就是最后总额。
> **解释**:用 **累加变量** 保存中间结果,
>  **加法初值 0, 乘法初值 1**。


````python
# 累加器: 求 1+2+...+10
total = 0               # 累加初值 = 0
for i in range(1, 11):
    total = total + i   # 每轮加到 total
print(f"1到10的和={total}")

# --- 执行过程 ---
# 初值 total = 0
# i=1: 0+1=1   i=2: 1+2=3   i=3: 3+3=6
# i=4: 6+4=10  ... i=10: 45+10=55
````

口诀:**加 0 乘 1**

累加初值必为 0, 累乘初值必为 1,
搞反结果就全错。


````python
# 累乘器: 求 1×2×3×4×5 (即 5 的阶乘)
product = 1                # 累乘初值 = 1
for i in range(1, 6):
    product = product * i  # 每轮乘到 product
print(f"5 的阶乘={product}")

# --- 执行过程 ---
# 初值 product = 1
# i=1: 1*1=1  i=2: 1*2=2  i=3: 2*3=6
# i=4: 6*4=24  i=5: 24*5=120
````

> **BREAK IT: 累乘初值写成 0,看看结果**

注意下面 product 初值写错为 0。


````python
# 故意写错: 累乘初值 = 0
product_wrong = 0
for i in range(1, 6):
    product_wrong = product_wrong * i
print(f"累乘初值=0,结果={product_wrong}")  # 必定是 0

# --- 执行过程 ---
# i=1: 0*1=0  i=2: 0*2=0
# ... 永远乘不出非零值!
````

> **逐行解剖 + 常见错误**

> - `total = 0` / `product = 1` —— 初值必须正确
> - `total += i` 是 `total = total + i` 的简写
> - 先想清楚初值再写循环
> - 1. **累乘初值写成 0**:0 乘任何数都为 0
> - 2. **累加初值写成 1**:结果多 1

#### 综合练习 —— 九九乘法表

> **目标**:用嵌套循环完成经典题目,
>  外层管行, 内层管列, 列数随行递增。


````python
# 9×9 乘法表
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j}*{i}={i*j}", end="	")
    print()                       # 每行结束换行

# --- 执行过程 (i=3 为例) ---
# i=3: j=1 → "1*3=3"; j=2 → "2*3=6"; j=3 → "3*3=9"
# 内层结束 print() 换行; 外层继续 i=4
````

```
每行比上一行多一式子 → 下三角
```

> - 外层 `range(1,10)` 共 9 行
> - 内层 `range(1,i+1)` 第 i 行有 i 个式子
> - `end="	"` 用制表符对齐, 不换行

#### while 与 for 的选择

> - **for**:已知次数, 用 range, 简单
> - **while**:次数不确定, 看条件, 灵活

```
┌──────────────┬────────────────────┐
│ 场景          │ 推荐                │
├──────────────┼────────────────────┤
│ 遍历 1-N      │ for i in range(1,N) │
│ 遍历字符串    │ for ch in s         │
│ 猜数字直到对  │ while True+break    │
│ 累加 / 累乘   │ for + 累加变量       │
│ 等用户输入    │ while 条件           │
└──────────────┴────────────────────┘
```


````python
# 同样逻辑的两种写法对比

# for 写法: 打印 1-5
print("for:", end=" ")
for i in range(1, 6):
    print(i, end=" ")
print()

# while 写法: 同样效果
print("while:", end=" ")
count = 1
while count <= 5:
    print(count, end=" ")
    count += 1
print()

# 首选 for 更简洁, while 适合条件复杂的场景
````

**今日小结**

| 概念 | 解决的痛点 |
|---|---|
| **while** 条件循环 | 不知次数, True 就继续, False 退出 |
| **for + range** 计数循环 | 已知次数, 比 while 简洁 |
| **嵌套循环** 行列结构 | 矩形、表格式数据 |
| **break** 立刻退出 | 找到目标立刻停, 不浪费时间 |
| **continue** 跳过本次 | 过滤特定元素, 不终止整个循环 |
| **累加/累乘器** 累积结果 | 求和求积, 加 0 乘 1 |

> 易错点:while 缺步进变死循环;
>  累乘初值写成 0 结果恒 0;
>  range 第二个参数不含, 倒序注意 stop。
