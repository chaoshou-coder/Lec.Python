**1. while 循环 —— 当条件成立就一直干**

类比:`while` = 跑步机,调到某个速度就一直跑,直到按下停止键。口诀:**while 三要素 —— 起始值、终止条件、递增步进,缺一不可**。忘记递增步进会变死循环!

```python
# 打印 1 到 5
count = 1
while count <= 5:       # 冒号 + 缩进
    print(count)
    count += 1          # 没有这行就是死循环!
print("循环结束")

# while True + break:主动退出
while True:
    word = input("输入 quit 退出:")
    if word == "quit":
        break           # 立刻跳出整个循环
print("你退出了")

# break 与 continue 的区别
# break:彻底结束循环(辞职)
for i in range(10):
    if i == 5:
        break           # 遇到 5 直接退出,后续不再执行
    print(i)            # 输出 0,1,2,3,4

# continue:跳过本次,继续下一趟(请一次假继续上班)
for i in range(10):
    if i % 2 == 0:
        continue        # 偶数跳过不打印
    print(i)            # 输出 1,3,5,7,9
```

**2. for 循环与 range —— 遍历**

`for item in 序列` 最适合"已知长度、要处理每一个元素"的场景。`range(start, stop, step)` 生成整数数列:**从 start 开始(含),到 
stop 结束(不含),每次跳 step**。


```python
# 遍历字符串
for ch in "Python":
    print(ch)           # P y t h o n 各一行

# 遍历列表
fruits = ["苹果", "香蕉", "橘子"]
for fruit in fruits:
    print(f"我喜欢吃 {fruit}")

# range 生成整数数列
for i in range(5):
    print(i)            # 0,1,2,3,4

for i in range(1, 6):
    print(i)            # 1,2,3,4,5

for i in range(0, 10, 2):
    print(i)            # 0,2,4,6,8

for i in range(100, 0, -2):
    print(i)            # 100,98,...,2
```

**3. 累加器与累乘器模式**

累加:初始化为 0,循环里 `total += i`。累乘:初始化为 1(不是 0!),循环里 `product *= i`。口诀:**加 0 乘 1,初始值不要搞反**。

```python
# 求 1 + 2 + ... + 100
total = 0                    # 1. 初始化为 0
for i in range(1, 101):
    total += i               # 2. 累加
print(total)                 # 5050

# 求奇数的乘积 1 * 3 * 5 * 7 * 9
product = 1                  # 1. 乘法的初始值为 1(不是 0!)
for i in range(1, 10, 2):
    product *= i
print(product)               # 945

# 实战:计算输入整数的位数
n = int(input("请输入一个整数:"))
count = 0
while n != 0:
    n //= 10                 # 整除 10,去掉最后一位
    count += 1
print(f"共 {count} 位")
```

**4. 嵌套循环 —— 九九乘法表**

循环里再套循环,外层控制行,内层控制列。`end=''` 可以取消 print 默认的末尾换行,让内层打印不换行。

```python
# 九九乘法表
for i in range(1, 10):               # 外层:行
    for j in range(1, i + 1):        # 内层:列(每行列数 = 行号)
        print(f"{j}×{i}={i*j}", end="\t")
    print()                          # 每行结束换行

# 鸡兔同笼:35 个头,94 只脚
for chicken in range(36):
    rabbit = 35 - chicken
    if chicken * 2 + rabbit * 4 == 94:
        print(f"鸡 {chicken} 只,兔 {rabbit} 只")
        break

# 字符串重复:* 运算符
print("*" * 20)     # ********************
print(" " * 4 + "***")  # 三角形第一行
```

**5. 综合实战 —— 菜单循环**

`while True + break` 模式是**菜单类程序的骨架**,几乎后续所有小项目都会复用。

```python
# 简易菜单
while True:
    print("\n=== 简易菜单 ===")
    print("1. 喂食  2. 玩耍  3. 睡觉  0. 退出")
    choice = input("请选择:")
    if choice == "1":
        print("喂食中...")
    elif choice == "2":
        print("玩耍中...")
    elif choice == "3":
        print("睡觉中...")
    elif choice == "0":
        print("再见!")
        break
    else:
        print("无效输入,请重新选择")
```

**✏️ 练一练 ⭐⭐⭐⭐**

输入整数直到 0 结束,统计奇数的均值与偶数的积(0 不参与统计)。如果没有奇数,均值输出 `无奇数`。

示例输入:`1 2 3 4 0` → 输出:`奇数均值: 2.0,偶数积: 8`

```python
# ======================
# 学员代码区
# ======================
```

```python
# 参考答案
odd_sum = 0
odd_count = 0
even_product = 1
even_count = 0
while True:
    n = int(input("请输入一个整数(0 结束):"))
    if n == 0:
        break
    if n % 2 == 1:
        odd_sum += n
        odd_count += 1
    else:
        even_product *= n
        even_count += 1
if odd_count > 0:
    print(f"奇数均值: {odd_sum / odd_count}")
else:
    print("奇数均值: 无奇数")
if even_count > 0:
    print(f"偶数积: {even_product}")
else:
    print("偶数积: 无偶数")
```

**今日小结**

本节让代码拥有"循环"能力。对比:

| 循环 | 适用场景 |
|---|---|
| `while` | 条件控制,不确定循环次数 |
| `for` | 遍历序列,已知循环次数 |
| `range(start, stop, step)` | 生成整数数列(左闭右开) |
| `break` | 彻底结束整个循环 |
| `continue` | 跳过本次,继续下一趟 |

> 🔴 **易错点**:
> 1. `while` 缺少递增步进会变死循环 —— 三要素:起始、终止、步进
> 2. 累乘初始值为 1,累加初始值为 0 —— 初值为 0 则乘积恒 0
> 3. `range(start, stop)` 左闭右开 —— `stop` 不会被取到
> 4. `break` 跳出整个循环,`continue` 只是跳过本次
