**1. while 循环 —— 当条件成立就一直干**

**while** 是最基础的循环结构: 先判断条件,条件为真就执行循环体,然后回头再判断。
口诀: 起始值、终止条件、递增步进,三者缺一不可!

```python
# 示例: 用 while 打印 1 到 5
count = 1               # 起始值
while count <= 5:       # 终止条件
    print(count)
    count += 1          # 递增步进(没有这行就是死循环!)
print("循环结束")
```

**2. break 与 continue —— 循环的"刹车"和"跳课"**

**break** = 立刻退出整个循环(像辞职)。**continue** = 跳过本次循环剩下的语句,
直接进入下一趟(像请假)。两者都只能用在循环内部。
生产中常用 **while True + break** 写菜单类程序骨架。

```python
# 示例: break —— 遇到 5 直接退出整个循环
print("break 示例:")
for i in range(10):
    if i == 5:
        break           # 遇到 5 立刻结束循环
    print(i, end=" ")
print()                 # 换行

# 示例: continue —— 偶数跳过,只打印奇数
print("continue 示例:")
for i in range(10):
    if i % 2 == 0:
        continue        # 偶数跳过,不执行后面的 print
    print(i, end=" ")
print()
```

**3. for 循环 + range —— 遍历序列**

**for item in 序列** 可以依次取出字符串、列表中的每个元素。
**range(start, stop, step)** 生成整数数列,起点含、终点不含(左闭右开)。

```python
# 遍历字符串: 逐个打印字符
for ch in "Python":
    print(ch)           # P y t h o n 各占一行

# 遍历列表
fruits = ["苹果", "香蕉", "橘子"]
for fruit in fruits:
    print("我喜欢吃 " + fruit)

# range() 的各种用法
print("range(5):")
for i in range(5):      # 0,1,2,3,4
    print(i, end=" ")
print()

print("range(1, 6):")
for i in range(1, 6):   # 1,2,3,4,5
    print(i, end=" ")
print()

print("range(0, 10, 2):")
for i in range(0, 10, 2):  # 0,2,4,6,8
    print(i, end=" ")
print()
```

**4. 累加器与累乘器**

累加器模式: 设 **total = 0**,循环里 **total += i**。
累乘器模式: 设 **product = 1**,循环里 **product *= i**。
口诀: **加 0 乘 1,初始值不要搞反**,否则乘积会永远为 0!

```python
# 累加器: 求 1 + 2 + ... + 100
total = 0                       # 累加初始值为 0
for i in range(1, 101):
    total += i                  # 逐个累加
print("1 到 100 的和为:", total)  # 5050

# 累乘器: 求 1 到 9 所有奇数的乘积
product = 1                     # 累乘初始值为 1(不是 0!)
for i in range(1, 10, 2):       # 1,3,5,7,9
    product *= i                # 逐个累乘
print("奇数的乘积为:", product)  # 945
```

关于初始值的小结: 累加器的 **total** 初始化为 0 是因为加 0 不影响结果;
累乘器的 **product** 初始化为 1 是因为乘 1 不影响结果。
如果不小心写成 0,无论乘以多少结果都还是 0,这是初学者最容易踩的坑!

**5. 嵌套循环 —— 九九乘法表**

外层循环控制行数,内层循环控制每行的列数。
内层 print 用 **end='\t'** 不换行,外层结束用 **print()** 换行。

```python
# 九九乘法表
for i in range(1, 10):              # 外层: 行
    for j in range(1, i + 1):       # 内层: 列(只打印左下三角)
        # end='\t' 让同一行用制表符隔开
        print(str(j) + "×" + str(i) + "=" + str(i * j), end="\t")
    print()                         # 每行结束换行
```

**6. 综合实战 —— 菜单循环与鸡兔同笼**

**while True + break** 是菜单类程序的经典骨架: 无限循环 + 特定条件退出。
鸡兔同笼问题用嵌套循环枚举所有可能,再用条件判断找到答案。

```python
# 简易菜单循环(while True + break)
while True:
    print("\n=== 简易菜单 ===")
    print("1. 喂食  2. 玩耍  3. 睡觉  0. 退出")
    choice = input("请选择:")

    if choice == "1":
        print("喂食中...")
    elif choice == "2":
        print("玩耍中...")
    elif choice == "3":
        print("睡觉中...")
    elif choice == "0":
        print("再见!")
        break               # 退出整个循环
    else:
        print("无效输入,请重新选择")
```

下面用**枚举法**解决鸡兔同笼问题: 假设鸡有 0~35 只,算出兔的数量,
再看总脚数是否匹配。这是一个经典的嵌套循环 + 条件判断的综合应用。

```python
# 鸡兔同笼: 35 个头,94 只脚,几只鸡几只兔?
total_heads = 35
total_feet = 94

# 枚举鸡的数量 0~35
for chicken in range(total_heads + 1):
    rabbit = total_heads - chicken
    # 判断脚的总数是否匹配
    if chicken * 2 + rabbit * 4 == total_feet:
        print("鸡 " + str(chicken) + " 只, 兔 " + str(rabbit) + " 只")
        break               # 找到答案,退出循环
```

**7. 易错点与今日小结**

今天学了 **while / for / range / break / continue / 累加累乘 / 嵌套循环**。
核心是理解: 循环让电脑帮我们重复干活,但一定要控制好入口和出口。

**本日易错 Top 4:**
1. **while 忘记加递增步进** → 条件永远成立,变成死循环
2. **累乘初值写成 0** → 无论乘多少结果永远是 0
3. **range(start, stop) 左闭右开** → stop 不会被取到
4. **break 与 continue 混淆** → break 是退出整个循环,continue 只是跳过本次

```python
# ✏️ 练习: 输入若干整数,输入 0 时退出,
# 统计奇数之和与偶数的个数
odd_sum = 0             # 奇数之和
even_count = 0          # 偶数个数

while True:
    n = int(input("请输入整数(0 结束):"))
    if n == 0:
        break           # 输入 0 退出循环
    if n % 2 == 1:
        odd_sum += n    # 奇数累加
    else:
        even_count += 1 # 偶数计数

print("奇数之和为:" + str(odd_sum))
print("偶数个数为:" + str(even_count))
```

参考答案: 上面用 while True + break 实现不确定次数的循环。
如果已知输入序列(比如列表),也可以用 for + break 改写,逻辑完全一致。

```python
# 参考答案: 用 for 循环重写上一练习
odd_sum = 0
even_count = 0
nums = [5, 2, 7, 4, 9, 6, 0]   # 模拟输入序列

for n in nums:
    if n == 0:
        break
    if n % 2 == 1:
        odd_sum += n
    else:
        even_count += 1

print("奇数之和为:" + str(odd_sum))
print("偶数个数为:" + str(even_count))
```
