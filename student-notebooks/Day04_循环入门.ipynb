## 第一讲(15 分钟)—— `while` 循环

### `while`:当条件成立就一直干

```python
while 条件:
    重复做的事
```

**口诀**: while = 当...时,就继续。先判断,再执行。

````python
# 示例代码:打印 1 到 5
count = 1
while count <= 5:       # 冒号 + 缩进
    print(count)
    count += 1          # 没有这行就是死循环!
print("循环结束")
````

> 🔴 **教学红线**: **while 三要素 —— 起始值、终止条件、递增步进,缺一不可!**

### 死循环与 `break`

````python
# 示例代码:用 break 退出循环
while True:
    word = input("输入 quit 退出:")
    if word == "quit":
        break           # 立刻跳出整个循环
print("你退出了")
````

### `break` 与 `continue` 的区别

- `break`: **彻底结束**循环(辞职)
- `continue`: **跳过本次**,继续下一趟(请假)

````python
# 示例代码:break vs continue
# break:遇到 5 直接退出
print("break 示例:")
for i in range(10):
    if i == 5:
        break           # 遇到 5 直接退出
    print(i, end=" ")  # 输出 0 1 2 3 4
print()

# continue:偶数跳过不打印
print("continue 示例:")
for i in range(10):
    if i % 2 == 0:
        continue        # 偶数跳过
    print(i, end=" ")  # 输出 1 3 5 7 9
print()
````

## 第二讲(15 分钟)—— `for` 循环与 `range`

### `for item in 序列`:遍历

````python
# 示例代码:遍历字符串
for ch in "Python":
    print(ch)           # P y t h o n 各一行

# 遍历列表
fruits = ["苹果", "香蕉", "橘子"]
for fruit in fruits:
    print(f"我喜欢吃 {fruit}")
````

### `range(start, stop, step)`:生成整数数列

**从 start 开始(含),到 stop 结束(不含),每次跳 step**

````python
# 示例代码:range 的各种用法
for i in range(5):
    print(i, end=" ")   # 0 1 2 3 4
print()

for i in range(1, 6):
    print(i, end=" ")   # 1 2 3 4 5
print()

for i in range(0, 10, 2):
    print(i, end=" ")   # 0 2 4 6 8
print()

for i in range(100, 0, -2):
    print(i, end=" ")   # 100 98 ... 2
print()
````

### 累加器模式

求 1 + 2 + ... + 100:

````python
# 示例代码:累加器
total = 0                    # 1. 初始化为 0
for i in range(1, 101):
    total += i               # 2. 累加
print(total)                 # 5050
````

### 累乘器:求奇数的乘积

````python
# 示例代码:累乘器
product = 1                  # ⚠️ 乘法的初始值为 1(不是 0!)
for i in range(1, 10, 2):
    product *= i
print(product)               # 945
````

> 🔴 **教学红线**: 累乘初值为 1! 如果写成 0,结果恒为 0。口诀: **加 0 乘 1,初始值不要搞反**。

## 第三讲(15 分钟)—— 综合实战

### 鸡兔同笼:循环 + 条件合作

````python
# 示例代码:鸡兔同笼
# 35 个头,94 只脚,几只鸡几只兔?
for chicken in range(36):
    rabbit = 35 - chicken
    if chicken * 2 + rabbit * 4 == 94:
        print(f"鸡 {chicken} 只,兔 {rabbit} 只")
        break
````

### 菜单循环:死循环 + 分支退出

这种 `while True + break` 模式是**菜单类程序的骨架**。

````python
# 示例代码:简易菜单
while True:
    print("\n=== 简易菜单 ===")
    print("1. 喂食  2. 玩耍  3. 睡觉  0. 退出")
    choice = input("请选择:")

    if choice == "1":
        print("喂食中...")
    elif choice == "2":
        print("玩耍中...")
    elif choice == "3":
        print("睡觉中...")
    elif choice == "0":
        print("再见!")
        break
    else:
        print("无效输入,请重新选择")
````

### 字符串重复:`*` 运算符

"*" * 20 = "********************"

````python
# 示例代码:用 * 打印图案
print("*" * 20)      # ********************
print(" " * 4 + "***")  # 三角形第一行
````

**题目**: 输入整数直到 0 结束,统计奇数均值与偶数乘积。

要求:
- 0 不参与统计
- 没有奇数时输出"没有奇数"

示例:
```
请输入数字(0结束): 5
请输入数字(0结束): 2
请输入数字(0结束): 3
请输入数字(0结束): 0
奇数均值: 4.0
偶数乘积: 2
```

````python
# ======================
# 学员代码区
# ======================
pass
````

<details>
<summary>💡 点击查看参考答案</summary>

````python
# 参考答案
odd_sum = 0
odd_count = 0
even_product = 1
has_even = False

while True:
    n = int(input("请输入数字(0结束):"))
    if n == 0:
        break
    if n % 2 == 1:
        odd_sum += n
        odd_count += 1
    else:
        even_product *= n
        has_even = True

if odd_count > 0:
    print(f"奇数均值: {odd_sum / odd_count}")
else:
    print("没有奇数")

if has_even:
    print(f"偶数乘积: {even_product}")
else:
    print("没有偶数")
````

## 综合练习:打印等腰三角形

### ✏️ 练习 6(小项目) ⭐⭐⭐⭐

**题目**: 输入行数 n,打印由 `*` 组成的等腰三角形。

示例(n=5):
```
    *
   ***
  *****
 *******
*********
```

**规律**: 第 i 行(从 0 开始) = `(n - i - 1)` 个空格 + `(2 * i + 1)` 个星号

提示: 
- i=0: 4 个空格 + 1 个星
- i=1: 3 个空格 + 3 个星
- i=2: 2 个空格 + 5 个星

````python
# ======================
# 学员代码区
# ======================
pass
````

<details>
<summary>💡 点击查看参考答案</summary>

````python
# 参考答案
n = int(input("请输入行数:"))

for i in range(n):
    spaces = n - i - 1
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)
````

</details>
