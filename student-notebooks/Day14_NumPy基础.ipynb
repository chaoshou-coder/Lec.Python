#### np.array 创建数组

[痛点] Python 列表做百万次数值运算慢,因为每个元素存类型信息,逐个解包。

[解释] `np.array()` 从列表创建数组,一维列表 → 向量,嵌套列表 → 矩阵。所有元素共享同一类型(dtype),底层 C 实现,省掉类型检查开销。

#### dtype / shape / ndim 数组身份证

[痛点] 不知道数组里存的是什么类型、什么形状,导致后续运算报错。
[类比] dtype 是"货物种类",shape 是"集装箱尺寸",ndim 是"堆叠层数"。
[解释] `dtype` 数据类型(所有元素共享),`shape` 形状(元组表示各维度长度),`ndim` 维度数。混合类型会被强制统一为精度更高的类型。

```python
import numpy as np

# 创建整数数组,默认 dtype 是 int64
a = np.array([1, 2, 3])
print(a.dtype)        # int64,自动推断
print(a.shape)        # (3,),一维,长度 3(逗号不能少,是元组)
print(a.ndim)         # 1,一维数组

# 混合 int + float 自动提升为 float64
b = np.array([1, 2.5, 3])
print(b.dtype)        # float64,混合时统一为精度更高的类型
```

② 第5行: 打印 a.dtype → int64

④ 第7行: 打印 a.ndim → 1

⑤ 第10行: 创建混合数组 b,含整数和浮点数

⑥ 第11行: 打印 b.dtype → float64,自动提升

**逐行解剖**

- `dtype`:data type,数组中所有元素共享同一类型
- `shape`:元组形式,`(3,)` 表示一维长度 3,`(2, 3)` 表示 2 行 3 列
- `ndim`:number of dimensions,维度数,等于 len(shape)
- 混合 int + float → float64;混合数字和字符串 → 全部变字符串

#### 常用创建函数 zeros / ones / arange / linspace

[痛点] 需要快速创建全零矩阵、等差数列等,手写循环太繁琐。
[类比] zeros 是"空货架",ones 是"满货架",arange 是"等差刻度尺",linspace 是"等分绳子"。
[解释] `np.zeros((m,n))` 全零,`np.ones((m,n))` 全一,`np.arange(start,stop,step)` 等差数列(不含尾),`np.linspace(start,stop,num)` 等分点(含尾)。

```python
import numpy as np

# np.zeros:全零矩阵,形状参数必须是元组 (2, 3)
z = np.zeros((2, 3))
print(z)

# np.ones:全一矩阵
o = np.ones((3, 2))
print(o)

# np.arange:等差数列,不含尾
a = np.arange(0, 10, 2)
print(a)              # [0 2 4 6 8]

# np.linspace:等分点,含尾
x = np.linspace(0, 1, 5)
print(x)              # [0.   0.25 0.5  0.75 1.  ]
```

#### reshape 改变形状

[痛点] 数组形状不匹配,需要重新塑形,但不想复制数据。
[类比] reshape 像"重新排列积木"——总数不变,摆放方式变了。
[解释] `reshape(m, n)` 把数组重新塑形,返回**视图**(共享内存)。`-1` 表示这一维自动计算。修改视图会影响原数组,需要独立副本用 `.copy()`。

```python
import numpy as np

# 0~11 共 12 个元素,reshape 为 3 行 4 列
a = np.arange(12)
b = a.reshape(3, 4)
print(b)

# -1 表示自动计算:12 个元素 / 2 行 = 6 列
c = a.reshape(2, -1)
print(c.shape)        # (2, 6)

# 验证视图共享内存:修改 b 后 a 跟着变
b[0, 0] = 999
print(a[0])           # 999 —— a 也被改了
```

# --- 执行过程 ---

① 第2行: 创建 0~11 的一维数组 a

② 第3行: `a.reshape(3, 4)` 把 a 重塑为 3 行 4 列,赋值给 b

③ 第5行: `a.reshape(2, -1)` 2 行,列数自动计算为 6

④ 第8行: 修改 b[0,0] 为 999

⑤ 第9行: 打印 a[0],发现也变成 999,说明共享内存

**逐行解剖**

- `reshape(3, 4)`:3 行 4 列,元素总数必须匹配
- `reshape(2, -1)`:-1 让 NumPy 自动计算这一维
- 视图共享内存:修改 b 会影响 a,需要独立副本用 `.copy()`

#### 向量化运算 告别 for 循环

[痛点] Python 列表做百万次运算慢,必须手写 for 或列表推导。
[类比] 向量化像"批量印刷"——一次印整张纸,而不是一个字一个字写。
[解释] 对整个数组逐元素运算,底层 C 循环一次性处理,不需要手写 for。写法更简洁,速度快 10-100 倍。

```python
import numpy as np

# 对整个数组逐元素平方再加 1,无需 range 或 for
a = np.arange(5)      # [0 1 2 3 4]
result = a ** 2 + 1   # 逐元素: [(0^2+1), (1^2+1), ...]
print(result)         # [ 1  2  5 10 17]

# 对比 Python 列表写法(慢)
lst = [0, 1, 2, 3, 4]
result_list = [x ** 2 + 1 for x in lst]
print(result_list)    # [1, 2, 5, 10, 17]
```

# --- 执行过程 ---

① 第2行: `np.arange(5)` 创建数组 [0 1 2 3 4]

② 第3行: `a ** 2 + 1` 逐元素平方再加 1

③ 第4行: 打印结果 [1 2 5 10 17]

④ 第7行: Python 列表写法,需要列表推导

⑤ 第8行: 结果相同,但写法更繁琐

**逐行解剖**

- `a ** 2`:逐元素平方,不是矩阵乘法
- `a ** 2 + 1`:向量化运算,每个元素独立计算
- 对比列表推导:NumPy 写法更简洁,速度更快

#### 广播 broadcasting

[痛点] 两个形状不同的数组想做运算,只能手动循环补齐,代码冗长。
[类比] 广播像"复印机"——小数组被自动复制放大,补满到大数组的形状。
[解释] 从最后一个维度开始对齐,要么相等,要么其中一方是 1,要么一方缺失。例如 shape `(2,3)` 的矩阵加上 shape `(3,)` 的向量,向量会被广播到每一行。

```python
import numpy as np

# 矩阵 shape (2, 3)
mat = np.array([[1, 2, 3],
                [4, 5, 6]])
# 向量 shape (3,),会被广播到每一行
vec = np.array([10, 20, 30])
print(mat + vec)      # [[11 22 33], [14 25 36]]

# 标量与矩阵广播:标量被拉伸成同样形状后相加
print(np.array([[1, 2], [3, 4]]) + 18)  # [[19 20], [21 22]]
```

# --- 执行过程 ---

① 第2行: 创建 shape (2,3) 的矩阵 mat

② 第5行: 创建 shape (3,) 的向量 vec

② 第6行: `mat + vec`,vec 被广播到每一行,每行都加 [10,20,30]

③ 第9行: 标量 18 与 2x2 矩阵相加,标量被广播

**逐行解剖**

- 广播规则:从最后一个维度对齐,要么相等要么一方是 1 或缺失
- `(2,3)` + `(3,)` → 向量被复制到每一行
- 标量与数组运算:标量被拉伸成同样形状

#### 常用数学函数 sqrt / exp / log

[痛点] 需要对数组每个元素做数学运算,手写循环太慢。
[类比] 数学函数像"批量计算器"——一次处理整个数组。
[解释] NumPy 提供逐元素数学函数:`np.sqrt` 平方根,`np.exp` 自然指数,`np.log` 自然对数。都是向量化运算,速度快。

```python
import numpy as np

# 准备一个整数数组,函数返回 float 数组
x = np.array([1, 4, 9, 16, 25])

# np.sqrt:逐元素开平方
print(np.sqrt(x))     # [1. 2. 3. 4. 5.]

# np.exp:逐元素计算 e 的 x 次方
print(np.exp(np.array([0.0, 1.0])))  # [1.        2.71828183]

# np.log:逐元素计算自然对数(底为 e)
print(np.log(np.array([1.0, np.e])))  # [0. 1.]
```

# --- 执行过程 ---

① 第2行: 创建数组 [1 4 9 16 25]

② 第5行: `np.sqrt(x)` 逐元素开平方

② 第8行: `np.exp([0,1])` 计算 e^0=1, e^1=e

③ 第11行: `np.log([1,e])` 计算 ln(1)=0, ln(e)=1

**逐行解剖**

- `np.sqrt(x)`:逐元素开平方,返回 float 数组
- `np.exp(x)`:e 的 x 次方,e ≈ 2.71828
- `np.log(x)`:自然对数,底为 e

#### 聚合函数 sum / mean

[痛点] 需要求和、求平均,手写 for 循环累加太繁琐。
[类比] 聚合函数像"压缩机"——把整个数组压成一个标量。
[解释] `np.sum` 求和,`np.mean` 求平均。不指定 axis 时对整个数组计算,返回标量。指定 axis 后沿该维度压缩(详见 Day12)。

```python
import numpy as np

x = np.arange(1, 6)   # [1 2 3 4 5]

# np.sum 对所有元素求和
print(np.sum(x))      # 15

# np.mean 求算术平均
print(np.mean(x))     # 3.0,返回浮点数
```

# --- 执行过程 ---

① 第1行: 创建数组 [1 2 3 4 5]

② 第4行: `np.sum(x)` 求和 = 1+2+3+4+5 = 15

③ 第7行: `np.mean(x)` 求平均 = 15/5 = 3.0

**逐行解剖**

- `np.sum(x)`:所有元素求和,返回标量
- `np.mean(x)`:算术平均,返回浮点数
- 两种写法等价:`np.sum(x)` 和 `x.sum()`
