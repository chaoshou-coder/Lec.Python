**从列表创建数组 —— np.array**

`np.array()` 是 NumPy 最常用的创建方式,接受 Python 列表或嵌套列表,返回 `ndarray` 对象。一维列表生成一维数组,二维列表生成二维数组。

```python
# 每个 cell 独立 import,这样单击 shift+enter 即可单独运行
import numpy as np

# 从一维列表创建向量(一维数组)
a = np.array([1, 2, 3, 4, 5])
print(a)              # 输出紧凑格式: [1 2 3 4 5]
print(type(a))        # <class 'numpy.ndarray'>,NumPy 的核心数据结构
```

**数组身份证 —— dtype / shape / ndim**

`dtype` 是数据类型,**所有元素共享同一类型**,混合类型会被**强制统一**为精度更高的类型。`shape` 是形状,用**元组**表示各维度长度。`ndim` 是维度数(秩)。

```python
import numpy as np

# 创建整数数组,默认 dtype 是 int64
a = np.array([1, 2, 3])
print(a.dtype)        # int64,自动推断
print(a.shape)        # (3,),一维,长度 3(逗号不能少,是元组)
print(a.ndim)         # 1,一维数组

# 混合 int + float 自动提升为 float64
b = np.array([1, 2.5, 3])
print(b.dtype)        # float64,混合时统一为精度更高的类型
```

**创建二维数组(矩阵)**

嵌套列表 → 二维数组。外层列表的每个元素是一行,内层列表长度必须一致,否则变成“参差不齐”的 object 数组,性能大幅退化。

```python
import numpy as np

# 2 行 3 列矩阵 —— 外层 2 个元素,内层每个 3 个元素
m = np.array([[1, 2, 3],
              [4, 5, 6]])
print(m)
print(m.shape)        # (2, 3) → 2 行 3 列
print(m.ndim)         # 2,二维
```

**全零数组 —— np.zeros**

`np.zeros(shape)` 创建**全 0** 数组,元素为 float64。形状参数必须传**元组 `(m, n)`**,不能写成 `np.zeros(2, 3)` —— 那是传两个独立参数,会 TypeError。

```python
import numpy as np

# 创建 3 行 4 列全 0 矩阵,形状参数必须是元组 (3, 4)
z = np.zeros((3, 4))
print(z)
print(z.dtype)        # float64,默认浮点
```

**全一数组 —— np.ones**

`np.ones(shape)` 创建**全 1** 数组,用法与 `np.zeros` 完全一致,同样需要元组作为形状参数。

```python
import numpy as np

# 创建 2 行 3 列全 1 矩阵
o = np.ones((2, 3))
print(o)
print(o.dtype)        # float64
```

**等差数列 —— np.arange**

`np.arange([start], stop, [step])` 类似 Python 内置 `range()`,但返回 **ndarray** 且**支持小数步长**。左闭右开 `[start, stop)`,不包含终点。

```python
import numpy as np

# 从 0 开始,步长 2,到 10 之前停止(不含 10)
a = np.arange(0, 10, 2)
print(a)              # [0 2 4 6 8]

# 省略 start 和 step,默认从 0 开始、步长 1
b = np.arange(5)
print(b)              # [0 1 2 3 4]
```

**等间隔分割 —— np.linspace**

`np.linspace(start, stop, num)` 在起止区间内生成 **num 个等分点**,**包含终点**(与 arange 不同)。适合画函数图像时构造横坐标。

```python
import numpy as np

# 在 0~1 区间等分 5 个点,包含终点 1.0
x = np.linspace(0, 1, 5)
print(x)              # [0.   0.25 0.5  0.75 1.  ]
```

**向量化运算 —— 告别 for 循环**

**向量化** = 对整个数组逐元素运算,底层 C 循环一次性处理,**不需要手写 for**。对比 Python 列表必须循环或列表推导,NumPy 写法更简洁、速度快 10-100 倍。

```python
import numpy as np

# 对整个数组逐元素平方再加 1,无需 range 或 for
a = np.arange(5)      # [0 1 2 3 4]
result = a ** 2 + 1   # 逐元素: [(0^2+1), (1^2+1), ...]
print(result)         # [ 1  2  5 10 17]
```

**广播 broadcasting —— 小数组“拉伸”满大数组**

**广播**规则:从最后一个维度开始对齐,要么相等,要么其中一方是 1,要么一方缺失。例如 shape `(2, 3)` 的矩阵 加上 shape `(3,)` 的向量,向量会被“广播”到每一行。

```python
import numpy as np

# 矩阵 shape (2, 3)
mat = np.array([[1, 2, 3],
                [4, 5, 6]])
# 向量 shape (3,),会被广播到每一行
vec = np.array([10, 20, 30])
print(mat + vec)      # [[11 22 33], [14 25 36]]

# 标量 18 与 2x2 矩阵广播:标量被拉伸成同样形状后相加
print(np.array([[1, 2], [3, 4]]) + 18)
```

**常用数学函数 —— sqrt / exp / log**

NumPy 提供了大量逐元素数学函数,命名与数学公式一一对应。`np.sqrt` 平方根 / `np.exp` 自然指数 / `np.log` 自然对数。

```python
import numpy as np

# 准备一个整数数组,函数返回 float 数组
x = np.array([1, 4, 9, 16, 25])

# np.sqrt:逐元素开平方
print(np.sqrt(x))     # [1. 2. 3. 4. 5.]

# np.exp:逐元素计算 e 的 x 次方
print(np.exp(np.array([0.0, 1.0])))

# np.log:逐元素计算自然对数(底为 e)
print(np.log(np.array([1.0, np.e])))
```

**聚合函数 —— sum 与 mean**

**聚合函数**把数组“压缩”成标量。`np.sum` 求和 / `np.mean` 求平均。不指定 `axis` 时对整个数组计算,返回标量;指定 `axis` 后沿该维度压缩,详见 Day12。

```python
import numpy as np

x = np.arange(1, 6)   # [1 2 3 4 5]

# np.sum 对所有元素求和
print(np.sum(x))      # 15

# np.mean 求算术平均
print(np.mean(x))     # 3.0,返回浮点数
```

**reshape 改变形状 —— 视图 vs 副本**

`reshape(m, n)` 把数组重新塑形,**不复制数据**,返回**视图** —— 修改视图会影响原数组。需要独立副本用 `.copy()`,否则可能踪到“改了 b,a 也变了”的坑。`-1` 表示这一维**自动计算**。

```python
import numpy as np

# 0~11 共 12 个元素,reshape 为 3 行 4 列
a = np.arange(12)
b = a.reshape(3, 4)
print(b)

# -1 表示自动计算:12 个元素 / 2 行 = 6 列
c = a.reshape(2, -1)
print(c.shape)        # (2, 6)

# 验证视图共享内存:修改 b 后 a 跟着变
b[0, 0] = 999
print(a[0])           # 999 —— a 也被改了

# 需要独立副本:d = a.reshape(3, 4).copy(),修改 d 不影响 a
```

**今日小结**

- **np.zeros / np.ones** 全零 / 全一,形状参数必须是元组

- **np.sqrt / np.exp / np.log** 逐元素数学函数
- **np.sum / np.mean** 聚合为标量
- **reshape** 返回视图(共享内存),需要副本用 **.copy()**
