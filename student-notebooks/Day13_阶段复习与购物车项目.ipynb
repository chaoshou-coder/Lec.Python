### 高频错误 1:if/elif 边界错误

错误写法:`elif score >= 80` 写在 `elif score >= 60` 之后,>= 60 先命中,>= 80 永远执行不到。正确做法:**从高分往低分写**,或者用 `and` 显式区间。

````python
# ===== 典型的 if/elif 边界错误 =====
score = 85

# 错误写法:低分条件在前,高分永远进不来
if score >= 60:
    print("及格")          # 命中,就停了
elif score >= 80:
    print("良好")          # 进不来
elif score >= 90:
    print("优秀")          # 进不来

# 正确写法:从高到低
if score >= 90:
    print("优秀")
elif score >= 80:
    print("良好")          # 命中
elif score >= 60:
    print("及格")
else:
    print("不及格")
````

### 高频错误 2:累加器在循环内重新初始化

错误写法:`total = 0` 写在循环里,每次迭代都被清零,结果只加到最后一个值。正确做法:累加器在循环外初始化,循环内只累加。

````python
# ===== 累加器位置错误 =====
prices = [5.5, 12.8, 8.0]

# 错误写法:total 在循环内被反复清零
total_wrong = 0
for p in prices:
    total_wrong = 0       # 每次都被清零!
    total_wrong += p
print("错误结果:", total_wrong)    # 8.0

# 正确写法:total 在循环外
total_right = 0
for p in prices:
    total_right += p     # 只累加,不清零
print("正确结果:", total_right)    # 26.3
````

### 高频错误 3:return 与 print 混淆

错误写法:函数里 `print` 了却没有 `return`,调用者拿到的永远是 `None`。正确做法:用 `return` 把值交出去,print 只负责给人看。

````python
# ===== return vs print + input() 返回类型 =====
def calc_print(nums):
    print(sum(nums))       # 只是打印,没有 return → 返回 None

def calc_return(nums):
    return sum(nums)       # 把结果交给调用者

r1 = calc_print([1, 2, 3])
r2 = calc_return([1, 2, 3])
print("print 版返回:", r1, type(r1))      # None <class 'NoneType'>
print("return 版返回:", r2, type(r2))     # 6 <class 'int'>

# input() 返回的是字符串,要转 int
# age = input("年龄:")    # 输入 20 → 字符串 "20"
# age = int(input("年龄:"))   # 转成整数 20
print("输入示例:age = int(input('年龄:'))")
````

### 读错方法论

遇到报错不要慌张,按以下顺序读错误信息:

1. 看**最后一行**异常类型(ValueError / TypeError / KeyError)
2. 看报错**行号**,定位出错位置
3. 看是哪一行的代码触发的
4. 遇到 `Traceback` 时,从下往上读,最后一行是直接原因

**口诀**:类型 → 行号 → 上下文 → 猜测 → 验证

### find_product() 与 browse_products() 辅助函数

find_product():按 id 在商品库查找商品,找到返回字典,找不到返回 None。browse_products():展示所有商品,用 for 循环遍历。

````python
# ===== find_product + browse_products =====
products = [
    {"id": 1, "name": "苹果", "price": 5.50},
    {"id": 2, "name": "香蕉", "price": 3.20},
    {"id": 3, "name": "牛奶", "price": 8.00},
    {"id": 4, "name": "面包", "price": 6.50},
]

def find_product(pid):
    for p in products:
        if p["id"] == pid:
            return p
    return None

def browse_products():
    print("\n--- 商品列表 ---")
    for p in products:
        print(f"  {p['id']}. {p['name']}  ￥{p['price']:.2f}")

browse_products()
print(find_product(2))        # 香蕉字典
print(find_product(99))       # None
````

** ---- 执行过程 ----**

以 `find_product(2)` 为例:

1. 进入 for 循环,p = {"id": 1, ...}
2. `p["id"] == 2` → False,继续
3. p = {"id": 2, ...},相等,**返回 p**
4. 如果循环结束都没命中,返回 None

````python
# ===== add_to_cart() 加入购物车 =====
products = [
    {"id": 1, "name": "苹果", "price": 5.50},
    {"id": 2, "name": "香蕉", "price": 3.20},
    {"id": 3, "name": "牛奶", "price": 8.00},
    {"id": 4, "name": "面包", "price": 6.50},
]
cart = []

def add_to_cart():
    global cart
    try:
        raw = input("输入商品 id:")
        pid = int(raw)        # 可能抛 ValueError
    except ValueError:
        print("请输入数字!")
        return
    p = find_product(pid)
    if p is None:
        print("无此商品")
        return
    # 已在购物车 → 数量 +1
    for item in cart:
        if item["id"] == pid:
            item["qty"] += 1
            print(f"已增加 {p['name']} 数量")
            return
    # 不在购物车 → 新增
    cart.append({"id": p["id"], "name": p["name"],
                 "price": p["price"], "qty": 1})
    print(f"已加入 {p['name']}")

# 由于需要真实 input,这里用直接调用模拟
# add_to_cart()
print("add_to_cart() 函数定义完成")
````

### view_cart() 与 checkout() 查看与结算

view_cart():遍历购物车打印每条小计与总价,**注意累加器 total 在循环外初始化**。checkout():计算总价,清掉购物车 `cart.clear()`。

````python
# ===== view_cart + checkout =====
cart = [
    {"id": 1, "name": "苹果", "price": 5.50, "qty": 2},
    {"id": 3, "name": "牛奶", "price": 8.00, "qty": 1},
]

def view_cart():
    if not cart:
        print("购物车为空")
        return
    print("\n--- 购物车 ---")
    total = 0                 # 累加器在循环外!
    for item in cart:
        subtotal = item["price"] * item["qty"]
        total += subtotal
        print(f"  {item['name']} x{item['qty']} = ￥{subtotal:.2f}")
    print(f"总价: ￥{total:.2f}")

def checkout():
    global cart
    if not cart:
        print("购物车为空")
        return
    view_cart()
    cart.clear()              # 清空购物车
    print("已结算,购物车已清空")

view_cart()
checkout()
````

### JSON 持久化:save_cart() 与 load_cart()

程序退出后数据不能丢,用 `json.dump()` 写入文件,`json.load()` 读取文件。`encoding="utf-8"` 必须写,否则中文会乱码。`try/except` 处理文件不存在的情况。

````python
# ===== save_cart + load_cart =====
import json, os

cart = [{"id": 1, "name": "苹果", "price": 5.5, "qty": 2},
        {"id": 3, "name": "牛奶", "price": 8.0, "qty": 1}]

def save_cart(path="_cart.json"):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(cart, f, ensure_ascii=False, indent=2)
    print("购物车已保存")

def load_cart(path="_cart.json"):
    global cart
    if not os.path.exists(path):
        print("无存档,重新开始")
        cart = []
        return
    try:
        with open(path, "r", encoding="utf-8") as f:
            cart = json.load(f)
        print("读取存档成功")
    except json.JSONDecodeError:
        print("存档损坏,清空")
        cart = []

save_cart()
cart = []                     # 模拟重启
load_cart()
print(cart)                   # 恢复回来

# 清理临时文件
if os.path.exists("_cart.json"):
    os.remove("_cart.json")
````

### 完整整合:启动加载 + 运行时保存

把 load / save 串入购物车生命周期:启动时 load,每次操作后 save。即使程序异常退出数据也不丢失。

````python
# ===== 完整整合 Demo =====
import json, os

products = [
    {"id": 1, "name": "苹果", "price": 5.50},
    {"id": 2, "name": "香蕉", "price": 3.20},
]
cart = []
DATA = "_cart.json"

def load():
    global cart
    if os.path.exists(DATA):
        with open(DATA, "r", encoding="utf-8") as f:
            cart = json.load(f)

def save():
    with open(DATA, "w", encoding="utf-8") as f:
        json.dump(cart, f, ensure_ascii=False, indent=2)

def browse():
    for p in products:
        print(f"{p['id']}. {p['name']}  ￥{p['price']:.2f}")

def add():
    global cart
    try:
        pid = int(input("id:"))
    except ValueError:
        print("请输入数字")
        return
    p = next((x for x in products if x["id"] == pid), None)
    if not p:
        print("无此商品")
        return
    for item in cart:
        if item["id"] == pid:
            item["qty"] += 1
            break
    else:
        cart.append({"id": p["id"], "name": p["name"],
                     "price": p["price"], "qty": 1})
    save()

# 由于需要真实 input,这里只演示 browse + save
browse()
save()                        # 保存空购物车演示一下
print("--- 查看保存内容 ---")
if os.path.exists(DATA):
    with open(DATA, "r", encoding="utf-8") as f:
        print(f.read())
if os.path.exists(DATA):
    os.remove(DATA)
````

** ---- 执行过程 ----**

整合版生命周期:

1. 启动 → `load()`:读出上次的 cart
2. 用户操作 → `browse / add / view / checkout`
3. 每次操作后 → `save()`:写回文件
4. 崩溃 → 启动时 `load()` 恢复到最后一次 save,不丢数据

**逐行解剖** — 下面演示 `save()` 被连续调用三次,文件系统始终保存最新状态:

````python
import json, os
path = "_demo.json"

# 第一次 save
data = [1, 2, 3]
with open(path, "w", encoding="utf-8") as f:
    json.dump(data, f)

# 第二次 save
data = [1, 2, 3, 4]
with open(path, "w", encoding="utf-8") as f:
    json.dump(data, f)

# 读:拿到最新一次 save 的内容
with open(path, "r", encoding="utf-8") as f:
    print(json.load(f))      # [1, 2, 3, 4]

if os.path.exists(path):
    os.remove(path)
````

### 购物车 v2:OOP 版 Product 与 Cart 类

把商品抽象成类,每个商品是一个对象。类构造函数 `__init__` 初始化属性,`__str__` 定义打印格式。把 Cart 也用类封装,/cart.clear() 变成方法。

````python
# ===== Product 类 + Cart 类 =====
class Product:
    def __init__(self, pid, name, price):
        self.pid = pid
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.pid}. {self.name}  ￥{self.price:.2f}"

products = [
    Product(1, "苹果", 5.50),
    Product(2, "香蕉", 3.20),
]
for p in products:
    print(p)                  # 利用 __str__ 直接打印

class Cart:
    def __init__(self):
        self.items = []       # 每项:{"product": Product, "qty": 1}

    def add(self, product, qty=1):
        for item in self.items:
            if item["product"].pid == product.pid:
                item["qty"] += qty
                return
        self.items.append({"product": product, "qty": qty})

    def total(self):
        # 累加器在循环外!
        t = 0
        for item in self.items:
            t += item["product"].price * item["qty"]
        return t

    def clear(self):
        self.items.clear()

cart = Cart()
cart.add(products[0])
cart.add(products[0])         # 数量 +1
cart.add(products[1])
print(f"总价: ￥{cart.total():.2f}")
````

### OOP 整合 Demo

把 Product 和 Cart 串入 main 循环,实现面向对象版购物车。对比函数式版本,OOP 把数据和行为绑定,扩展(如打折、会员)更容易。

````python
# ===== OOP 整合运行 =====
class Product:
    def __init__(self, pid, name, price):
        self.pid = pid
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.pid}. {self.name}  ￥{self.price:.2f}"

class Cart:
    def __init__(self):
        self.items = []

    def add(self, product, qty=1):
        for item in self.items:
            if item["product"].pid == product.pid:
                item["qty"] += qty
                return
        self.items.append({"product": product, "qty": qty})

    def view(self):
        if not self.items:
            print("购物车为空")
            return
        total = 0
        for item in self.items:
            sub = item["product"].price * item["qty"]
            total += sub
            print(f"  {item['product'].name} x{item['qty']} = ￥{sub:.2f}")
        print(f"总价: ￥{total:.2f}")

    def checkout(self):
        self.view()
        self.items.clear()

products = [Product(1, "苹果", 5.50), Product(2, "香蕉", 3.20)]
cart = Cart()
# 模拟几次加入操作
cart.add(products[0])
cart.add(products[0])
cart.add(products[1])
cart.view()
cart.checkout()
cart.view()                   # 已清空
````

### 练习 5:综合函数设计 -- view_cart 返回格式化字符串

请实现 `view_cart(cart)` 函数,**不直接 print**,而是**返回**一个格式化的字符串:

```
--- 购物车 ---
  苹果 x2 = ￥11.00
  牛奶 x1 = ￥8.00
总价: ￥19.00
```

空购物车时返回 '购物车为空'。

**要点**:
- 累加器在循环外初始化
- 用 `return` 把结果交出去,不要 print
- 用 `\n` 拼接多行字符串

> 问自己:为什么用 return 比直接 print 更好?

````python
# ==============================
# 学员代码区:view_cart 返回字符串
# ==============================
pass
````

````python
# 参考答案
def view_cart(cart):
    if not cart:
        return "购物车为空"
    lines = ["--- 购物车 ---"]
    total = 0                       # 累加器在循环外
    for item in cart:
        subtotal = item["price"] * item["qty"]
        total += subtotal
        lines.append(f"  {item['name']} x{item['qty']} = ￥{subtotal:.2f}")
    lines.append(f"总价: ￥{total:.2f}")
    return "\n".join(lines)          # 用换行符拼接多行

cart = [{"id": 1, "name": "苹果", "price": 5.5, "qty": 2},
        {"id": 3, "name": "牛奶", "price": 8.0, "qty": 1}]
result = view_cart(cart)            # 拿到字符串
print(result)                       # 由调用者决定打印
print(view_cart([]))                # 购物车为空
````
