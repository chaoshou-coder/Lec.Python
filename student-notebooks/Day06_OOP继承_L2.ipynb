#### 为什么需要继承(代码复用 / is-a 关系)

> **痛点**  
写几个类发现彼此有大量重复代码,比如实体书和电子书几乎完全一样,
复制粘贴越多,改 BUG 越痛。

> **类比**  
动物 → 猫 / 狗:猫 'is-a' 动物,继承了动物的共同特征,再扩展自己的'叫'。

> **解释**  
继承让子类**自动拥有**父类的属性和方法,在此基础上添加或修改,实现代码复用。

````python
# 父类(基类)
class Animal:
    def __init__(self, name):
        self.name = name

    def breathe(self):
        print(f"{self.name} 在呼吸")

# 子类:自动继承 Animal 的所有内容
class Dog(Animal):
    def bark(self):  # 子类扩展自己的方法
        print(f"{self.name} 汪汪叫")

dog = Dog("旺财")
dog.breathe()   # 继承自 Animal
dog.bark()      # 子类自己的方法
````

> **逐行解剖**  
① `class Animal:` 创建父类  
② `def __init__(self, name)` 绑定 name  
③ `class Dog(Animal):` Dog 继承 Animal 的所有成员  
④ `dog = Dog('旺财')` → Dog 没有 __init__,沿'子类→父类'找到 Animal.__init__  
⑤ `dog.breathe()` → Dog 没有 breathe,找到 Animal.breathe

#### 单继承语法 + super().__init__()

> **痛点**  
子类想扩展父类属性,又不想重复写父类的绑定逻辑。
直接写 父类.__init__(self, ...) 看起来能用,但层级一深就出错。

> **类比**  
儿子继承了父亲的房子(父类属性),再加盖一层(扩展属性),但必须**先**把父亲的房子盖好。

> **解释**  
`super()` 返回父类对象,用来**安全地调用父类方法**;写子类构造函数时,第一行就要 `super().__init__()`。

````python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # 先让父类把 name 绑好
        self.breed = breed      # 再扩展自己的

dog = Dog("旺财", "柯基")
print(dog.name, dog.breed)
dog2 = Dog("来福", "柴犬")
print(dog2.name)  # 来福(独立)
````

> **逐行解剖**  
① `class Animal` 定义父类 `__init__` 绑定 name  
② `class Dog(Animal)` 子类  
③ `super().__init__(name)` → 调用 Animal.__init__ → self.name 绑好  
④ `self.breed = breed` 绑子类自己的属性  
⑤ `dog = Dog('旺财', '柯基')` → Dog.__init__ → super() → Animal.__init__

> **逐行解剖**  
① `class Animal` 定义父类 `__init__` 绑定 name  
② `class Dog(Animal)` 子类  
③ `super().__init__(name)` → 调用 Animal.__init__ → self.name 绑好  
④ `self.breed = breed` 绑子类自己的属性

#### 方法重写(override)

> **痛点**  
父类的方法不能满足子类需要,比如基类 `shipping_cost()` 返回 0,
但实体书要运费,电子书不要。

> **类比**  
法律是父类,地方法规是子类:国家规定最低工资 A,地方可**重写**为更高的 B。

> **解释**  
子类重新定义同名方法,调用时**优先执行子类版本**。

> **逐行解剖**  
① `class Product` 基类,`shipping_cost()` 返回 0  
② `class PhysicalProduct(Product)` 子类  
③ `super().__init__(name, price)` 继承基类属性  
④ `def shipping_cost(self)` **重写** → 覆盖基类的 0  
⑤ `DigitalProduct` 不重写 → 继承基类的 0

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def shipping_cost(self):
        return 0  # 基类默认无运费

class PhysicalProduct(Product):
    def __init__(self, name, price, weight):
        super().__init__(name, price)
        self.weight = weight

    def shipping_cost(self):
        # 重写:按重量算运费
        return self.weight * 5

class DigitalProduct(Product):
    # 数字产品不需要重写,继承基类的 0
    pass

p = PhysicalProduct("纸质书", 50, 2)
d = DigitalProduct("电子书", 30)
print(p.shipping_cost())  # 10
print(d.shipping_cost())  # 0
````

#### MRO 方法解析顺序 + isinstance 反模式

> **痛点**  
多级继承时,属性/方法的查找顺序是什么?
需要运行时判断类型时,`isinstance` 是最佳选择吗?

> **类比**  
找人:先问自己 → 问父母 → 问祖父母,一层层往上找。

> **解释**  
- 查找顺序:子类 → 父类 → 父类的父类 → object
- `isinstance` 可以做类型判断,但**用它做"类型分发"是反模式**
- 每次新加类型都要改 if-elif → 违反"开闭原则"
- 正确做法:用多态(留到 Day07)

> **逐行解剖**  
① `class Animal` 基类  
② `class Mammal(Animal)` 一级子类  
③ `class Dog(Mammal)` 二级子类  
④ `Dog.__mro__` 查看查找链: Dog → Mammal → Animal → object  
⑤ `isinstance(dog, Animal)` → 沿继承链找 → True

````python
class Animal:
    pass

class Mammal(Animal):
    pass

class Dog(Mammal):
    pass

# MRO 查看
print(Dog.__mro__)
# (<class '__main__.Dog'>, <class '__main__.Mammal'>,
#  <class '__main__.Animal'>, <class 'object'>)

# isinstance
print(isinstance(Dog(), Dog))      # True
print(isinstance(Dog(), Mammal))   # True(继承链上也算)
print(isinstance(Dog(), Animal))   # True
````

#### 反模式:if-elif 类型分发(为何不好)

以下代码"能跑",但请找出 3 个坏味道:

```python
def pay(employee):
    if isinstance(employee, Manager):
        return employee.base + employee.bonus
    elif isinstance(employee, Sales):
        return employee.base + employee.commission
```

**问题**:  
1. 每新增一种员工类型,都要改 `pay` 函数  
2. `pay` 函数越来越长,难维护  
3. 违反"开闭原则"(对扩展开放,对修改关闭)  

> Day07 我们会用多态**一个表达式**替换整个 if-elif!

#### 综合练习:员工薪资系统

设计系统:
1. 基类 `Employee`:属性 `name`、`base_salary`,方法 `pay()` 返回 base_salary
2. 子类 `Sales`:新增 `commission`(提成),**重写** pay()
3. 子类 `Manager`:新增 `bonus`(奖金),**重写** pay()
4. 写函数 `show_pay(emp)` 传入任意员工,打印 pay()

> 提示:`show_pay` 体现的是 **多态**,今天先"能跑"即可,
> Day07 会把它改成无 if-elif 的版本。

````python
# ============ 学员代码区 ============
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def pay(self):
        return self.base_salary

# class Sales(Employee): ...
# class Manager(Employee): ...
# def show_pay(emp): ...
# employees = [Employee(...), Sales(...)]
# for e in employees:
#     show_pay(e)
pass

# ============ 参考答案 ============
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def pay(self):
        return self.base_salary

class Sales(Employee):
    def __init__(self, name, base, commission):
        super().__init__(name, base)
        self.commission = commission

    def pay(self):
        return super().pay() + self.commission

class Manager(Employee):
    def __init__(self, name, base, bonus):
        super().__init__(name, base)
        self.bonus = bonus

    def pay(self):
        return super().pay() + self.bonus

def show_pay(emp):
    if isinstance(emp, Employee):
        print(f"{emp.name} 薪资: {emp.pay()}")
    else:
        print("不是员工")

for e in [Employee("张三", 5000), Sales("李四", 5000, 2000)]:
    show_pay(e)
````
