**1. Matplotlib 四种基本图形**

`matplotlib.pyplot` 是 Python 数据可视化的基础库,提供多种基础图形。折线图看趋势,柱状图比大小,散点图看关系,直方图看分布,饼图看占比。


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ===== 折线图:看趋势 =====
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.title("折线图示例")
# plt.show()  # Jupyter中需%matplotlib inline;脚本中取消注释

# ===== 柱状图:比大小 =====
city = ["北京", "上海", "深圳", "广州"]
gdp = [4.1, 4.3, 3.2, 2.8]
plt.bar(city, gdp)
plt.title("柱状图示例")
# plt.show()

# ===== 散点图:看关系 =====
height = [160, 165, 170, 175, 180, 185]
weight = [55, 60, 65, 70, 75, 80]
plt.scatter(height, weight)
plt.title("散点图示例")
# plt.show()

# ===== 直方图:看分布 =====
ages = np.random.randint(18, 60, size=200)
plt.hist(ages, bins=20)
plt.title("直方图示例")
# plt.show()

# ===== 饼图:看占比 =====
sizes = [30, 25, 20, 15, 10]
labels = ["苹果", "香蕉", "橘子", "葡萄", "其他"]
plt.pie(sizes, labels=labels, autopct="%.1f%%")
plt.title("饼图示例")
# plt.show()
```

**2. 图表标注:标题/轴标签/图例/颜色**

一张图能不能读懂,全靠标注。`title` 标题、`xlabel/ylabel` 轴标签、`legend` 图例、`grid` 网格 —— 四大件缺一不可。


```python
import matplotlib.pyplot as plt
import numpy as np

# 两条折线数据
sales_a = [120, 135, 150, 165, 180]
sales_b = [100, 110, 140, 170, 200]
months = ["1月", "2月", "3月", "4月", "5月"]

# 画两条线:不同颜色、线型、标记
plt.plot(months, sales_a, color="r", linestyle="-",
         marker="o", label="产品A")
plt.plot(months, sales_b, color="b", linestyle="--",
         marker="s", label="产品B")

# 图表标注四大件
plt.title("销售额趋势对比", fontsize=16)
plt.xlabel("月份")
plt.ylabel("销售额(万元)")
plt.legend()
plt.grid(True)
# plt.show()

# 常用参数速查:
# color: r(红) g(绿) b(蓝) k(黑)
# linestyle: - 实线 -- 虚线 -. 点划线 : 点线
# marker: o 圆点 s 方块 * 星形
```

**3. 子图 subplots**

多张图并排时,用 `plt.subplots(rows, cols)` 创建网格布局,返回 `fig` 和 `axes` 数组,用 `axes[i,j]` 定位每个子图。


```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 2x2 子图网格
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 子图1:折线
axes[0, 0].plot([1, 2, 3], [1, 4, 9])
axes[0, 0].set_title("折线")

# 子图2:柱状
axes[0, 1].bar(["A", "B", "C"], [5, 3, 7])
axes[0, 1].set_title("柱状")

# 子图3:散点
axes[1, 0].scatter([1, 2, 3], [1, 2, 3])
axes[1, 0].set_title("散点")

# 子图4:直方
axes[1, 1].hist(np.random.randn(100), bins=10)
axes[1, 1].set_title("直方")

# 自动调整子图间距
plt.tight_layout()
# plt.show()
```

**4. Seaborn 高级图表**

`seaborn` 基于 matplotlib 封装,默认样式更美观,一行代码即可绘制统计图表。常用:热力图(相关性)、箱线图、pairplot。


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 设置样式
sns.set_theme(style="whitegrid")

# ===== 箱线图:数据分布 =====
tips = pd.DataFrame({
    "day": ["Sun", "Sun", "Sat", "Sat", "Fri", "Fri",
            "Thur", "Thur"],
    "total_bill": [17, 21, 24, 28, 16, 19, 15, 18]
})
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("箱线图示例")
# plt.show()

# ===== 热力图:相关性 =====
df = pd.DataFrame({
    "数学": [90, 80, 70, 85],
    "语文": [85, 75, 80, 90],
    "英语": [70, 85, 90, 75]
})
corr = df.corr()  # 计算相关系数矩阵
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("热力图示例")
# plt.show()

# ===== pairplot:多变量关系 =====
iris = pd.DataFrame({
    "sepal_length": [5.1, 4.9, 5.8, 6.2, 5.0, 5.5],
    "sepal_width":  [3.5, 3.0, 3.7, 3.4, 2.3, 2.6],
    "petal_length": [1.4, 1.4, 4.7, 5.4, 3.4, 4.2],
    "species": ["setosa", "setosa", "versicolor",
               "virginica", "versicolor", "versicolor"]
})
sns.pairplot(iris, hue="species")
# plt.show()
```

**5. 图表导出 savefig**

图表要放进报告时,用 `fig.savefig` 保存为图片文件。**务必在 `plt.show()` 之前调用**,因为 `show()` 会清空画布。


```python
import matplotlib.pyplot as plt

# 创建图和坐标轴
fig, ax = plt.subplots()

# 画图
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_title("导出示例")

# 保存为图片:dpi 控制清晰度,bbox_inches 控制白边
fig.savefig("output.png", dpi=300, bbox_inches="tight")
# plt.show()  # show 会清空画布,必须在 savefig 之前调用
print("图表已保存为 output.png")
```

| pairplot | `sns.pairplot` | 多变量关系 |
