**4. 子图 subplots**多张图并排时,用 plt.subplots(rows, cols) 创建网格布局,返回 fig 和 axes 数组。**推荐用 axes[i,j] 而非 axes[i][j]** 
定位每个子图。最后用 tight_layout() 自动调整间距。

**5. Seaborn 高级图表**seaborn 基于 matplotlib 封装,默认样式更美观,一行代码即可绘制统计图表。Seaborn 的函数**直接接受 DataFrame**,用 x、y、data 
参数指定列名,非常方便。常用:**箱线图**(数据分布)、**热力图**(相关性)、**pairplot**(多变量关系)。


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False# 设置 Seaborn 全局样式sns.set_theme(style="whitegrid")# ===== 箱线图:数据分布 =====tips = pd.DataFrame({    "day": ["Sun", "Sun", "Sat", "Sat", "Fri", "Fri", "Thur", "Thur"],    "total_bill": [17, 21, 24, 28, 16, 19, 15, 18]})sns.boxplot(x="day", y="total_bill", data=tips)plt.title("箱线图示例")# plt.show()# ===== 热力图:相关性 =====df = pd.DataFrame({    "数学": [90, 80, 70, 85],    "语文": [85, 75, 80, 90],    "英语": [70, 85, 90, 75]})corr = df.corr()  # 计算相关系数矩阵sns.heatmap(corr, annot=True, cmap="coolwarm")plt.title("热力图示例")# plt.show()# ===== pairplot:多变量关系 =====iris = pd.DataFrame({    "sepal_length": [5.1, 4.9, 5.8, 6.2, 5.0, 5.5],    "sepal_width":  [3.5, 3.0, 3.7, 3.4, 2.3, 2.6],    "petal_length": [1.4, 1.4, 4.7, 5.4, 3.4, 4.2],    "species": ["setosa", "setosa", "versicolor",               "virginica", "versicolor", "versicolor"]})sns.pairplot(iris, hue="species")# plt.show()
```

**7. 综合示例:销售数据可视化报告**综合运用本节所学,用一张 2x2 子图展示某公司的销售数据分析报告:折线图看趋势、柱状图比大小、箱线图看分布、热力图看相关性。

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False# 构造销售数据df = pd.DataFrame({    "月份": ["1月", "2月", "3月", "4月", "5月", "6月"],    "销售额": [120, 135, 150, 165, 180, 200],    "利润": [20, 25, 30, 35, 42, 50],    "成本": [100, 110, 120, 130, 138, 150]})fig, axes = plt.subplots(2, 2, figsize=(12, 9))# 子图1:销售额趋势(折线)axes[0, 0].plot(df["月份"], df["销售额"], marker="o")axes[0, 0].set_title("销售额趋势")axes[0, 0].grid(True)# 子图2:销售额 vs 成本(柱状)x_pos = np.arange(len(df))width = 0.35axes[0, 1].bar(x_pos - width/2, df["销售额"], width, label="销售额")axes[0, 1].bar(x_pos + width/2, df["成本"], width, label="成本")axes[0, 1].set_xticks(x_pos)axes[0, 1].set_xticklabels(df["月份"])axes[0, 1].set_title("销售额 vs 成本")axes[0, 1].legend()# 子图3:利润分布(箱线图)sns.boxplot(y=df["利润"], ax=axes[1, 0])axes[1, 0].set_title("利润分布")# 子图4:指标相关性(热力图)corr = df[["销售额", "利润", "成本"]].corr()sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[1, 1])axes[1, 1].set_title("指标相关性")plt.tight_layout()fig.savefig("report.png", dpi=300, bbox_inches="tight")# plt.show()print("报告已保存为 report.png")
```
