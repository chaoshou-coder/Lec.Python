### 2.1 图表标注: title/label/legend📖 通过标注参数让图表更清晰可读。

```python
import matplotlib.pyplot as pltimport numpy as npx = np.linspace(0, 10, 100)y1 = x ** 2  # 平方曲线y2 = x * 3   # 线性曲线plt.figure(figsize=(8, 5))# 绘制两条不同样式的线plt.plot(x, y1, label="平方", color="red",         linestyle="--")plt.plot(x, y2, label="线性", color="blue",         marker="o", markevery=10)# 添加标注plt.title("平方 vs 线性", fontsize=16)plt.xlabel("X 轴", fontsize=12)plt.ylabel("Y 轴", fontsize=12)plt.legend()  # 显示图例plt.grid(True, alpha=0.3)# plt.show()  # 实际环境中需取消注释print("📊 [此处显示带标注的图表]")
```

### 2.2 subplots 子图布局💡 `plt.subplots` 创建网格,一画布多子图。

```python
import matplotlib.pyplot as pltimport numpy as np# 创建 2×2 子图网格fig, axes = plt.subplots(2, 2, figsize=(10, 8))x = np.linspace(0, 2 * np.pi, 100)# 子图 1: 折线图axes[0, 0].plot(x, np.sin(x), color="blue")axes[0, 0].set_title("折线图")# 子图 2: 柱状图axes[0, 1].bar(["A", "B", "C"], [12, 19, 7],               color="tomato")axes[0, 1].set_title("柱状图")# 子图 3: 散点图np.random.seed(42)axes[1, 0].scatter(np.random.rand(30),                   np.random.rand(30),                   color="green")axes[1, 0].set_title("散点图")# 子图 4: 饼图axes[1, 1].pie([3, 2, 1], labels=["X", "Y", "Z"],               autopct="%1.0f%%")axes[1, 1].set_title("饼图")plt.tight_layout()  # 自动调整间距# plt.show()  # 实际环境中需取消注释print("📊 [此处显示 2×2 子图网格]")
```

### 3.3 图表导出 savefig💡 使用 `fig.savefig()` 保存图表为图片文件。

```python
import matplotlib.pyplot as pltimport numpy as npx = np.linspace(0, 10, 50)y = np.sin(x)fig, ax = plt.subplots(figsize=(6, 4))ax.plot(x, y, color="steelblue")ax.set_title("sine 曲线")ax.set_xlabel("x")ax.set_ylabel("sin(x)")# 保存图表(高分辨率)fig.savefig("output.png", dpi=300,            bbox_inches="tight")# plt.show()  # 实际环境中需取消注释print("📊 图表已保存为 output.png")
```

> 🔴 **易错点**: `savefig` 放在 `show()` 之前!因为 `show()` 执行后会清空画布,此时再 `savefig` 只会保存一张空白图。
