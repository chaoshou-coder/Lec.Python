### 1.2 四种常用图形
Matplotlib 支持多种常用图形,下面我们逐一演示:

```python
import matplotlib.pyplot as plt
import numpy as np

# ========== 1. 柱状图: 各城市 GDP ==========
cities = ["北京", "上海", "广州", "深圳"]
gdp = [4.03, 4.32, 2.88, 3.24]  # 万亿元

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.bar(cities, gdp, color="steelblue")
plt.title("各城市 GDP(万亿元)")
plt.ylabel("GDP")

# ========== 2. 散点图: 身高 vs 体重 ==========
np.random.seed(42)
height = np.random.normal(170, 8, 50)
weight = height * 0.4 + np.random.normal(0, 5, 50)

plt.subplot(2, 2, 2)
plt.scatter(height, weight, alpha=0.7, color="tomato")
plt.title("身高 vs 体重 散点图")
plt.xlabel("身高(cm)")
plt.ylabel("体重(kg)")

# ========== 3. 直方图: 年龄分布 ==========
ages = np.random.normal(30, 10, 200).astype(int)
ages = ages[(ages >= 0) & (ages <= 80)]

plt.subplot(2, 2, 3)
plt.hist(ages, bins=15, color="mediumseagreen",
         edgecolor="black")
plt.title("年龄分布直方图")
plt.xlabel("年龄")
plt.ylabel("人数")

# ========== 4. 饼图: 编程语言占比 ==========
langs = ["Python", "Java", "C++", "JavaScript", "其他"]
ratio = [35, 25, 15, 15, 10]
colors = ["#ff9999", "#66b3ff", "#99ff99",
          "#ffcc99", "#ff99cc"]

plt.subplot(2, 2, 4)
plt.pie(ratio, labels=langs, colors=colors,
        autopct="%1.1f%%", startangle=90)
plt.title("编程语言使用占比")

plt.tight_layout()
# 实际环境中此行需执行以显示图表
# plt.show()
print("📊 [此处显示四种常用图表:",
      "柱状图/散点图/直方图/饼图]")
```

## 第二讲: 图表标注与子图

### 2.1 图表标注
通过 legend/title/color/linestyle/marker 等参数,
可以让图表更清晰、更具可读性。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 5))

# 绘制三条不同样式的线
plt.plot(x, np.sin(x), color="blue",
         linestyle="-", marker="",
         label="sin(x)")
plt.plot(x, np.cos(x), color="red",
         linestyle="--", marker="",
         label="cos(x)")
plt.plot(x, np.sin(x) + np.cos(x),
         color="green", linestyle=":",
         marker="o", markevery=10,
         label="sin(x)+cos(x)")

# 标注
plt.title("三角函数示例", fontsize=16)
plt.xlabel("x 轴", fontsize=12)
plt.ylabel("y 轴", fontsize=12)
plt.legend(loc="upper right")
plt.grid(True, alpha=0.3)

# 实际环境中此行需执行以显示图表
# plt.show()
print("📊 [此处显示带标注的图表]")
```

### 2.2 subplots 子图
使用 `plt.subplots` 创建网格布局,
在一个画布上绘制多个子图。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 2×2 子图网格
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 2 * np.pi, 100)

# 子图 1: 折线图
axes[0, 0].plot(x, np.sin(x), color="blue")
axes[0, 0].set_title("折线图")

# 子图 2: 柱状图
axes[0, 1].bar(["A", "B", "C"], [12, 19, 7],
               color="tomato")
axes[0, 1].set_title("柱状图")

# 子图 3: 散点图
axes[1, 0].scatter(np.random.rand(30),
                   np.random.rand(30),
                   color="green")
axes[1, 0].set_title("散点图")

# 子图 4: 饼图
axes[1, 1].pie([3, 2, 1], labels=["X", "Y", "Z"],
               autopct="%1.0f%%")
axes[1, 1].set_title("饼图")

plt.tight_layout()
# 实际环境中此行需执行以显示图表
# plt.show()
print("📊 [此处显示 2×2 子图网格]")
```

## 第三讲: Seaborn 高级图表

### 3.1 Seaborn 简介
Seaborn 是建立在 Matplotlib 之上的高阶库,
专注统计图表,默认样式更美观。

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 设置主题样式
sns.set_theme(style="whitegrid")

# ===== boxplot: tips 数据集 =====
tips = sns.load_dataset("tips")
plt.figure(figsize=(6, 4))
sns.boxplot(x="day", y="total_bill",
            data=tips)
plt.title("Tips 箱线图")
# plt.show()  # 实际环境中此行需执行以显示图表

# ===== heatmap: 相关性矩阵 =====
import numpy as np
import pandas as pd

np.random.seed(0)
df = pd.DataFrame(np.random.randn(100, 4),
                  columns=["A", "B", "C", "D"])
corr = df.corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("相关性热力图")
# plt.show()  # 实际环境中此行需执行以显示图表

# ===== pairplot: iris 数据集 =====
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
# plt.show()  # 实际环境中此行需执行以显示图表

print("📊 [此处显示 Seaborn 图表:",
      "boxplot/heatmap/pairplot]")
```

### 3.2 图表导出 savefig
使用 `fig.savefig()` 将图表保存为图片文件,
常用于报告生成。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y)
ax.set_title("sine 曲线")

# 保存图表
fig.savefig("output.png", dpi=300,
            bbox_inches="tight")
# plt.show()  # 实际环境中此行需执行以显示图表
print("图表已保存为 output.png")
```

> 🔴 易错点: savefig 放在 `show()` 之前!
>
因为 `show()` 执行后会清空画布,
此时再 `savefig` 只会保存一张空白图。

## 总结

| 库 | 用途 | 常用图表 |
|---|---|---|
| Matplotlib | 基础绘图 | 折线/柱状/散点/直方/饼图 |
| Matplotlib | 标注 | legend/title/color/marker |
| Matplotlib | 布局 | subplots/savefig |
| Seaborn | 高级统计图 | boxplot/heatmap/pairplot |

**今日核心要点:**
- `plt.plot/bar/scatter/hist/pie` 对应五种基础图形
- `plt.subplots` 快速创建子图网格
- `sns.boxplot/heatmap/pairplot` 绘制高级统计图
- `fig.savefig()` 必须放在 `plt.show()` **之前**

## ⭐ 挑战题
综合可视化 — Titanic 数据集的探索:
- 年龄分布直方图
- 各舱位生存率柱状图
- 数值特征相关性热力图

要求通过 3 个子图展示在同一画布中。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 加载 Titanic 数据集
titanic = sns.load_dataset("titanic")

# ======================
# 学员代码区
# 1. 创建 1×3 子图
# 2. 画出年龄分布直方图
# 3. 画出舱位生存率柱状图
# 4. 画出相关性热力图
# ======================
pass

# plt.show()  # 实际环境中此行需执行以显示图表
```

```python
# 参考答案
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic = sns.load_dataset("titanic")

fig, axes = plt.subplots(1, 3,
                         figsize=(15, 4))

# 1. 年龄分布直方图
axes[0].hist(titanic["age"].dropna(),
             bins=20, color="steelblue",
             edgecolor="black")
axes[0].set_title("乘客年龄分布")
axes[0].set_xlabel("年龄")
axes[0].set_ylabel("人数")

# 2. 各舱位生存率柱状图
surv_rate = (titanic.groupby("class")
             ["survived"].mean())
axes[1].bar(surv_rate.index, surv_rate.values,
            color=["gold", "silver", "chocolate"])
axes[1].set_title("各舱位生存率")
axes[1].set_xlabel("舱位等级")
axes[1].set_ylabel("生存率")
axes[1].set_ylim(0, 1)

# 3. 数值特征相关性热力图
num_cols = titanic.select_dtypes("number")
corr = num_cols.corr()
im = axes[2].imshow(corr, cmap="coolwarm",
                    vmin=-1, vmax=1)
axes[2].set_xticks(range(len(corr)))
axes[2].set_yticks(range(len(corr)))
axes[2].set_xticklabels(corr.columns,
                         rotation=45)
axes[2].set_yticklabels(corr.columns)
axes[2].set_title("数值特征相关性")
fig.colorbar(im, ax=axes[2])

plt.tight_layout()
# plt.show()  # 实际环境中此行需执行以显示图表
```
