## 什么是 Jupyter Notebook?

Jupyter Notebook 是一个**交互式编程环境**,
你可以把它想象成一个"电子实验本":

- 📝 可以在代码旁边写笔记(Markdown)
- ▶️ 可以**逐段运行代码**,立即看到结果
- 📊 可以直接显示图表、表格
- 💾 可以保存代码+笔记+输出为一个文件(.ipynb)

## 为什么用 Jupyter Notebook 学 Python?

| 方式 | 特点 |
|---|---|
| 终端 `python script.py` | 一次运行整个脚本,看不到中间结果 |
| IDE (VSCode/PyCharm) | 功能强大但复杂,初学者易迷失 |
| **Jupyter Notebook** | **逐段运行,即时反馈,笔记+代码一体** ✅ |

## Cell:Notebook 的基本单元

**Cell** 是 Notebook 的"单元格",每个 Cell 可以是两种类型:

| 类型 | 用途 | 快捷键 |
|---|---|---|
| **Code** | 写 Python 代码,可以运行出结果 | `Y` 切换为 Code |
| **Markdown** | 写笔记、标题、说明文字 | `M` 切换为 Markdown |

### 创建 Cell

- 点击工具栏的 **`+`** 按钮
- 或按快捷键 **`B`** (在下方插入) / **`A`** (在上方插入)

### 编辑 Cell

- **单击** Cell → 进入编辑模式(光标闪烁)
- **双击** Markdown Cell → 查看源码

### 运行 Cell

- **`Shift + Enter`**:运行当前 Cell,并跳到下一个
  (没有下一个则新建)
- **`Ctrl + Enter`**:运行当前 Cell,停留在当前
- **`Alt + Enter`**:运行当前 Cell,在下方新建一个

````python
# 试试运行这个 Cell!
# 1. 单击这个 Cell(进入编辑模式)
# 2. 按 Shift + Enter 运行
print("🎉 恭喜!你成功运行了第一个 Jupyter Cell!")
````

### 删除 Cell

- 选中 Cell → 按 **`D`** 两次(或 `Edit → Delete Cells`)

## ⚡ 快捷键速查表

> 快捷键是提高效率的关键!先记住最常用的 5 个,
> 其余慢慢熟悉。

### 命令模式(按 `Esc` 进入,Cell 边框变蓝)

| 快捷键 | 功能 |
|---|---|

| `↑` / `↓` | 上下选择 Cell |
| `Enter` | 进入编辑模式 |

### 编辑模式(按 `Enter` 进入,Cell 边框变绿)

| 快捷键 | 功能 |
|---|---|
| `Shift + Enter` | 运行并跳到下一个 ⭐最常用 |
| `Ctrl + Enter` | 运行并停留 |
| `Alt + Enter` | 运行并在下方新建 |
| `Esc` | 回到命令模式 |

## 🔄 Kernel 操作

Kernel 是运行 Python 代码的"引擎"。
有时需要重启或中断:

| 操作 | 菜单路径 | 何时用 |
|---|---|---|
| **Restart Kernel** | Kernel → Restart | 代码卡住/变量混乱时 |
| **Interrupt Kernel** | Kernel → Interrupt | 代码死循环/运行太久 |
| **Run All Cells** | Cell → Run All | 一次性运行整个 Notebook |
| **Restart & Run All** | Kernel → Restart & Run All | 从头重新运行整个 Notebook |

## ❓ 常见问题排查

### Q1: 运行 Cell 没有输出?

**A**: 检查:
1. 是不是 Markdown Cell?(Markdown 不会执行代码)
2. 按 `Y` 切换为 Code Cell
3. 确认代码没有语法错误

````python
# 测试:运行这个 Cell,应该看到数字 42
42  # 在 Jupyter 中,最后一个表达式的值会自动显示
````

### Q2: 变量在后面的 Cell 中找不到?

**A**: Jupyter 的变量是**按运行顺序**记忆的,
不是按 Cell 位置!

- 如果 Cell 1 定义了 `x = 10`,必须**先运行 Cell 1**,
  Cell 2 才能用 `x`
- 不确定时,按 **Kernel → Restart & Run All** 从头运行

````python
# 演示:先运行下一个 Cell,再运行这个 Cell,观察报错
# 然后 Restart & Run All,再观察
print(f"x = {x}")
````

````python
# 先运行这个 Cell
x = 42
````

### Q3: Jupyter 启动后浏览器没打开?

**A**: 手动在浏览器输入终端中显示的 URL
(通常是 `http://localhost:8888/?token=xxxxx`)

### Q4: 代码一直运行不出结果(死循环)?

**A**: 按工具栏的 **⏹(Stop)** 按钮,
或菜单 **Kernel → Interrupt**

### Q5: 如何关闭 Jupyter?

**A**:
1. 在浏览器中关闭标签页(这不会关闭 Jupyter)
2. 回到终端,按 **`Ctrl + C`** 两次 → 真正关闭
